```python
import ipywidgets as widgets
from IPython.display import display, clear_output

all_quizzes = {
    "19_1_1: Anatomy of a Network": [
        {
            "question": "What is the nn.Module contract \u2014 the two things every PyTorch model class defines?",
            "options": [
                "__init__ defines and registers the layers; forward(x) defines the computation",
                "fit() trains the model; predict() makes predictions",
                "encode() compresses the input; decode() reconstructs it",
                "train() runs epochs; eval() reports metrics"
            ],
            "answer": "__init__ defines and registers the layers; forward(x) defines the computation"
        },
        {
            "question": "What does nn.Linear(2, 4) contain, and what does its forward computation do?",
            "options": [
                "A weight matrix of shape (4, 2) and a bias of shape (4,); it computes input @ weight.T + bias",
                "A weight matrix of shape (2, 4) and no bias; it computes weight @ input",
                "Two neurons with four activations each",
                "Four copies of a 2-element input"
            ],
            "answer": "A weight matrix of shape (4, 2) and a bias of shape (4,); it computes input @ weight.T + bias"
        },
        {
            "question": "Why can XORNet (Linear \u2192 ReLU \u2192 Linear \u2192 sigmoid) solve XOR when logistic regression cannot?",
            "options": [
                "The hidden layer learns a transformation of the inputs in which the classes become linearly separable",
                "It has a lower learning rate than logistic regression",
                "BCELoss is more powerful than the loss logistic regression uses",
                "The sigmoid output allows probabilities instead of hard labels"
            ],
            "answer": "The hidden layer learns a transformation of the inputs in which the classes become linearly separable"
        },
        {
            "question": "When should you write a custom nn.Module class instead of using nn.Sequential?",
            "options": [
                "When the forward pass needs custom logic \u2014 branching, multiple inputs, skip connections \u2014 rather than a straight pipeline of layers",
                "Whenever the network has more than two layers",
                "When you want the layers to be automatically named by index",
                "nn.Sequential is deprecated, so always use a class"
            ],
            "answer": "When the forward pass needs custom logic \u2014 branching, multiple inputs, skip connections \u2014 rather than a straight pipeline of layers"
        }
    ],
    "19_1_2: The Training Loop": [
        {
            "question": "With 341 training samples and batch_size=32, how many parameter updates happen per epoch?",
            "options": [
                "11 \u2014 one per batch (ceil(341/32))",
                "1 \u2014 one per epoch",
                "341 \u2014 one per sample",
                "32 \u2014 one per batch element"
            ],
            "answer": "11 \u2014 one per batch (ceil(341/32))"
        },
        {
            "question": "What does the combination model.eval() + torch.no_grad() accomplish in the validation phase?",
            "options": [
                "eval() switches layers like Dropout/BatchNorm to evaluation behavior; no_grad() skips computation-graph building since no backward pass is needed",
                "eval() freezes the weights permanently; no_grad() deletes the gradients",
                "Both calls do the same thing \u2014 one is a backup for the other",
                "They convert the model's outputs from logits to probabilities"
            ],
            "answer": "eval() switches layers like Dropout/BatchNorm to evaluation behavior; no_grad() skips computation-graph building since no backward pass is needed"
        },
        {
            "question": "In the deliberately oversized (256, 256) network, training loss collapsed to ~0 while validation loss bottomed out around epoch 12 and then climbed. What is this pattern called?",
            "options": [
                "Overfitting \u2014 the model is memorizing training noise instead of learning generalizable patterns",
                "Underfitting \u2014 the model is too simple for the data",
                "Vanishing gradients \u2014 the backward pass is failing",
                "Convergence \u2014 the model has found the global minimum"
            ],
            "answer": "Overfitting \u2014 the model is memorizing training noise instead of learning generalizable patterns"
        },
        {
            "question": "Why does the three-way split reserve a test set that is never touched during training or model selection?",
            "options": [
                "The validation set influences decisions (architecture, epochs), making it slightly optimistic; only data unseen at every stage gives an unbiased estimate",
                "The test set is needed to compute the class weights",
                "Three splits triple the amount of training data",
                "The DataLoader requires three datasets to shuffle correctly"
            ],
            "answer": "The validation set influences decisions (architecture, epochs), making it slightly optimistic; only data unseen at every stage gives an unbiased estimate"
        }
    ],
    "19_1_3: Multiclass Classification": [
        {
            "question": "What is the most common mistake when using nn.CrossEntropyLoss, and why is it silent?",
            "options": [
                "Applying softmax in the model's forward method \u2014 the loss applies softmax internally, so it gets applied twice and the gradients are quietly wrong",
                "Forgetting to call .backward() \u2014 the loss is then never computed",
                "Using integer targets \u2014 CrossEntropyLoss requires one-hot vectors",
                "Using more than two classes \u2014 the loss only supports binary problems"
            ],
            "answer": "Applying softmax in the model's forward method \u2014 the loss applies softmax internally, so it gets applied twice and the gradients are quietly wrong"
        },
        {
            "question": "What target format does PyTorch's CrossEntropyLoss expect?",
            "options": [
                "Integer class labels (torch.long) of shape (n,) \u2014 not one-hot vectors",
                "One-hot encoded float vectors of shape (n, K)",
                "Float probabilities of shape (n, 1)",
                "String class names"
            ],
            "answer": "Integer class labels (torch.long) of shape (n,) \u2014 not one-hot vectors"
        },
        {
            "question": "How does CrossEntropyLoss(weight=class_weights) help with class imbalance?",
            "options": [
                "Each sample's loss is multiplied by its true class's weight, so rare classes contribute more gradient signal during training",
                "It removes majority-class samples until the classes are balanced",
                "It raises the decision threshold for the majority class",
                "It duplicates minority-class samples inside each batch"
            ],
            "answer": "Each sample's loss is multiplied by its true class's weight, so rare classes contribute more gradient signal during training"
        },
        {
            "question": "How do you obtain the same N\u00d7K probability matrix from a PyTorch multiclass model that predict_proba() gave you in XGBoost?",
            "options": [
                "torch.softmax(model(X), dim=1) \u2014 softmax over the K logits of each row",
                "model(X) directly \u2014 the forward pass already returns probabilities",
                "model.predict_proba(X) \u2014 the method exists on every nn.Module",
                "torch.sigmoid(model(X)) \u2014 sigmoid maps each logit to a probability"
            ],
            "answer": "torch.softmax(model(X), dim=1) \u2014 softmax over the K logits of each row"
        }
    ]
}

class QuizApp:
    def __init__(self, quiz_name, data):
        self.quiz_name = quiz_name
        self.data = data
        self.current_index = 0
        self.score = 0

        self.header = widgets.HTML(value="<h2>" + quiz_name + "</h2>")
        self.progress = widgets.HTML(value="")
        self.question_text = widgets.HTML(value="")
        self.options = widgets.RadioButtons(options=[], layout={"width": "max-content"})
        self.submit_btn = widgets.Button(description="Submit Answer", button_style="primary")
        self.next_btn = widgets.Button(description="Next Question", button_style="info", disabled=True)
        self.back_btn = widgets.Button(description="Back to Menu", button_style="warning")
        self.feedback = widgets.HTML(value="")

        self.submit_btn.on_click(self.handle_submit)
        self.next_btn.on_click(self.handle_next)
        self.back_btn.on_click(self.handle_back)
        self.render_question()

    def render_question(self):
        clear_output(wait=True)
        q = self.data[self.current_index]
        self.progress.value = "<b>Question " + str(self.current_index + 1) + " of " + str(len(self.data)) + "</b>"
        self.question_text.value = "<p style='font-size:14px;'>" + q["question"] + "</p>"
        self.options.options = q["options"]
        self.options.value = None
        self.submit_btn.disabled = False
        self.next_btn.disabled = True
        self.feedback.value = ""
        display(self.header, self.progress, self.question_text, self.options,
                widgets.HBox([self.submit_btn, self.next_btn]), self.feedback)

    def handle_submit(self, b):
        if self.options.value is None:
            self.feedback.value = "<span style='color:red;'>Please select an option!</span>"
            return
        correct = self.data[self.current_index]["answer"]
        if self.options.value == correct:
            self.score += 1
            self.feedback.value = "<span style='color:green; font-weight:bold;'>Correct!</span>"
        else:
            self.feedback.value = "<span style='color:red; font-weight:bold;'>Incorrect.</span><br>The correct answer was: <b>" + correct + "</b>"
        self.submit_btn.disabled = True
        self.next_btn.disabled = False

    def handle_next(self, b):
        self.current_index += 1
        if self.current_index < len(self.data):
            self.render_question()
        else:
            self.show_results()

    def handle_back(self, b):
        show_menu()

    def show_results(self):
        clear_output()
        pct = (self.score / len(self.data)) * 100
        html = """<div style="border: 2px solid #0366d6; padding: 20px; border-radius: 10px;">
            <h2>Quiz Complete!</h2>
            <p style="font-size:16px;">Your Final Score: <b>""" + str(self.score) + " / " + str(len(self.data)) + """</b></p>
            <p style="font-size:20px; color:#0366d6;"><b>Percentage: """ + "{:.1f}".format(pct) + "%</b></p></div>"
        display(widgets.HTML(html), self.back_btn)

def show_menu():
    clear_output(wait=True)
    header = widgets.HTML(value="<h2>19_1 Neural Networks — Practice Quizzes</h2>")
    instruction = widgets.HTML(value="<p>Select a notebook quiz below to begin.</p>")
    quiz_selector = widgets.Dropdown(options=list(all_quizzes.keys()), description="Select Quiz:", layout={"width": "50%"})
    start_btn = widgets.Button(description="Start Quiz", button_style="success")

    def on_start(b):
        QuizApp(quiz_selector.value, all_quizzes[quiz_selector.value])

    start_btn.on_click(on_start)
    display(header, instruction, quiz_selector, start_btn)

show_menu()
```
