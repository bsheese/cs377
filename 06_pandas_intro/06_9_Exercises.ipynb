### Exercise 5 — Adding a Computed Column

## Part 5 — GroupBy and Aggregation

*Covers: 06.5*

### Exercise 12 — Average Mass by Species

Compute the average body mass for each species. Sort the result from lightest to heaviest.

Which species is the heaviest? By how much does it exceed the others?

```
# Your code here
```

```
#@title 💡 Solution — Exercise 12 {display-mode: "form"}
avg_mass = (
    df.groupby("species")["body_mass_g"]
    .mean()
    .round(1)
    .sort_values()
)
print(avg_mass)
print()
print(f"Gentoo are by far the heaviest: {avg_mass['Gentoo']:.0f} g")
print(f"That is {avg_mass['Gentoo'] - avg_mass['Adelie']:.0f} g heavier than Adelie")
print(f"and {avg_mass['Gentoo'] - avg_mass['Chinstrap']:.0f} g heavier than Chinstrap.")
print()
print("Adelie and Chinstrap are nearly equal in mass (~3700 g).")
print("Gentoo are about 1375 g heavier — 37% more mass. A meaningful biological difference.")
```

### Exercise 13 — Multiple Aggregations by Two Keys

Group by both `species` and `sex`. Compute:
- The count of penguins in each group
- The average body mass
- The average flipper length

Which species-sex combination is the heaviest? The lightest?

```
# Your code here
```

```
#@title 💡 Solution — Exercise 13 {display-mode: "form"}
summary = (
    df.groupby(["species", "sex"])
    .agg(
        count          = ("body_mass_g",      "count"),
        avg_mass_g     = ("body_mass_g",      "mean"),
        avg_flipper_mm = ("flipper_length_mm", "mean"),
    )
    .round(1)
)
print(summary)
print()
print("Heaviest: Gentoo MALE  — avg 5484.8 g")
print("Lightest: Adelie FEMALE — avg 3368.8 g")
print()
print("Within every species, males are heavier than females by ~600–800g.")
print("This sexual dimorphism is consistent across all three species.")
```

### Exercise 14 — GroupBy with a Twist

Which **island** has the most penguins? Which has the highest average body mass?

Compute count and average body mass grouped by island, sorted by count descending.

Can you explain why one island might have heavier penguins on average? *(Hint: think about which species lives on which island.)*

```
# Your code here
```

```
#@title 💡 Solution — Exercise 14 {display-mode: "form"}
island_summary = (
    df.groupby("island")
    .agg(
        count    = ("body_mass_g", "count"),
        avg_mass = ("body_mass_g", "mean"),
    )
    .round(1)
    .sort_values("count", ascending=False)
)
print(island_summary)
print()
print("Biscoe has the most penguins (168) AND the highest average mass (4716g).")
print()
print("Why? Because Biscoe is home to all 124 Gentoo penguins in this dataset,")
print("and Gentoo are by far the heaviest species (~5076g average).")
print("The island's average mass is driven almost entirely by species composition,")
print("not by anything about Biscoe itself — a classic example of Simpson's Paradox risk.")
```

### Exercise 16 — Pivot Table: Body Mass by Species and Sex

Build a pivot table showing the **average body mass** for each species (rows) and sex (columns).

Then add `margins=True` to see the overall averages.

What is the gender mass gap within each species?

```
# Your code here
```

```
#@title 💡 Solution — Exercise 16 {display-mode: "form"}
pt = pd.pivot_table(
    df,
    values       = "body_mass_g",
    index        = "species",
    columns      = "sex",
    aggfunc      = "mean",
    margins      = True,
    margins_name = "Overall",
).round(1)
print(pt)
print()
print("Gender mass gap within each species:")
for sp in ["Adelie", "Chinstrap", "Gentoo"]:
    female = pt.loc[sp, "FEMALE"]
    male   = pt.loc[sp, "MALE"]
    print(f"  {sp}: males are {male - female:.0f}g heavier ({(male-female)/female:.0%} more)")
print()
print("The gap is consistent (~600–800g) and proportionally similar across species.")
print("Gentoo have the largest absolute gap because they are the largest species overall.")
```

### Exercise 17 — Correlation: Which Measurements Go Together?

Compute the correlation matrix for the four numeric measurement columns:
`bill_length_mm`, `bill_depth_mm`, `flipper_length_mm`, `body_mass_g`.

1. Which two columns are most strongly positively correlated?
2. Which pair has a **negative** correlation? Why might that be?
3. Does a strong correlation mean one measurement *causes* the other?

```
# Your code here
```

```
#@title 💡 Solution — Exercise 17 {display-mode: "form"}
numeric_cols = ["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]
corr = df[numeric_cols].corr().round(2)
print(corr)
print()
print("Strongest positive correlation: flipper_length_mm ↔ body_mass_g (r = 0.87)")
print("  Heavier penguins tend to have longer flippers — both reflect overall body size.")
print()
print("Negative correlation: bill_depth_mm ↔ bill_length_mm (r = -0.24)")
print("  and bill_depth_mm ↔ body_mass_g (r = -0.47)")
print()
print("This seems counterintuitive — why would deeper bills go with smaller penguins?")
print("The answer is species composition: Adelie penguins have short, DEEP bills and")
print("are small. Gentoo have long, NARROW bills and are large. The species structure")
print("creates a negative correlation in the combined dataset that doesn't exist within")
print("any single species. This is another instance of Simpson's Paradox.")
print()
print("Correlation does not imply causation — longer flippers don't cause heavier mass.")
print("Both are caused by a third variable: the penguin's overall size.")
```
