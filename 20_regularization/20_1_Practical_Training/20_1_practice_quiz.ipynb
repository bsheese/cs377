```python
import ipywidgets as widgets
from IPython.display import display, clear_output

all_quizzes = {
    "20_1_1: Dropout": [
        {
            "question": "Where does nn.Dropout belong in a network block?",
            "options": [
                "After the activation: Linear \u2192 ReLU \u2192 Dropout",
                "Before the linear layer: Dropout \u2192 Linear \u2192 ReLU",
                "Between the linear layer and the activation: Linear \u2192 Dropout \u2192 ReLU",
                "Only after the final output layer"
            ],
            "answer": "After the activation: Linear \u2192 ReLU \u2192 Dropout"
        },
        {
            "question": "With nn.Dropout(p=0.3) in training mode, what happens to the activations that are NOT zeroed?",
            "options": [
                "They are scaled up by 1 / (1 \u2212 0.3) \u2248 1.43 so the overall magnitude stays the same",
                "They pass through unchanged",
                "They are scaled down by 0.3",
                "They are renormalized to mean 0 and std 1"
            ],
            "answer": "They are scaled up by 1 / (1 \u2212 0.3) \u2248 1.43 so the overall magnitude stays the same"
        },
        {
            "question": "You forget model.eval() before making predictions with a dropout network. What happens?",
            "options": [
                "Every prediction comes from a randomly crippled network and differs on each call \u2014 with no error raised",
                "PyTorch raises a RuntimeError reminding you to switch modes",
                "Nothing \u2014 dropout only affects the backward pass",
                "The predictions are deterministic but shifted by the dropout rate"
            ],
            "answer": "Every prediction comes from a randomly crippled network and differs on each call \u2014 with no error raised"
        },
        {
            "question": "In 20_1_1, adding Dropout(0.3) to the oversized (256, 256) network left the best loss, end-of-training loss, and collapsed training loss essentially unchanged. What explains this?",
            "options": [
                "With ~217 parameters per training sample, the network has enough spare capacity to memorize the training set through the dropout noise",
                "Dropout was accidentally disabled by model.train()",
                "The dropout layer was placed after the sigmoid output",
                "BCELoss is incompatible with dropout"
            ],
            "answer": "With ~217 parameters per training sample, the network has enough spare capacity to memorize the training set through the dropout noise"
        }
    ],
    "20_1_2: Batch Normalization": [
        {
            "question": "What is the standard placement of BatchNorm in a fully connected block?",
            "options": [
                "After the linear layer, before the activation: Linear \u2192 BatchNorm \u2192 ReLU",
                "After the activation: Linear \u2192 ReLU \u2192 BatchNorm",
                "Before the linear layer: BatchNorm \u2192 Linear \u2192 ReLU",
                "In place of the activation function"
            ],
            "answer": "After the linear layer, before the activation: Linear \u2192 BatchNorm \u2192 ReLU"
        },
        {
            "question": "Why does BatchNorm need the model.train()/model.eval() switch \u2014 what changes between the modes?",
            "options": [
                "Train mode normalizes with the current batch's statistics and updates the running averages; eval mode uses the frozen running statistics",
                "Train mode disables the learned gamma and beta parameters; eval mode enables them",
                "Train mode normalizes each feature; eval mode normalizes each sample",
                "Nothing \u2014 only Dropout responds to the mode switch"
            ],
            "answer": "Train mode normalizes with the current batch's statistics and updates the running averages; eval mode uses the frozen running statistics"
        },
        {
            "question": "What subtle bug occurs if you evaluate a BatchNorm network while it is still in train() mode?",
            "options": [
                "Every forward pass keeps updating running_mean and running_var \u2014 evaluating the model literally changes the model",
                "The gamma and beta parameters receive gradient updates during evaluation",
                "The network refuses to process batches smaller than the training batch size",
                "The loss becomes negative"
            ],
            "answer": "Every forward pass keeps updating running_mean and running_var \u2014 evaluating the model literally changes the model"
        },
        {
            "question": "In 20_1_2's experiments with Adam on the small breast-cancer problem, what did BatchNorm actually buy?",
            "options": [
                "Stability at a 50\u00d7 larger learning rate (the plain network thrashed; the BatchNorm network kept descending) \u2014 not faster convergence at the normal rate",
                "A 10\u00d7 speedup in convergence at the normal learning rate",
                "A substantially lower best validation loss",
                "Immunity to overfitting"
            ],
            "answer": "Stability at a 50\u00d7 larger learning rate (the plain network thrashed; the BatchNorm network kept descending) \u2014 not faster convergence at the normal rate"
        }
    ],
    "20_1_3: Learning Rate Schedules": [
        {
            "question": "Where does scheduler.step() belong for StepLR and CosineAnnealingLR?",
            "options": [
                "After optimizer.step(), once per epoch, outside the batch loop",
                "Before optimizer.step(), once per batch",
                "Inside the batch loop, after loss.backward()",
                "Once, before training begins"
            ],
            "answer": "After optimizer.step(), once per epoch, outside the batch loop"
        },
        {
            "question": "You accidentally call scheduler.step() inside the batch loop (11 batches per epoch) with CosineAnnealingLR(T_max=200). What happens?",
            "options": [
                "The schedule advances once per batch, so the 200-epoch decay finishes in about 18 epochs and the learning rate sits near zero for the rest of training",
                "PyTorch raises a warning and ignores the extra calls",
                "The learning rate decays 11\u00d7 more slowly than intended",
                "Nothing \u2014 schedulers count epochs internally"
            ],
            "answer": "The schedule advances once per batch, so the 200-epoch decay finishes in about 18 epochs and the learning rate sits near zero for the rest of training"
        },
        {
            "question": "In the fixed-vs-step-vs-cosine comparison on the real network, what was the honest conclusion?",
            "options": [
                "All three reach the same best validation loss; the schedules win at the endpoint by controlling late-epoch drift \u2014 and step vs. cosine differences were within noise",
                "Cosine annealing reached a dramatically lower best loss than the other two",
                "The fixed learning rate was best because schedules underfit",
                "Step decay diverged while cosine converged"
            ],
            "answer": "All three reach the same best validation loss; the schedules win at the endpoint by controlling late-epoch drift \u2014 and step vs. cosine differences were within noise"
        },
        {
            "question": "What is learning-rate warmup, and where is it standard practice?",
            "options": [
                "Starting with a very small rate and ramping up over the first few epochs before decaying \u2014 standard when training Transformers",
                "Running a few epochs on a subset of the data before full training",
                "Doubling the learning rate whenever the loss plateaus",
                "Pre-training the BatchNorm running statistics before the first epoch"
            ],
            "answer": "Starting with a very small rate and ramping up over the first few epochs before decaying \u2014 standard when training Transformers"
        }
    ],
    "20_1_4: Saving & Checkpointing": [
        {
            "question": "What does torch.save(model.state_dict(), 'model.pt') store?",
            "options": [
                "Only the named parameter tensors \u2014 nothing about the architecture, which is why loading requires a structurally identical model",
                "The full Python class definition of the model plus its weights",
                "The model, the optimizer, and the training data",
                "A compressed copy of the training loop"
            ],
            "answer": "Only the named parameter tensors \u2014 nothing about the architecture, which is why loading requires a structurally identical model"
        },
        {
            "question": "After early stopping triggers, why must you call model.load_state_dict on the saved checkpoint?",
            "options": [
                "The in-memory model is from the stopping epoch \u2014 patience epochs past the best \u2014 so without the restore you keep the wrong (worse) model",
                "PyTorch deletes the in-memory weights when training stops",
                "The optimizer overwrites the model weights at the final step",
                "load_state_dict is what disables dropout for inference"
            ],
            "answer": "The in-memory model is from the stopping epoch \u2014 patience epochs past the best \u2014 so without the restore you keep the wrong (worse) model"
        },
        {
            "question": "Why is weights_only=True the recommended habit for torch.load?",
            "options": [
                "torch.save uses pickle, and unpickling an untrusted file can execute arbitrary code; weights_only=True loads only tensors and safe primitives",
                "It halves the size of the checkpoint file",
                "It loads the weights onto the GPU automatically",
                "It is required for state_dicts saved with torch.save"
            ],
            "answer": "torch.save uses pickle, and unpickling an untrusted file can execute arbitrary code; weights_only=True loads only tensors and safe primitives"
        },
        {
            "question": "To resume an interrupted training run (not just reuse final weights), what must the checkpoint contain beyond model.state_dict()?",
            "options": [
                "The optimizer's state_dict (Adam keeps per-parameter running averages) plus bookkeeping like the epoch number",
                "The full training dataset",
                "The random seed used at import time",
                "The classification report from the last epoch"
            ],
            "answer": "The optimizer's state_dict (Adam keeps per-parameter running averages) plus bookkeeping like the epoch number"
        }
    ],
    "20_1_5: The Full Pipeline": [
        {
            "question": "In the CTG ablation (baseline / +dropout / +batchnorm / +both), the test macro F1 spread was 0.005. What is the right conclusion?",
            "options": [
                "The plain network was not overfitting, so there was nothing for the regularizers to fix \u2014 regularization is a fix for overfitting, not a free upgrade",
                "Dropout and BatchNorm cancel each other out when combined",
                "The ablation was invalid because the seeds differed across configs",
                "CTG is too small a dataset for neural networks"
            ],
            "answer": "The plain network was not overfitting, so there was nothing for the regularizers to fix \u2014 regularization is a fix for overfitting, not a free upgrade"
        },
        {
            "question": "On the breast-cancer comparison, the unregularized baseline still tested as well as the full pipeline. Why?",
            "options": [
                "train_model restores the best checkpoint, so the baseline was evaluated at its early best epoch rather than its drifted endpoint",
                "The test set was easier than the validation set",
                "The baseline secretly contained dropout layers",
                "The full pipeline overfit the validation set"
            ],
            "answer": "train_model restores the best checkpoint, so the baseline was evaluated at its early best epoch rather than its drifted endpoint"
        },
        {
            "question": "Why is the fully regularized pipeline still recommended as a default even on problems where it barely moves the result?",
            "options": [
                "You rarely know in advance whether you are overfitting \u2014 the pipeline rescues you when you are and costs almost nothing when you are not",
                "It always improves macro F1 by at least a few points",
                "It removes the need for a validation set",
                "It makes the network train faster in every case"
            ],
            "answer": "You rarely know in advance whether you are overfitting \u2014 the pipeline rescues you when you are and costs almost nothing when you are not"
        },
        {
            "question": "In the standard regularized block, what is the conventional ordering of the four pieces?",
            "options": [
                "Linear \u2192 BatchNorm \u2192 ReLU \u2192 Dropout",
                "Linear \u2192 Dropout \u2192 BatchNorm \u2192 ReLU",
                "BatchNorm \u2192 Linear \u2192 Dropout \u2192 ReLU",
                "Linear \u2192 ReLU \u2192 BatchNorm \u2192 Dropout"
            ],
            "answer": "Linear \u2192 BatchNorm \u2192 ReLU \u2192 Dropout"
        }
    ]
}

class QuizApp:
    def __init__(self, quiz_name, data):
        self.quiz_name = quiz_name
        self.data = data
        self.current_index = 0
        self.score = 0

        self.header = widgets.HTML(value="<h2>" + quiz_name + "</h2>")
        self.progress = widgets.HTML(value="")
        self.question_text = widgets.HTML(value="")
        self.options = widgets.RadioButtons(options=[], layout={"width": "max-content"})
        self.submit_btn = widgets.Button(description="Submit Answer", button_style="primary")
        self.next_btn = widgets.Button(description="Next Question", button_style="info", disabled=True)
        self.back_btn = widgets.Button(description="Back to Menu", button_style="warning")
        self.feedback = widgets.HTML(value="")

        self.submit_btn.on_click(self.handle_submit)
        self.next_btn.on_click(self.handle_next)
        self.back_btn.on_click(self.handle_back)
        self.render_question()

    def render_question(self):
        clear_output(wait=True)
        q = self.data[self.current_index]
        self.progress.value = "<b>Question " + str(self.current_index + 1) + " of " + str(len(self.data)) + "</b>"
        self.question_text.value = "<p style='font-size:14px;'>" + q["question"] + "</p>"
        self.options.options = q["options"]
        self.options.value = None
        self.submit_btn.disabled = False
        self.next_btn.disabled = True
        self.feedback.value = ""
        display(self.header, self.progress, self.question_text, self.options,
                widgets.HBox([self.submit_btn, self.next_btn]), self.feedback)

    def handle_submit(self, b):
        if self.options.value is None:
            self.feedback.value = "<span style='color:red;'>Please select an option!</span>"
            return
        correct = self.data[self.current_index]["answer"]
        if self.options.value == correct:
            self.score += 1
            self.feedback.value = "<span style='color:green; font-weight:bold;'>Correct!</span>"
        else:
            self.feedback.value = "<span style='color:red; font-weight:bold;'>Incorrect.</span><br>The correct answer was: <b>" + correct + "</b>"
        self.submit_btn.disabled = True
        self.next_btn.disabled = False

    def handle_next(self, b):
        self.current_index += 1
        if self.current_index < len(self.data):
            self.render_question()
        else:
            self.show_results()

    def handle_back(self, b):
        show_menu()

    def show_results(self):
        clear_output()
        pct = (self.score / len(self.data)) * 100
        html = """<div style="border: 2px solid #0366d6; padding: 20px; border-radius: 10px;">
            <h2>Quiz Complete!</h2>
            <p style="font-size:16px;">Your Final Score: <b>""" + str(self.score) + " / " + str(len(self.data)) + """</b></p>
            <p style="font-size:20px; color:#0366d6;"><b>Percentage: """ + "{:.1f}".format(pct) + "%</b></p></div>"
        display(widgets.HTML(html), self.back_btn)

def show_menu():
    clear_output(wait=True)
    header = widgets.HTML(value="<h2>20_1 Regularization & Practical Training — Practice Quizzes</h2>")
    instruction = widgets.HTML(value="<p>Select a notebook quiz below to begin.</p>")
    quiz_selector = widgets.Dropdown(options=list(all_quizzes.keys()), description="Select Quiz:", layout={"width": "50%"})
    start_btn = widgets.Button(description="Start Quiz", button_style="success")

    def on_start(b):
        QuizApp(quiz_selector.value, all_quizzes[quiz_selector.value])

    start_btn.on_click(on_start)
    display(header, instruction, quiz_selector, start_btn)

show_menu()
```
