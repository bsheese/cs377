```python
# @title
import ipywidgets as widgets
from IPython.display import display, clear_output

# --- ALL QUIZ DATA ---
all_quizzes = {
    "18_5_0: Intro to Ensemble Methods": [
        {
            "question": "All ensemble methods share a core idea. What is it?",
            "options": [
                "Train one very deep model that captures all patterns in the data",
                "Combine multiple models to produce a single, more accurate prediction",
                "Use cross-validation to select the single best model from a candidate set",
                "Apply regularization to prevent any single model from overfitting"
            ],
            "answer": "Combine multiple models to produce a single, more accurate prediction"
        },
        {
            "question": "In the ensemble hierarchy, what is the primary goal of bagging?",
            "options": [
                "Reduce bias by sequentially correcting the errors of previous models",
                "Reduce variance by averaging the predictions of many independently trained models",
                "Provide uncertainty quantification through Bayesian posterior distributions",
                "Maximize interpretability by producing a single, simple decision tree"
            ],
            "answer": "Reduce variance by averaging the predictions of many independently trained models"
        },
        {
            "question": "How does boosting differ from bagging in how trees are built?",
            "options": [
                "Boosting builds trees in parallel on bootstrap samples, while bagging builds them sequentially",
                "Boosting builds trees sequentially, each correcting the errors of the ensemble so far, while bagging builds trees independently in parallel",
                "Boosting uses deep trees while bagging uses shallow trees",
                "Boosting uses random feature subsets at each split while bagging considers all features"
            ],
            "answer": "Boosting builds trees sequentially, each correcting the errors of the ensemble so far, while bagging builds trees independently in parallel"
        },
        {
            "question": "What is the primary goal of boosting in the ensemble framework?",
            "options": [
                "Reduce variance by averaging many diverse, uncorrelated predictions",
                "Reduce bias by taking weak models and making them stronger through sequential error correction",
                "Decorrelate trees by forcing each one to consider only a random subset of features",
                "Provide credible intervals for each individual prediction"
            ],
            "answer": "Reduce bias by taking weak models and making them stronger through sequential error correction"
        },
        {
            "question": "What unique capability does BART offer that standard ensemble methods do not?",
            "options": [
                "It trains significantly faster than Random Forest on large datasets",
                "It provides uncertainty quantification through credible intervals for each prediction",
                "It automatically handles missing values without any preprocessing",
                "It eliminates the need for hyperparameter tuning entirely"
            ],
            "answer": "It provides uncertainty quantification through credible intervals for each prediction"
        },
        {
            "question": "The Wisconsin Breast Cancer dataset contains how many samples and features?",
            "options": [
                "2,930 samples and 82 features",
                "569 samples and 30 features",
                "1,000 samples and 21 features",
                "569 samples and 10 features"
            ],
            "answer": "569 samples and 30 features"
        },
        {
            "question": "What is the approximate class distribution in the Wisconsin Breast Cancer dataset?",
            "options": [
                "50% benign, 50% malignant",
                "70% benign, 30% malignant",
                "63% benign, 37% malignant",
                "80% benign, 20% malignant"
            ],
            "answer": "63% benign, 37% malignant"
        },
        {
            "question": "In the breast cancer context, what does a false negative represent?",
            "options": [
                "The model predicts malignant but the tumor is actually benign — an unnecessary biopsy",
                "The model predicts benign but the tumor is actually malignant — a missed cancer",
                "The model correctly predicts benign and the tumor is actually benign",
                "The model produces a probability of exactly 0.5 for a malignant tumor"
            ],
            "answer": "The model predicts benign but the tumor is actually malignant — a missed cancer"
        },
        {
            "question": "Why is accuracy alone insufficient for evaluating a model on the breast cancer dataset?",
            "options": [
                "Because the dataset is too small for accuracy to be a reliable metric",
                "Because accuracy does not distinguish between false positives and false negatives, and in a medical context, missing a cancer is far more costly than an unnecessary biopsy",
                "Because accuracy can only be computed on balanced datasets",
                "Because accuracy requires the model to output probabilities rather than class labels"
            ],
            "answer": "Because accuracy does not distinguish between false positives and false negatives, and in a medical context, missing a cancer is far more costly than an unnecessary biopsy"
        },
        {
            "question": "A naive model that always predicts 'benign' on the breast cancer dataset would achieve approximately what accuracy?",
            "options": [
                "37%",
                "50%",
                "63%",
                "95%"
            ],
            "answer": "63%"
        },
        {
            "question": "What does the 'wisdom of the crowd' principle mean in the context of ensemble methods?",
            "options": [
                "The largest model in the ensemble should dominate the final prediction",
                "Averaging the predictions of many diverse, independent models produces a more accurate result than any single model",
                "The model with the highest cross-validation score should be selected as the final model",
                "Human experts should always review and override model predictions in high-stakes applications"
            ],
            "answer": "Averaging the predictions of many diverse, independent models produces a more accurate result than any single model"
        },
        {
            "question": "How does a Random Forest differ from plain bagging?",
            "options": [
                "Random Forest uses fewer trees, making it faster to train",
                "Random Forest builds trees sequentially while bagging builds them in parallel",
                "Random Forest considers only a random subset of features at each split, decorrelating the trees beyond what bagging achieves",
                "Random Forest uses shallow trees while bagging uses deep trees"
            ],
            "answer": "Random Forest considers only a random subset of features at each split, decorrelating the trees beyond what bagging achieves"
        },
        {
            "question": "Why is the breast cancer dataset described as a 'high-stakes classification problem'?",
            "options": [
                "Because the dataset is very large and requires significant computational resources",
                "Because the features are difficult to measure accurately in clinical practice",
                "Because the costs of the two types of errors are asymmetric — missing a cancer has far more severe consequences than an unnecessary biopsy",
                "Because the class distribution is perfectly balanced, making it hard to achieve high accuracy"
            ],
            "answer": "Because the costs of the two types of errors are asymmetric — missing a cancer has far more severe consequences than an unnecessary biopsy"
        },
        {
            "question": "In the ensemble hierarchy table, what does 'Bayesian tree perturbation' refer to?",
            "options": [
                "The process of randomly removing trees from the ensemble to improve speed",
                "The MCMC-based process in BART of randomly adding branches, removing splits, and changing thresholds across the ensemble",
                "The bootstrap sampling process used in bagging to create diverse training sets",
                "The sequential reweighting of samples in AdaBoost"
            ],
            "answer": "The MCMC-based process in BART of randomly adding branches, removing splits, and changing thresholds across the ensemble"
        },
        {
            "question": "What is the relationship between bagging/random forests and variance?",
            "options": [
                "They increase variance by training many different models on different data",
                "They reduce variance by averaging many independently overfitting models, canceling out their individual errors",
                "They have no effect on variance — variance is only affected by the number of features",
                "They reduce variance by using shallow trees that cannot overfit"
            ],
            "answer": "They reduce variance by averaging many independently overfitting models, canceling out their individual errors"
        },
        {
            "question": "What is the relationship between boosting and bias?",
            "options": [
                "Boosting increases bias by using shallow trees that cannot capture complex patterns",
                "Boosting has no effect on bias — it only reduces variance",
                "Boosting reduces bias by sequentially correcting errors, turning weak learners into a strong ensemble",
                "Boosting reduces bias by using deeper trees than bagging"
            ],
            "answer": "Boosting reduces bias by sequentially correcting errors, turning weak learners into a strong ensemble"
        },
        {
            "question": "The 18_5 series uses the Wisconsin Breast Cancer dataset instead of the Ames Housing dataset used in 18_1. Why is this a meaningful change?",
            "options": [
                "Because the breast cancer dataset is larger and provides more reliable results",
                "Because the breast cancer dataset is a classification problem with asymmetric error costs, requiring different evaluation metrics than the regression problem in 18_1",
                "Because the Ames Housing dataset was too clean and did not demonstrate real-world challenges",
                "Because the breast cancer dataset has fewer features, making it easier to visualize"
            ],
            "answer": "Because the breast cancer dataset is a classification problem with asymmetric error costs, requiring different evaluation metrics than the regression problem in 18_1"
        },
        {
            "question": "What does it mean that ensemble methods 'combine multiple weak models to create a strong one'?",
            "options": [
                "Each individual model in the ensemble must perform worse than random guessing",
                "The ensemble's combined prediction is more accurate than any individual model's prediction, even if some individual models are quite good",
                "The ensemble discards the weakest models and only uses the strongest ones",
                "The ensemble assigns zero weight to models that perform below a certain threshold"
            ],
            "answer": "The ensemble's combined prediction is more accurate than any individual model's prediction, even if some individual models are quite good"
        },
        {
            "question": "In the context of the breast cancer dataset, what would be the clinical consequence of optimizing a model purely for accuracy?",
            "options": [
                "The model would achieve perfect accuracy because the dataset is well-separated",
                "The model might achieve high accuracy by correctly classifying most benign tumors while missing a significant fraction of malignant ones",
                "The model would automatically balance false positives and false negatives",
                "The model would require more features to achieve high accuracy"
            ],
            "answer": "The model might achieve high accuracy by correctly classifying most benign tumors while missing a significant fraction of malignant ones"
        },
        {
            "question": "What is the 'majority vote' mechanism in bagging and random forests?",
            "options": [
                "Each tree votes for a class, and the class with the most votes becomes the final prediction",
                "The model with the highest cross-validation score is selected as the final model",
                "Each feature votes for its importance score, and the top features are used for prediction",
                "The training samples vote on which class they belong to based on their nearest neighbors"
            ],
            "answer": "Each tree votes for a class, and the class with the most votes becomes the final prediction"
        },
        {
            "question": "Why does the notebook recommend reviewing the 18_1_5 notebook before starting 18_5?",
            "options": [
                "Because 18_1_5 covers the same dataset and provides the data cleaning steps needed for 18_5",
                "Because 18_1_5 introduced tree-based methods that 18_5 builds upon with deeper analysis",
                "Because 18_1_5 covers the mathematical proofs behind ensemble methods",
                "Because 18_1_5 contains the code for BART that is used in 18_5"
            ],
            "answer": "Because 18_1_5 introduced tree-based methods that 18_5 builds upon with deeper analysis"
        },
        {
            "question": "What does 'decorrelate' mean in the context of Random Forests?",
            "options": [
                "Remove correlated features from the dataset before training",
                "Make individual trees less similar to each other by restricting the features they can consider at each split, so that averaging them is more effective",
                "Ensure that the training and test sets have no correlated features",
                "Force each tree to use a completely different set of features"
            ],
            "answer": "Make individual trees less similar to each other by restricting the features they can consider at each split, so that averaging them is more effective"
        },
        {
            "question": "If a model on the breast cancer dataset achieves 95% accuracy but has a malignant recall of 70%, what does this mean?",
            "options": [
                "The model is excellent — 95% accuracy is very high for any medical application",
                "The model correctly identifies 95% of all tumors but misses 30% of actual cancers, which may be clinically unacceptable",
                "The model is overfitting and should be retrained with more data",
                "The model's accuracy and recall are inconsistent, indicating a bug in the evaluation code"
            ],
            "answer": "The model correctly identifies 95% of all tumors but misses 30% of actual cancers, which may be clinically unacceptable"
        },
        {
            "question": "What is the 'weighted sum' mechanism in boosting?",
            "options": [
                "Each tree's prediction is multiplied by a weight reflecting its performance, and the weighted predictions are summed for the final output",
                "Each feature is assigned a weight based on its importance, and the weighted features are summed",
                "Each training sample is weighted by its class frequency, and the weighted samples are summed",
                "The final prediction is the sum of all tree predictions divided by the number of trees"
            ],
            "answer": "Each tree's prediction is multiplied by a weight reflecting its performance, and the weighted predictions are summed for the final output"
        },
        {
            "question": "The notebook states that BART addresses 'both' bias and variance. What does this mean?",
            "options": [
                "BART uses both deep and shallow trees in the same ensemble",
                "BART combines the variance-reduction benefits of ensemble averaging with the bias-reduction benefits of sequential refinement, while also providing uncertainty estimates",
                "BART trains two separate models — one for bias and one for variance — and combines their predictions",
                "BART uses both L1 and L2 regularization to control bias and variance simultaneously"
            ],
            "answer": "BART combines the variance-reduction benefits of ensemble averaging with the bias-reduction benefits of sequential refinement, while also providing uncertainty estimates"
        },
        {
            "question": "Why does the notebook emphasize looking 'beyond accuracy' for the breast cancer problem?",
            "options": [
                "Because accuracy is computationally expensive to calculate on large datasets",
                "Because accuracy treats all errors equally, but in medical diagnosis, the cost of a false negative far exceeds the cost of a false positive",
                "Because accuracy can only be used for regression problems, not classification",
                "Because the breast cancer dataset is too imbalanced for accuracy to be computable"
            ],
            "answer": "Because accuracy treats all errors equally, but in medical diagnosis, the cost of a false negative far exceeds the cost of a false positive"
        },
        {
            "question": "What is the significance of the 'mean, standard error, and worst' feature structure in the breast cancer dataset?",
            "options": [
                "It means each of the 10 cell nucleus characteristics is measured three ways, producing 30 features total",
                "It means the dataset has been preprocessed to remove outliers",
                "It means the features are already standardized and require no scaling",
                "It means the dataset contains three separate target variables"
            ],
            "answer": "It means each of the 10 cell nucleus characteristics is measured three ways, producing 30 features total"
        },
        {
            "question": "In the ensemble hierarchy, what distinguishes a single decision tree from all ensemble methods?",
            "options": [
                "A single tree can only handle binary classification while ensembles can handle multiclass",
                "A single tree is built top-down without combining multiple models, making it interpretable but unstable",
                "A single tree requires feature scaling while ensemble methods do not",
                "A single tree always overfits while ensemble methods never overfit"
            ],
            "answer": "A single tree is built top-down without combining multiple models, making it interpretable but unstable"
        }
    ],
    "18_5_1: Decision Trees \u2014 The Foundation": [
        {
            "question": "What does Gini impurity measure in a decision tree node?",
            "options": [
                "The number of samples in the node",
                "The depth of the node in the tree",
                "How mixed the classes are in the node (0 = pure, 0.5 = maximally mixed)",
                "The total error reduction from splitting on this node"
            ],
            "answer": "How mixed the classes are in the node (0 = pure, 0.5 = maximally mixed)"
        },
        {
            "question": "In a classification tree, what does each leaf node predict?",
            "options": [
                "The mean of target values in the leaf",
                "The majority class of samples in the leaf",
                "The median of target values in the leaf",
                "A weighted average of all class probabilities"
            ],
            "answer": "The majority class of samples in the leaf"
        },
        {
            "question": "In the depth experiment, why might the best depth for malignant recall differ from the best depth for accuracy?",
            "options": [
                "Because recall and accuracy use different formulas that always disagree",
                "Because deeper trees always have higher recall regardless of accuracy",
                "Because a tree optimized for overall accuracy may miss more cancers than a slightly deeper tree",
                "Because the training data is too small to produce reliable results"
            ],
            "answer": "Because a tree optimized for overall accuracy may miss more cancers than a slightly deeper tree"
        },
        {
            "question": "What does the gap between training accuracy and testing accuracy in the depth experiment represent?",
            "options": [
                "The model's bias",
                "The overfitting penalty",
                "The effect of class imbalance",
                "The measurement error in the test set"
            ],
            "answer": "The overfitting penalty"
        },
    ],
    "18_5_2: Bagging and Random Forests": [
        {
            "question": "Why does bagging use deep (unlimited depth) trees while boosting uses shallow trees?",
            "options": [
                "Because deep trees are faster to train than shallow trees",
                "Because bagging averages many overfitting trees to cancel their individual errors, while boosting builds weak learners that sequentially correct each other",
                "Because shallow trees cannot be used with bootstrap sampling",
                "Because deep trees have lower variance than shallow trees"
            ],
            "answer": "Because bagging averages many overfitting trees to cancel their individual errors, while boosting builds weak learners that sequentially correct each other"
        },
        {
            "question": "What is Out-of-Bag (OOB) error and why is it useful?",
            "options": [
                "The error on the training set, used to check for overfitting",
                "The error on a separate validation set, used for hyperparameter tuning",
                "The error on samples left out of each bootstrap sample, providing a built-in validation estimate without needing a separate test set",
                "The error on the test set, used to report final model performance"
            ],
            "answer": "The error on samples left out of each bootstrap sample, providing a built-in validation estimate without needing a separate test set"
        },
        {
            "question": "How does Random Forest differ from plain bagging?",
            "options": [
                "Random Forest uses fewer trees than bagging",
                "Random Forest considers only a random subset of features at each split, decorrelating the trees",
                "Random Forest builds trees sequentially rather than in parallel",
                "Random Forest uses shallow trees while bagging uses deep trees"
            ],
            "answer": "Random Forest considers only a random subset of features at each split, decorrelating the trees"
        },
        {
            "question": "What effect does class_weight='balanced' typically have on a Random Forest in a medical context?",
            "options": [
                "It increases accuracy by ignoring the minority class",
                "It shifts the decision boundary to catch more malignancies at the cost of more false positives",
                "It reduces the number of trees needed in the forest",
                "It eliminates the need for cross-validation"
            ],
            "answer": "It shifts the decision boundary to catch more malignancies at the cost of more false positives"
        },
        {
            "question": "Why do correlated features share importance in a Random Forest?",
            "options": [
                "Because the model assigns equal importance to all features by design",
                "Because the model picks one correlated feature for splits and the others get no credit",
                "Because correlated features are automatically removed during training",
                "Because the importance calculation only considers the first feature alphabetically"
            ],
            "answer": "Because the model picks one correlated feature for splits and the others get no credit"
        },
    ],
    "18_5_3: Boosting and BART": [
        {
            "question": "How does AdaBoost differ from Gradient Boosting in how it handles errors?",
            "options": [
                "AdaBoost reweights misclassified samples; Gradient Boosting predicts residuals",
                "AdaBoost predicts residuals; Gradient Boosting reweights misclassified samples",
                "AdaBoost uses deep trees; Gradient Boosting uses shallow trees",
                "AdaBoost builds trees in parallel; Gradient Boosting builds trees sequentially"
            ],
            "answer": "AdaBoost reweights misclassified samples; Gradient Boosting predicts residuals"
        },
        {
            "question": "What is the 'shrinkage' concept in Gradient Boosting?",
            "options": [
                "Reducing the number of trees to prevent overfitting",
                "Using a smaller learning rate that requires more trees but often produces more robust models",
                "Limiting tree depth to reduce model complexity",
                "Removing features with low importance scores"
            ],
            "answer": "Using a smaller learning rate that requires more trees but often produces more robust models"
        },
        {
            "question": "What unique advantage does BART offer over standard ensemble methods?",
            "options": [
                "It trains faster than Random Forest",
                "It provides credible intervals (uncertainty quantification) for each prediction",
                "It requires no hyperparameter tuning",
                "It can handle categorical features without encoding"
            ],
            "answer": "It provides credible intervals (uncertainty quantification) for each prediction"
        },
        {
            "question": "Why are decision stumps (max_depth=1) used as base learners in AdaBoost?",
            "options": [
                "Because they are the fastest to train",
                "Because they are weak learners, leaving room for the sequential correction process to add value",
                "Because deeper stumps cause numerical instability",
                "Because sklearn only supports stumps for AdaBoost"
            ],
            "answer": "Because they are weak learners, leaving room for the sequential correction process to add value"
        },
        {
            "question": "In probability calibration, what does it mean if a model's curve is below the diagonal?",
            "options": [
                "The model underestimates probabilities",
                "The model overestimates probabilities",
                "The model is perfectly calibrated",
                "The model has high variance"
            ],
            "answer": "The model overestimates probabilities"
        },
    ],
    "18_5_4: Model Comparison": [
        {
            "question": "What does nested cross-validation provide that a single train/test split does not?",
            "options": [
                "Faster model training",
                "An unbiased performance estimate with quantifiable variance across folds",
                "More accurate predictions on the test set",
                "Automatic hyperparameter tuning without GridSearchCV"
            ],
            "answer": "An unbiased performance estimate with quantifiable variance across folds"
        },
        {
            "question": "In the nested CV setup for 18_5_4, what does the inner loop do?",
            "options": [
                "Evaluates the model on the held-out outer fold",
                "Tunes hyperparameters using GridSearchCV within each outer training fold",
                "Trains the final model on the full dataset",
                "Calculates the standard deviation across outer folds"
            ],
            "answer": "Tunes hyperparameters using GridSearchCV within each outer training fold"
        },
        {
            "question": "In a medical context, why might you prefer a model with slightly lower mean recall but lower variance over the model with the highest mean recall?",
            "options": [
                "Because lower variance means the model is faster to train",
                "Because a more consistent recall means the model is more reliable in clinical deployment \u2014 it won't occasionally miss many cancers",
                "Because lower variance models always have higher accuracy",
                "Because variance has no clinical significance"
            ],
            "answer": "Because a more consistent recall means the model is more reliable in clinical deployment \u2014 it won't occasionally miss many cancers"
        },
        {
            "question": "Why are the nested CV estimates typically slightly lower than single-split test scores?",
            "options": [
                "Because nested CV uses less data for training",
                "Because nested CV corrects for the optimistic bias introduced by hyperparameter tuning on the same data",
                "Because nested CV uses a different scoring metric",
                "Because nested CV has fewer folds than a single split"
            ],
            "answer": "Because nested CV corrects for the optimistic bias introduced by hyperparameter tuning on the same data"
        },
        {
            "question": "In the 18_5_4 comparison, why are red dots overlaid on the boxplots?",
            "options": [
                "To highlight outliers that should be removed",
                "To show the actual 5 outer fold data points, honestly revealing the limited sample size",
                "To indicate the best-performing fold for each model",
                "To show the mean and median values"
            ],
            "answer": "To show the actual 5 outer fold data points, honestly revealing the limited sample size"
        },
    ],
}


# --- QUIZ APP ---
class QuizApp:
    def __init__(self, quiz_name, data):
        self.quiz_name = quiz_name
        self.data = data
        self.current_index = 0
        self.score = 0
        self.user_answers = []

        # UI Elements
        self.header = widgets.HTML(value="<h2>" + quiz_name + "</h2>")
        self.progress = widgets.HTML(value="")
        self.question_text = widgets.HTML(value="")
        self.options = widgets.RadioButtons(options=[], layout={"width": "max-content"})
        self.submit_btn = widgets.Button(description="Submit Answer", button_style="primary")
        self.next_btn = widgets.Button(description="Next Question", button_style="info", disabled=True)
        self.back_btn = widgets.Button(description="Back to Menu", button_style="warning")
        self.feedback = widgets.HTML(value="")

        self.submit_btn.on_click(self.handle_submit)
        self.next_btn.on_click(self.handle_next)
        self.back_btn.on_click(self.handle_back)

        self.render_question()

    def render_question(self):
        clear_output(wait=True)
        q = self.data[self.current_index]
        self.progress.value = "<b>Question " + str(self.current_index + 1) + " of " + str(len(self.data)) + "</b>"
        self.question_text.value = "<p style='font-size:14px;'>" + q["question"] + "</p>"
        self.options.options = q["options"]
        self.options.value = None
        self.submit_btn.disabled = False
        self.next_btn.disabled = True
        self.feedback.value = ""

        display(self.header, self.progress, self.question_text, self.options,
                widgets.HBox([self.submit_btn, self.next_btn]), self.feedback)

    def handle_submit(self, b):
        if self.options.value is None:
            self.feedback.value = "<span style='color:red;'>Please select an option!</span>"
            return

        correct_answer = self.data[self.current_index]["answer"]
        if self.options.value == correct_answer:
            self.score += 1
            self.feedback.value = "<span style='color:green; font-weight:bold;'>Correct!</span>"
        else:
            self.feedback.value = "<span style='color:red; font-weight:bold;'>Incorrect.</span><br>The correct answer was: <b>" + correct_answer + "</b>"

        self.submit_btn.disabled = True
        self.next_btn.disabled = False

    def handle_next(self, b):
        self.current_index += 1
        if self.current_index < len(self.data):
            self.render_question()
        else:
            self.show_results()

    def handle_back(self, b):
        show_menu()

    def show_results(self):
        clear_output()
        percent = (self.score / len(self.data)) * 100
        result_html = """
        <div style="border: 2px solid #0366d6; padding: 20px; border-radius: 10px;">
            <h2>Quiz Complete!</h2>
            <p style="font-size:16px;">Your Final Score: <b>" + str(self.score) + " / " + str(len(self.data)) + "</b></p>
            <p style="font-size:20px; color:#0366d6;"><b>Percentage: " + "{:.1f}".format(percent) + "%</b></p>
        </div>
        """
        display(widgets.HTML(result_html), self.back_btn)


# --- MENU ---
def show_menu():
    clear_output(wait=True)
    header = widgets.HTML(value="<h2>18_5 Ensemble Methods -- Practice Quizzes</h2>")
    instruction = widgets.HTML(value="<p>Select a notebook quiz below to begin.</p>")
    quiz_selector = widgets.Dropdown(
        options=list(all_quizzes.keys()),
        description="Select Quiz:",
        layout={"width": "50%"}
    )
    start_btn = widgets.Button(description="Start Quiz", button_style="success")

    def on_start(b):
        selected = quiz_selector.value
        QuizApp(selected, all_quizzes[selected])

    start_btn.on_click(on_start)
    display(header, instruction, quiz_selector, start_btn)


# Start the app
show_menu()
```
