```python
import ipywidgets as widgets
from IPython.display import display, clear_output

all_quizzes = {
    "Part 1: Classification Foundations": [
        {
            "question": "In XGBoost, what does the scale_pos_weight parameter do?",
            "options": [
                "Increases the learning rate for positive class predictions",
                "Scales the weight of the positive class to handle class imbalance",
                "Normalizes all predictions to be between 0 and 1",
                "Scales feature values to have zero mean and unit variance"
            ],
            "answer": "Scales the weight of the positive class to handle class imbalance"
        },
        {
            "question": "For the South German Credit dataset with 70% good and 30% bad credit, what is the naive baseline accuracy?",
            "options": [
                "30%",
                "50%",
                "70%",
                "100%"
            ],
            "answer": "70%"
        },
        {
            "question": "XGBoost builds an ensemble of decision trees:",
            "options": [
                "Sequentially, correcting errors from prior trees",
                "In parallel, averaging predictions across all trees",
                "Randomly, selecting the best tree at the end"
            ],
            "answer": "Sequentially, correcting errors from prior trees"
        },
        {
            "question": "Why is stratified splitting important for imbalanced datasets?",
            "options": [
                "It makes training and test sets proportionally equal in size",
                "It ensures both train and test sets have the same class proportions",
                "It automatically balances the class weights during training",
                "It removes minority class samples to simplify training"
            ],
            "answer": "It ensures both train and test sets have the same class proportions"
        },
        {
            "question": "Why do tree-based models like XGBoost NOT require feature scaling?",
            "options": [
                "They use gradient descent, which is scale-invariant",
                "They make threshold-based splits, so relative rank matters more than magnitude",
                "They normalize features internally before training",
                "They only work with binary features"
            ],
            "answer": "They make threshold-based splits, so relative rank matters more than magnitude"
        }
    ],
    "Part 2: Confusion Matrix & Basic Metrics": [
        {
            "question": "In a confusion matrix, what does a False Negative represent?",
            "options": [
                "Predicting default when the customer is actually good",
                "Predicting good when the customer actually defaults",
                "Correctly identifying a defaulter",
                "Correctly identifying a good customer"
            ],
            "answer": "Predicting good when the customer actually defaults"
        },
        {
            "question": "Precision measures:",
            "options": [
                "Of all actual positives, how many were correctly identified",
                "Of all positive predictions, how many are actually correct",
                "The total number of correct predictions",
                "The ratio of true positives to true negatives"
            ],
            "answer": "Of all positive predictions, how many are actually correct"
        },
        {
            "question": "Recall measures:",
            "options": [
                "Of all positive predictions, how many are actually correct",
                "Of all actual positives, how many were correctly identified",
                "The total number of correct predictions",
                "The ratio of true positives to total predictions"
            ],
            "answer": "Of all actual positives, how many were correctly identified"
        },
        {
            "question": "Using the Denominator Trick: precision's denominator is ___; recall's denominator is ___.",
            "options": [
                "actual positives; predicted positives",
                "predicted positives; actual positives",
                "all samples; all positives",
                "true positives; all samples"
            ],
            "answer": "predicted positives; actual positives"
        },
        {
            "question": "Why does F1-score use harmonic mean instead of arithmetic mean?",
            "options": [
                "Harmonic mean is faster to compute",
                "Harmonic mean penalizes extreme imbalances between precision and recall",
                "Arithmetic mean cannot handle percentages",
                "Harmonic mean always gives higher scores"
            ],
            "answer": "Harmonic mean penalizes extreme imbalances between precision and recall"
        },
        {
            "question": "If precision = 100% and recall = 1%, what is the F1-score?",
            "options": [
                "About 50%",
                "About 2%",
                "About 100%",
                "About 25%"
            ],
            "answer": "About 2%"
        },
        {
            "question": "When would you prefer weighted average over macro average?",
            "options": [
                "When all classes are equally important regardless of size",
                "When you want overall performance reflecting the dataset's class distribution",
                "When you want to treat all classes equally regardless of sample size",
                "When the dataset is perfectly balanced"
            ],
            "answer": "When you want overall performance reflecting the dataset's class distribution"
        },
        {
            "question": "If you raise the decision threshold from 0.5 to 0.7, what happens to precision and recall?",
            "options": [
                "Both increase",
                "Both decrease",
                "Precision increases, recall decreases",
                "Precision decreases, recall increases"
            ],
            "answer": "Precision increases, recall decreases"
        },
        {
            "question": "Type I error is another name for:",
            "options": [
                "True Negative",
                "False Negative",
                "False Positive",
                "True Positive"
            ],
            "answer": "False Positive"
        }
    ],
    "Part 3: Examples and Practice": [
        {
            "question": "In the Adult Census Income dataset (~76% earn <=50K, ~24% earn >50K), what is the naive baseline accuracy?",
            "options": [
                "24%",
                "50%",
                "76%",
                "100%"
            ],
            "answer": "76%"
        },
        {
            "question": "After training XGBoost on the Adult Census dataset, the '>50K' class has lower precision and recall than the '<=50K' class. What is the most likely reason?",
            "options": [
                "XGBoost cannot handle more than two classes",
                "The model sees fewer positive examples during training, making the minority class harder to predict",
                "The features in the dataset are not relevant to income",
                "scale_pos_weight was not set correctly"
            ],
            "answer": "The model sees fewer positive examples during training, making the minority class harder to predict"
        },
        {
            "question": "For the Adult Census task, a false negative means the model predicted '<=50K' for someone who actually earns '>50K'. In a marketing context, what is the business cost of this error?",
            "options": [
                "Wasted marketing budget on an uninterested prospect",
                "A missed high-income prospect who would have responded to a premium offer",
                "Lower model accuracy on the training set",
                "Higher precision on the positive class"
            ],
            "answer": "A missed high-income prospect who would have responded to a premium offer"
        }
    ],
    "Part 4: ROC, AUC & Threshold Tuning": [
        {
            "question": "What does the ROC curve plot?",
            "options": [
                "Precision vs Recall",
                "True Positive Rate vs False Positive Rate",
                "Accuracy vs F1-Score",
                "Loss vs number of trees"
            ],
            "answer": "True Positive Rate vs False Positive Rate"
        },
        {
            "question": "An AUC of 0.79 means:",
            "options": [
                "The model correctly classifies 79% of samples",
                "The model correctly ranks a random positive above a random negative 79% of the time",
                "79% of predictions are true positives",
                "The model has 79% precision"
            ],
            "answer": "The model correctly ranks a random positive above a random negative 79% of the time"
        },
        {
            "question": "Why is AUC described as threshold-independent?",
            "options": [
                "It uses a fixed threshold of 0.5",
                "It evaluates performance across all possible thresholds simultaneously",
                "It only works with binary classification",
                "It doesn't require a threshold to compute"
            ],
            "answer": "It evaluates performance across all possible thresholds simultaneously"
        },
        {
            "question": "Youden's J statistic is calculated as:",
            "options": [
                "TPR + FPR",
                "TPR - FPR",
                "Precision + Recall",
                "Accuracy - Baseline"
            ],
            "answer": "TPR - FPR"
        },
        {
            "question": "Why might ROC curves be over-optimistic on imbalanced data?",
            "options": [
                "The large number of true negatives keeps FPR artificially low even with many false alarms",
                "ROC curves ignore false positives entirely",
                "ROC is not affected by class imbalance",
                "ROC uses precision instead of recall"
            ],
            "answer": "The large number of true negatives keeps FPR artificially low even with many false alarms"
        },
        {
            "question": "What is the baseline for a Precision-Recall curve?",
            "options": [
                "0.5 (random)",
                "The positive class prevalence",
                "The majority class proportion",
                "1.0 (perfect)"
            ],
            "answer": "The positive class prevalence"
        },
        {
            "question": "If false negatives cost more than false positives, where should the threshold be set relative to 0.5?",
            "options": [
                "At 0.5 (default)",
                "Higher than 0.5 (more conservative)",
                "Lower than 0.5 (flag more positives, catch more true cases)",
                "At 1.0"
            ],
            "answer": "Lower than 0.5 (flag more positives, catch more true cases)"
        },
        {
            "question": "Youden's J is the most appropriate threshold selection method when:",
            "options": [
                "False positives cost more than false negatives",
                "False negatives cost more than false positives",
                "False positives and false negatives are equally costly",
                "The dataset has no class imbalance"
            ],
            "answer": "False positives and false negatives are equally costly"
        },
        {
            "question": "Why do we use out-of-fold (OOF) training probabilities to select the threshold, rather than test-set probabilities?",
            "options": [
                "OOF probabilities are more accurate than test-set probabilities",
                "Using the test set for threshold selection leaks information and makes the final evaluation dishonest",
                "The test set does not contain enough samples for reliable threshold estimation",
                "OOF probabilities produce a smoother ROC curve"
            ],
            "answer": "Using the test set for threshold selection leaks information and makes the final evaluation dishonest"
        }
    ],
    "Part 5: Credit Card Fraud Detection": [
        {
            "question": "The credit card fraud dataset has 0.17% fraud. A model that labels every transaction 'Not Fraud' achieves 99.83% accuracy. Why is this model useless?",
            "options": [
                "Its precision is too low",
                "It catches zero fraud cases — the thing we actually care about",
                "It has a high false positive rate",
                "It has not been trained on enough data"
            ],
            "answer": "It catches zero fraud cases — the thing we actually care about"
        },
        {
            "question": "With 99.8% legitimate transactions, the weighted average F1 in the classification report is nearly indistinguishable from accuracy. Why?",
            "options": [
                "Weighted average is always equal to accuracy",
                "The legitimate class dominates the support counts, so its performance overwhelms the fraud class in the weighted calculation",
                "F1 and accuracy use the same formula",
                "The model performs equally well on both classes"
            ],
            "answer": "The legitimate class dominates the support counts, so its performance overwhelms the fraud class in the weighted calculation"
        },
        {
            "question": "For the fraud dataset, the ROC AUC looks impressive (e.g., ~0.97). Why is this misleading?",
            "options": [
                "AUC is always high when the dataset is large",
                "With 85,000+ legitimate transactions, even thousands of false alarms barely move the FPR denominator",
                "The ROC curve only measures recall, not precision",
                "XGBoost always produces high AUC on fraud data"
            ],
            "answer": "With 85,000+ legitimate transactions, even thousands of false alarms barely move the FPR denominator"
        },
        {
            "question": "The F2-score weights recall twice as heavily as precision (beta=2). When is this the right choice?",
            "options": [
                "When false positives are more costly than false negatives",
                "When missing positive cases (e.g., fraud) is more costly than generating false alarms",
                "When the dataset is perfectly balanced",
                "When you want to maximize overall accuracy"
            ],
            "answer": "When missing positive cases (e.g., fraud) is more costly than generating false alarms"
        },
        {
            "question": "On the fraud dataset, the PR curve baseline is approximately 0.0017 (0.17%). A model whose PR curve barely rises above this line would tell you:",
            "options": [
                "The model is performing well on the minority class",
                "The model is barely better than randomly guessing 'fraud' with 0.17% probability",
                "The model has high recall but low precision",
                "The positive class prevalence is too high to detect fraud"
            ],
            "answer": "The model is barely better than randomly guessing 'fraud' with 0.17% probability"
        },
        {
            "question": "Why must you use out-of-fold (OOF) probabilities — not test-set probabilities — when selecting the optimal threshold?",
            "options": [
                "OOF probabilities are always more accurate",
                "Using test-set probabilities to choose a threshold means your final evaluation is no longer on truly unseen data",
                "The test set is too small for reliable threshold estimation",
                "OOF avoids overfitting the model itself"
            ],
            "answer": "Using test-set probabilities to choose a threshold means your final evaluation is no longer on truly unseen data"
        }
    ]
}

class QuizApp:
    def __init__(self, quiz_name, data):
        self.quiz_name = quiz_name
        self.data = data
        self.current_index = 0
        self.score = 0

        self.header = widgets.HTML(value="<h2>" + quiz_name + "</h2>")
        self.progress = widgets.HTML(value="")
        self.question_text = widgets.HTML(value="")
        self.options = widgets.RadioButtons(options=[], layout={"width": "max-content"})
        self.submit_btn = widgets.Button(description="Submit Answer", button_style="primary")
        self.next_btn = widgets.Button(description="Next Question", button_style="info", disabled=True)
        self.back_btn = widgets.Button(description="Back to Menu", button_style="warning")
        self.feedback = widgets.HTML(value="")

        self.submit_btn.on_click(self.handle_submit)
        self.next_btn.on_click(self.handle_next)
        self.back_btn.on_click(self.handle_back)
        self.render_question()

    def render_question(self):
        clear_output(wait=True)
        q = self.data[self.current_index]
        self.progress.value = "<b>Question " + str(self.current_index + 1) + " of " + str(len(self.data)) + "</b>"
        self.question_text.value = "<p style='font-size:14px;'>" + q["question"] + "</p>"
        self.options.options = q["options"]
        self.options.value = None
        self.submit_btn.disabled = False
        self.next_btn.disabled = True
        self.feedback.value = ""
        display(self.header, self.progress, self.question_text, self.options,
                widgets.HBox([self.submit_btn, self.next_btn]), self.feedback)

    def handle_submit(self, b):
        if self.options.value is None:
            self.feedback.value = "<span style='color:red;'>Please select an option!</span>"
            return
        correct = self.data[self.current_index]["answer"]
        if self.options.value == correct:
            self.score += 1
            self.feedback.value = "<span style='color:green; font-weight:bold;'>Correct!</span>"
        else:
            self.feedback.value = "<span style='color:red; font-weight:bold;'>Incorrect.</span><br>The correct answer was: <b>" + correct + "</b>"
        self.submit_btn.disabled = True
        self.next_btn.disabled = False

    def handle_next(self, b):
        self.current_index += 1
        if self.current_index < len(self.data):
            self.render_question()
        else:
            self.show_results()

    def handle_back(self, b):
        show_menu()

    def show_results(self):
        clear_output()
        pct = (self.score / len(self.data)) * 100
        html = """<div style="border: 2px solid #0366d6; padding: 20px; border-radius: 10px;">
            <h2>Quiz Complete!</h2>
            <p style="font-size:16px;">Your Final Score: <b>""" + str(self.score) + " / " + str(len(self.data)) + """</b></p>
            <p style="font-size:20px; color:#0366d6;"><b>Percentage: """ + "{:.1f}".format(pct) + "%</b></p></div>"
        display(widgets.HTML(html), self.back_btn)

def show_menu():
    clear_output(wait=True)
    header = widgets.HTML(value="<h2>18_1 Classification Basics — Practice Quizzes</h2>")
    instruction = widgets.HTML(value="<p>Select a notebook quiz below to begin.</p>")
    quiz_selector = widgets.Dropdown(options=list(all_quizzes.keys()), description="Select Quiz:", layout={"width": "50%"})
    start_btn = widgets.Button(description="Start Quiz", button_style="success")

    def on_start(b):
        QuizApp(quiz_selector.value, all_quizzes[quiz_selector.value])

    start_btn.on_click(on_start)
    display(header, instruction, quiz_selector, start_btn)

show_menu()
```
