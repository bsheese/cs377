```python
# @title
import ipywidgets as widgets
from IPython.display import display, clear_output

# --- ALL QUIZ DATA ---
all_quizzes = {
    "18_1_1: Classification Foundations": [
        {
            "question": "In XGBoost, what does the scale_pos_weight parameter do?",
            "options": [
                "Increases the learning rate for positive class predictions",
                "Scales the weight of the positive class to handle class imbalance",
                "Normalizes all predictions to be between 0 and 1",
                "Scales feature values to have zero mean and unit variance"
            ],
            "answer": "Scales the weight of the positive class to handle class imbalance"
        },
        {
            "question": "For the South German Credit dataset with 70% good and 30% bad credit, what is the naive baseline accuracy?",
            "options": ["30%", "50%", "70%", "100%"],
            "answer": "70%"
        },
        {
            "question": "Why doesn't XGBoost require feature scaling?",
            "options": [
                "It automatically normalizes all input features",
                "Tree-based models make decisions based on threshold splits, not distances",
                "The gradient descent algorithm is immune to feature scale",
                "XGBoost converts all features to categorical before processing"
            ],
            "answer": "Tree-based models make decisions based on threshold splits, not distances"
        },
        {
            "question": "What does feature importance (gain) measure in XGBoost?",
            "options": [
                "How much the feature correlates with the target variable",
                "How much the feature contributes to reducing the loss function when making splits",
                "The raw coefficient value for each feature",
                "How many missing values the feature contains"
            ],
            "answer": "How much the feature contributes to reducing the loss function when making splits"
        },
        {
            "question": "What is the difference between .predict() and .predict_proba()?",
            "options": [
                ".predict() returns probabilities, .predict_proba() returns class labels",
                ".predict() returns class labels, .predict_proba() returns probabilities",
                ".predict() is faster than .predict_proba()",
                "There is no difference - they are identical"
            ],
            "answer": ".predict() returns class labels, .predict_proba() returns probabilities"
        },
        {
            "question": "The probability distribution overlap shows 41.1% of actual defaulters scored below 0.5. What does this mean?",
            "options": [
                "41.1% of good customers were incorrectly classified as defaulters",
                "41.1% of actual defaulters were missed (false negatives)",
                "The model has 41.1% precision",
                "The model accuracy is 41.1%"
            ],
            "answer": "41.1% of actual defaulters were missed (false negatives)"
        },
        {
            "question": "XGBoost builds an ensemble of decision trees:",
            "options": [
                "In parallel, all at once",
                "Sequentially, with each tree correcting the errors of previous trees",
                "Randomly, with no particular order",
                "From the bottom up, starting with the root node"
            ],
            "answer": "Sequentially, with each tree correcting the errors of previous trees"
        },
        {
            "question": "Why is stratified splitting important for imbalanced datasets?",
            "options": [
                "It makes the training set larger than the test set",
                "It ensures both train and test sets have the same class proportions",
                "It automatically balances the class weights",
                "It removes all minority class samples to simplify training"
            ],
            "answer": "It ensures both train and test sets have the same class proportions"
        }
    ],
    "18_1_2: Confusion Matrix and Basic Metrics": [
        {
            "question": "In a confusion matrix, what does a False Negative represent?",
            "options": [
                "Predicting default when the customer is actually good",
                "Predicting good when the customer actually defaults",
                "Correctly identifying a defaulter",
                "Correctly identifying a good customer"
            ],
            "answer": "Predicting good when the customer actually defaults"
        },
        {
            "question": "Precision measures:",
            "options": [
                "Of all actual positives, how many were correctly identified",
                "Of all positive predictions, how many are actually correct",
                "The total number of correct predictions",
                "The ratio of true positives to true negatives"
            ],
            "answer": "Of all positive predictions, how many are actually correct"
        },
        {
            "question": "Recall measures:",
            "options": [
                "Of all positive predictions, how many are actually correct",
                "Of all actual positives, how many were correctly identified",
                "The total number of correct predictions",
                "The ratio of true positives to total predictions"
            ],
            "answer": "Of all actual positives, how many were correctly identified"
        },
        {
            "question": "Why does F1-score use harmonic mean instead of arithmetic mean?",
            "options": [
                "Harmonic mean is faster to compute",
                "Harmonic mean gives more weight to lower values, penalizing extreme imbalances",
                "Arithmetic mean cannot handle percentages",
                "Harmonic mean always gives higher scores"
            ],
            "answer": "Harmonic mean gives more weight to lower values, penalizing extreme imbalances"
        },
        {
            "question": "If precision = 100% and recall = 1%, what is the F1-score?",
            "options": ["About 50%", "About 2%", "About 100%", "About 25%"],
            "answer": "About 2%"
        },
        {
            "question": "When would you prefer weighted average over macro average?",
            "options": [
                "When all classes are equally important",
                "When you want overall performance on the dataset population",
                "When you want to treat all classes equally regardless of sample size",
                "When the dataset is perfectly balanced"
            ],
            "answer": "When you want overall performance on the dataset population"
        },
        {
            "question": "If you raise the decision threshold from 0.5 to 0.7, what happens to precision and recall?",
            "options": [
                "Both increase",
                "Both decrease",
                "Precision increases, recall decreases",
                "Precision decreases, recall increases"
            ],
            "answer": "Precision increases, recall decreases"
        },
        {
            "question": "Type I error is another name for:",
            "options": ["True Negative", "False Negative", "False Positive", "True Positive"],
            "answer": "False Positive"
        }
    ],
    "18_1_3: ROC, AUC, and Threshold Tuning": [
        {
            "question": "What does the ROC curve plot?",
            "options": [
                "Precision vs Recall",
                "True Positive Rate vs False Positive Rate",
                "Accuracy vs F1-Score",
                "Loss vs Epoch"
            ],
            "answer": "True Positive Rate vs False Positive Rate"
        },
        {
            "question": "An AUC of 0.79 means:",
            "options": [
                "The model correctly classifies 79% of samples",
                "The model correctly ranks a random positive above a random negative 79% of the time",
                "79% of predictions are true positives",
                "The model has 79% precision"
            ],
            "answer": "The model correctly ranks a random positive above a random negative 79% of the time"
        },
        {
            "question": "Why is AUC described as threshold-independent?",
            "options": [
                "It uses a fixed threshold of 0.5",
                "It evaluates performance across all possible thresholds",
                "It only works with binary classification",
                "It doesn't require a threshold to compute"
            ],
            "answer": "It evaluates performance across all possible thresholds"
        },
        {
            "question": "Youden's J statistic is calculated as:",
            "options": ["TPR + FPR", "TPR - FPR", "Precision + Recall", "Accuracy - Baseline"],
            "answer": "TPR - FPR"
        },
        {
            "question": "Why might ROC curves be over-optimistic on imbalanced data?",
            "options": [
                "ROC includes true negatives in the calculation",
                "ROC curves ignore false positives",
                "ROC is not affected by class imbalance",
                "ROC uses precision instead of recall"
            ],
            "answer": "ROC includes true negatives in the calculation"
        },
        {
            "question": "What is the baseline for a Precision-Recall curve?",
            "options": ["0.5 (random)", "The positive class prevalence", "The majority class proportion", "1.0 (perfect)"],
            "answer": "The positive class prevalence"
        },
        {
            "question": "If false negatives cost more than false positives, where should you set the threshold?",
            "options": [
                "At 0.5 (default)",
                "Higher than 0.5 (more conservative)",
                "Lower than 0.5 (more aggressive)",
                "At 1.0"
            ],
            "answer": "Lower than 0.5 (more aggressive)"
        },
        {
            "question": "A perfect ROC curve would:",
            "options": [
                "Follow the diagonal line",
                "Hug the top-left corner",
                "Be a horizontal line at y=0.5",
                "Be a vertical line at x=0"
            ],
            "answer": "Hug the top-left corner"
        }
    ],
    "18_1_4: Model Selection via Cross-Validation": [
        {
            "question": "Why is 5-fold cross-validation preferred over a single train/test split?",
            "options": [
                "It is faster to compute",
                "It provides a more reliable performance estimate by averaging across multiple splits",
                "It always gives higher accuracy",
                "It requires less data"
            ],
            "answer": "It provides a more reliable performance estimate by averaging across multiple splits"
        },
        {
            "question": "A wide box in a cross-validation boxplot indicates:",
            "options": [
                "The model is always accurate",
                "The model's performance varies significantly across different data splits",
                "The model has high bias",
                "The model is underfitting"
            ],
            "answer": "The model's performance varies significantly across different data splits"
        },
        {
            "question": "Why do we score models on both accuracy AND F1?",
            "options": [
                "F1 is always more important than accuracy",
                "Accuracy can be misleading on imbalanced data; F1 reveals how well the model handles the minority class",
                "Accuracy is required by law",
                "They always give the same result"
            ],
            "answer": "Accuracy can be misleading on imbalanced data; F1 reveals how well the model handles the minority class"
        },
        {
            "question": "Which models were compared in the notebook's model competition?",
            "options": [
                "Logistic Regression and Neural Network",
                "XGBoost, Decision Tree, Random Forest, and SVM",
                "K-Means and DBSCAN",
                "Linear Regression and Ridge"
            ],
            "answer": "XGBoost, Decision Tree, Random Forest, and SVM"
        },
        {
            "question": "Does the notebook cover L1/L2/ElasticNet regularization?",
            "options": ["Yes, in detail", "Yes, briefly mentioned", "No, it was not covered", "Only L1 was covered"],
            "answer": "No, it was not covered"
        },
        {
            "question": "If Random Forest has the highest mean F1 but also the widest spread, you should:",
            "options": [
                "Always choose it because it has the highest mean",
                "Consider both mean performance and stability; a slightly lower but more consistent model might be preferable",
                "Choose the lowest variance model regardless of mean",
                "Use a single train/test split instead of CV"
            ],
            "answer": "Consider both mean performance and stability; a slightly lower but more consistent model might be preferable"
        },
        {
            "question": "Why don't Decision Trees and Random Forests require feature scaling?",
            "options": [
                "They are immune to overfitting",
                "They use information gain, not distances",
                "Scaling is done automatically inside the algorithm",
                "Scaling would reduce accuracy"
            ],
            "answer": "They use information gain, not distances"
        },
        {
            "question": "Which XGBoost parameter controls how many trees are in the ensemble?",
            "options": ["max_depth", "learning_rate", "n_estimators", "subsample"],
            "answer": "n_estimators"
        }
    ],
    "18_1_5: Multiclass Classification": [
        {
            "question": "In One-vs-Rest (OvR) multiclass strategy, how many classifiers are trained for K classes?",
            "options": ["1", "K", "K-1", "K*(K-1)/2"],
            "answer": "K"
        },
        {
            "question": "How does Softmax differ from OvR?",
            "options": [
                "Softmax trains one model with K outputs that compete against each other; OvR trains K independent binary classifiers",
                "Softmax only works with 2 classes",
                "Softmax cannot handle imbalanced classes",
                "OvR is faster than Softmax"
            ],
            "answer": "Softmax trains one model with K outputs that compete against each other; OvR trains K independent binary classifiers"
        },
        {
            "question": "In a 3x3 confusion matrix, the diagonal contains:",
            "options": ["False positives", "False negatives", "Correct predictions", "Uncertain predictions"],
            "answer": "Correct predictions"
        },
        {
            "question": "Macro average treats all classes:",
            "options": [
                "Equally, regardless of sample size",
                "Based on their support (sample count)",
                "Based on their precision scores only",
                "Randomly"
            ],
            "answer": "Equally, regardless of sample size"
        },
        {
            "question": "When would macro and weighted averages give very different results?",
            "options": [
                "When the dataset is perfectly balanced",
                "When classes have very different sample sizes and performance varies significantly across classes",
                "When using binary classification",
                "When using OvR strategy"
            ],
            "answer": "When classes have very different sample sizes and performance varies significantly across classes"
        },
        {
            "question": "What does the OvR probability histogram overlap indicate?",
            "options": [
                "The classifier is performing well",
                "The classifier is confused between classes (areas of uncertainty)",
                "The model has perfect precision",
                "The dataset is balanced"
            ],
            "answer": "The classifier is confused between classes (areas of uncertainty)"
        },
        {
            "question": "What type of multiclass approach does XGBoost use natively?",
            "options": ["One-vs-Rest", "One-vs-One", "Softmax (Multinomial)", "Binary encoding"],
            "answer": "Softmax (Multinomial)"
        },
        {
            "question": "If class 0 and class 2 are frequently confused in a wine dataset, what might this indicate?",
            "options": [
                "The model is broken",
                "Those two cultivars share similar chemical properties",
                "The dataset is too small",
                "The features need to be scaled"
            ],
            "answer": "Those two cultivars share similar chemical properties"
        }
    ],
    "18_1_6: Decision Boundaries and Feature Importance": [
        {
            "question": "What does PCA do?",
            "options": [
                "Removes features that are not important",
                "Finds directions of maximum variance to project high-dimensional data into lower dimensions",
                "Converts categorical variables to numerical",
                "Normalizes data to have zero mean"
            ],
            "answer": "Finds directions of maximum variance to project high-dimensional data into lower dimensions"
        },
        {
            "question": "Decision tree boundaries are:",
            "options": [
                "Straight lines (hyperplanes)",
                "Axis-aligned rectangles (horizontal and vertical steps)",
                "Wavy and irregular curves",
                "Diagonal lines only"
            ],
            "answer": "Axis-aligned rectangles (horizontal and vertical steps)"
        },
        {
            "question": "KNN produces what type of decision boundaries?",
            "options": ["Straight lines", "Axis-aligned rectangles", "Wavy, irregular, and local", "Circular"],
            "answer": "Wavy, irregular, and local"
        },
        {
            "question": "If you increase K in KNN, what happens to the decision boundaries?",
            "options": ["They become more jagged", "They become smoother", "They become straight lines", "They disappear"],
            "answer": "They become smoother"
        },
        {
            "question": "Does feature importance tell you the direction (positive/negative) of the effect?",
            "options": [
                "Yes, always",
                "No, it only tells you which features matter, not how they affect predictions",
                "Only for binary features",
                "Only for numerical features"
            ],
            "answer": "No, it only tells you which features matter, not how they affect predictions"
        },
        {
            "question": "If two features are highly correlated, how does this affect feature importance?",
            "options": [
                "Both appear more important",
                "They share importance between them, so neither appears as dominant as they would individually",
                "Correlation has no effect on importance",
                "Only one feature will show any importance"
            ],
            "answer": "They share importance between them, so neither appears as dominant as they would individually"
        },
        {
            "question": "Why might decision boundaries in PCA-reduced 2D space differ from the true 13D boundaries?",
            "options": [
                "PCA introduces errors in the data",
                "PCA loses information when projecting to 2D; the true boundaries may be more complex",
                "2D boundaries are always more accurate",
                "The difference is negligible"
            ],
            "answer": "PCA loses information when projecting to 2D; the true boundaries may be more complex"
        },
        {
            "question": "Which feature was identified as a top discriminator for wine classification in the notebook?",
            "options": ["Alcohol", "Proline", "Magnesium", "Ash"],
            "answer": "Proline"
        }
    ]
}


# --- QUIZ APP ---
class QuizApp:
    def __init__(self, quiz_name, data):
        self.quiz_name = quiz_name
        self.data = data
        self.current_index = 0
        self.score = 0
        self.user_answers = []

        # UI Elements
        self.header = widgets.HTML(value="<h2>" + quiz_name + "</h2>")
        self.progress = widgets.HTML(value="")
        self.question_text = widgets.HTML(value="")
        self.options = widgets.RadioButtons(options=[], layout={"width": "max-content"})
        self.submit_btn = widgets.Button(description="Submit Answer", button_style="primary")
        self.next_btn = widgets.Button(description="Next Question", button_style="info", disabled=True)
        self.back_btn = widgets.Button(description="Back to Menu", button_style="warning")
        self.feedback = widgets.HTML(value="")

        self.submit_btn.on_click(self.handle_submit)
        self.next_btn.on_click(self.handle_next)
        self.back_btn.on_click(self.handle_back)

        self.render_question()

    def render_question(self):
        clear_output(wait=True)
        q = self.data[self.current_index]
        self.progress.value = "<b>Question " + str(self.current_index + 1) + " of " + str(len(self.data)) + "</b>"
        self.question_text.value = "<p style='font-size:14px;'>" + q["question"] + "</p>"
        self.options.options = q["options"]
        self.options.value = None
        self.submit_btn.disabled = False
        self.next_btn.disabled = True
        self.feedback.value = ""

        display(self.header, self.progress, self.question_text, self.options,
                widgets.HBox([self.submit_btn, self.next_btn]), self.feedback)

    def handle_submit(self, b):
        if self.options.value is None:
            self.feedback.value = "<span style='color:red;'>Please select an option!</span>"
            return

        correct_answer = self.data[self.current_index]["answer"]
        if self.options.value == correct_answer:
            self.score += 1
            self.feedback.value = "<span style='color:green; font-weight:bold;'>Correct!</span>"
        else:
            self.feedback.value = "<span style='color:red; font-weight:bold;'>Incorrect.</span><br>The correct answer was: <b>" + correct_answer + "</b>"

        self.submit_btn.disabled = True
        self.next_btn.disabled = False

    def handle_next(self, b):
        self.current_index += 1
        if self.current_index < len(self.data):
            self.render_question()
        else:
            self.show_results()

    def handle_back(self, b):
        show_menu()

    def show_results(self):
        clear_output()
        percent = (self.score / len(self.data)) * 100
        result_html = """
        <div style="border: 2px solid #0366d6; padding: 20px; border-radius: 10px;">
            <h2>Quiz Complete!</h2>
            <p style="font-size:16px;">Your Final Score: <b>""" + str(self.score) + """ / """ + str(len(self.data)) + """</b></p>
            <p style="font-size:20px; color:#0366d6;"><b>Percentage: """ + "{:.1f}".format(percent) + """%</b></p>
        </div>
        """
        display(widgets.HTML(result_html), self.back_btn)


# --- MENU ---
def show_menu():
    clear_output(wait=True)
    header = widgets.HTML(value="<h2>18_1 Classification Basics -- Practice Quizzes</h2>")
    instruction = widgets.HTML(value="<p>Select a notebook quiz below to begin.</p>")
    quiz_selector = widgets.Dropdown(
        options=list(all_quizzes.keys()),
        description="Select Quiz:",
        layout={"width": "50%"}
    )
    start_btn = widgets.Button(description="Start Quiz", button_style="success")

    def on_start(b):
        selected = quiz_selector.value
        QuizApp(selected, all_quizzes[selected])

    start_btn.on_click(on_start)
    display(header, instruction, quiz_selector, start_btn)


# Start the app
show_menu()
```
