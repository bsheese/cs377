```python
# @title
import ipywidgets as widgets
from IPython.display import display, clear_output

# --- ALL QUIZ DATA ---
all_quizzes = {
    "18_1_0: Classification Foundations": [
        {
            "question": "This is a temporary test question for 18_1_0. What is 2 + 2?",
            "options": ["3", "4", "5", "6"],
            "answer": "4"
        },
        {
            "question": "This is a temporary test question for 18_1_0. What color is the sky on a clear day?",
            "options": ["Green", "Blue", "Red", "Yellow"],
            "answer": "Blue"
        },
        {
            "question": "This is a temporary test question for 18_1_0. Which is a programming language?",
            "options": ["Python", "Banana", "Guitar", "Ocean"],
            "answer": "Python"
        }
    ],
    "18_1_1: Confusion Matrix and Basic Metrics": [
        {
            "question": "This is a temporary test question for 18_1_1. What is 3 x 3?",
            "options": ["6", "9", "12", "15"],
            "answer": "9"
        },
        {
            "question": "This is a temporary test question for 18_1_1. How many days are in a week?",
            "options": ["5", "6", "7", "8"],
            "answer": "7"
        },
        {
            "question": "This is a temporary test question for 18_1_1. Which shape has three sides?",
            "options": ["Square", "Circle", "Triangle", "Pentagon"],
            "answer": "Triangle"
        }
    ],
    "18_1_2: ROC, AUC, and Threshold Tuning": [
        {
            "question": "This is a temporary test question for 18_1_2. What is 10 minus 7?",
            "options": ["2", "3", "4", "5"],
            "answer": "3"
        },
        {
            "question": "This is a temporary test question for 18_1_2. Which planet is closest to the Sun?",
            "options": ["Venus", "Earth", "Mercury", "Mars"],
            "answer": "Mercury"
        },
        {
            "question": "This is a temporary test question for 18_1_2. Which is a mammal?",
            "options": ["Eagle", "Shark", "Whale", "Frog"],
            "answer": "Whale"
        }
    ],
    "18_1_3: Model Selection via Cross-Validation": [
        {
            "question": "This is a temporary test question for 18_1_3. What is 100 divided by 4?",
            "options": ["20", "25", "30", "35"],
            "answer": "25"
        },
        {
            "question": "This is a temporary test question for 18_1_3. How many continents are there?",
            "options": ["5", "6", "7", "8"],
            "answer": "7"
        },
        {
            "question": "This is a temporary test question for 18_1_3. Which is a primary color?",
            "options": ["Green", "Orange", "Red", "Purple"],
            "answer": "Red"
        }
    ],
    "18_1_4: Multiclass Classification": [
        {
            "question": "This is a temporary test question for 18_1_4. What is 5 x 6?",
            "options": ["25", "30", "35", "40"],
            "answer": "30"
        },
        {
            "question": "This is a temporary test question for 18_1_4. Which is the largest ocean?",
            "options": ["Atlantic", "Indian", "Arctic", "Pacific"],
            "answer": "Pacific"
        },
        {
            "question": "This is a temporary test question for 18_1_4. Which is a fruit?",
            "options": ["Carrot", "Potato", "Apple", "Onion"],
            "answer": "Apple"
        }
    ],
    "18_1_5: Decision Boundaries and Feature Importance": [
        {
            "question": "This is a temporary test question for 18_1_5. What is 8 + 7?",
            "options": ["13", "14", "15", "16"],
            "answer": "15"
        },
        {
            "question": "This is a temporary test question for 18_1_5. Which is the fastest land animal?",
            "options": ["Lion", "Cheetah", "Horse", "Elephant"],
            "answer": "Cheetah"
        },
        {
            "question": "This is a temporary test question for 18_1_5. Which is a metal?",
            "options": ["Wood", "Glass", "Iron", "Plastic"],
            "answer": "Iron"
        }
    ]
}


# --- QUIZ APP ---
class QuizApp:
    def __init__(self, quiz_name, data):
        self.quiz_name = quiz_name
        self.data = data
        self.current_index = 0
        self.score = 0
        self.user_answers = []

        # UI Elements
        self.header = widgets.HTML(value="<h2>" + quiz_name + "</h2>")
        self.progress = widgets.HTML(value="")
        self.question_text = widgets.HTML(value="")
        self.options = widgets.RadioButtons(options=[], layout={"width": "max-content"})
        self.submit_btn = widgets.Button(description="Submit Answer", button_style="primary")
        self.next_btn = widgets.Button(description="Next Question", button_style="info", disabled=True)
        self.back_btn = widgets.Button(description="Back to Menu", button_style="warning")
        self.feedback = widgets.HTML(value="")

        self.submit_btn.on_click(self.handle_submit)
        self.next_btn.on_click(self.handle_next)
        self.back_btn.on_click(self.handle_back)

        self.render_question()

    def render_question(self):
        clear_output(wait=True)
        q = self.data[self.current_index]
        self.progress.value = "<b>Question " + str(self.current_index + 1) + " of " + str(len(self.data)) + "</b>"
        self.question_text.value = "<p style='font-size:14px;'>" + q["question"] + "</p>"
        self.options.options = q["options"]
        self.options.value = None
        self.submit_btn.disabled = False
        self.next_btn.disabled = True
        self.feedback.value = ""

        display(self.header, self.progress, self.question_text, self.options,
                widgets.HBox([self.submit_btn, self.next_btn]), self.feedback)

    def handle_submit(self, b):
        if self.options.value is None:
            self.feedback.value = "<span style='color:red;'>Please select an option!</span>"
            return

        correct_answer = self.data[self.current_index]["answer"]
        if self.options.value == correct_answer:
            self.score += 1
            self.feedback.value = "<span style='color:green; font-weight:bold;'>Correct!</span>"
        else:
            self.feedback.value = "<span style='color:red; font-weight:bold;'>Incorrect.</span><br>The correct answer was: <b>" + correct_answer + "</b>"

        self.submit_btn.disabled = True
        self.next_btn.disabled = False

    def handle_next(self, b):
        self.current_index += 1
        if self.current_index < len(self.data):
            self.render_question()
        else:
            self.show_results()

    def handle_back(self, b):
        show_menu()

    def show_results(self):
        clear_output()
        percent = (self.score / len(self.data)) * 100
        result_html = """
        <div style="border: 2px solid #0366d6; padding: 20px; border-radius: 10px;">
            <h2>Quiz Complete!</h2>
            <p style="font-size:16px;">Your Final Score: <b>" + str(self.score) + " / " + str(len(self.data)) + "</b></p>
            <p style="font-size:20px; color:#0366d6;"><b>Percentage: " + "{:.1f}".format(percent) + "%</b></p>
        </div>
        """
        display(widgets.HTML(result_html), self.back_btn)


# --- MENU ---
def show_menu():
    clear_output(wait=True)
    header = widgets.HTML(value="<h2>18_1 Classification Basics -- Practice Quizzes</h2>")
    instruction = widgets.HTML(value="<p>Select a notebook quiz below to begin.</p>")
    quiz_selector = widgets.Dropdown(
        options=list(all_quizzes.keys()),
        description="Select Quiz:",
        layout={"width": "50%"}
    )
    start_btn = widgets.Button(description="Start Quiz", button_style="success")

    def on_start(b):
        selected = quiz_selector.value
        QuizApp(selected, all_quizzes[selected])

    start_btn.on_click(on_start)
    display(header, instruction, quiz_selector, start_btn)


# Start the app
show_menu()
```
