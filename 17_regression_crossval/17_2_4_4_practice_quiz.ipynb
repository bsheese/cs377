```python
# @title
import ipywidgets as widgets
from IPython.display import display, clear_output

# --- QUIZ DATA ---
quiz_data = [
    {
        "question": "In the manual alpha tuning loop, you plot both training R\u00b2 and cross-validation R\u00b2 against alpha. As alpha increases from 0.01 to 1000, what pattern do you expect to see in the two curves?",
        "options": [
            "Training R\u00b2 rises steadily while CV R\u00b2 falls steadily, reflecting increasing model complexity.",
            "Training R\u00b2 falls steadily while CV R\u00b2 rises then falls, peaking at the optimal alpha.",
            "Both curves rise in parallel, since regularization always improves model fit.",
            "Training R\u00b2 stays flat while CV R\u00b2 oscillates randomly around a constant value."
        ],
        "answer": "Training R\u00b2 falls steadily while CV R\u00b2 rises then falls, peaking at the optimal alpha."
    },
    {
        "question": "A colleague runs GridSearchCV with 5-fold CV on a training set of 2,000 samples, testing 10 alpha values. How many total model fits does GridSearchCV perform?",
        "options": [
            "10, because it only fits one model per alpha value.",
            "50, because it fits 10 alphas across 5 folds.",
            "2,000, because it fits one model per training sample.",
            "15, because it adds the number of alphas and folds together."
        ],
        "answer": "50, because it fits 10 alphas across 5 folds."
    },
    {
        "question": "You use GridSearchCV with refit=True (the default). After the search completes, what model does grid_search.best_estimator_ return?",
        "options": [
            "The model trained on the single best-performing fold during the search.",
            "The model retrained on the entire training set using the best hyperparameters found.",
            "An ensemble of all models tested during the search, weighted by their CV scores.",
            "The model from the fold with the lowest validation error, without retraining."
        ],
        "answer": "The model retrained on the entire training set using the best hyperparameters found."
    },
    {
        "question": "Why does the Pipeline prefix hyperparameter keys with 'model__' (e.g., 'model__alpha') when passing them to GridSearchCV?",
        "options": [
            "It is a naming convention that GridSearchCV ignores; any prefix would work.",
            "It tells GridSearchCV which step in the pipeline the parameter belongs to.",
            "It prevents GridSearchCV from trying to tune the scaler's parameters.",
            "It signals to scikit-learn that the parameter should be tuned on a logarithmic scale."
        ],
        "answer": "It tells GridSearchCV which step in the pipeline the parameter belongs to."
    },
    {
        "question": "You fit a StandardScaler on the full dataset, then pass the scaled data to cross_val_score with cv=5. What is the consequence?",
        "options": [
            "Nothing changes; scaling before CV is mathematically equivalent to scaling within each fold.",
            "The scaler's mean and variance incorporate information from every fold, leaking into each training split.",
            "Cross-validation automatically detects the pre-scaled data and adjusts its fold boundaries.",
            "The model trains faster because the scaling computation is performed only once instead of five times."
        ],
        "answer": "The scaler's mean and variance incorporate information from every fold, leaking into each training split."
    },
    {
        "question": "In ElasticNet, the l1_ratio parameter controls the mix between L1 and L2 penalties. What does l1_ratio=0.7 mean?",
        "options": [
            "70% of the penalty comes from the L1 (Lasso) term and 30% from the L2 (Ridge) term.",
            "70% of the features will be dropped to exactly zero by the model.",
            "The model will behave 70% like Ridge and 30% like Lasso.",
            "The overall penalty strength is reduced to 70% of the specified alpha value."
        ],
        "answer": "70% of the penalty comes from the L1 (Lasso) term and 30% from the L2 (Ridge) term."
    },
    {
        "question": "After tuning three regularized models (Ridge, Lasso, ElasticNet) with GridSearchCV, all three achieve nearly identical test R\u00b2 scores. What is the most likely explanation?",
        "options": [
            "The hyperparameter grids were too narrow to distinguish between the models.",
            "The remaining features are weakly informative, so all regularization approaches converge to similar solutions.",
            "GridSearchCV always returns the same model regardless of the algorithm specified.",
            "The test set is too small to distinguish between different modeling approaches."
        ],
        "answer": "The remaining features are weakly informative, so all regularization approaches converge to similar solutions."
    },
    {
        "question": "You run GridSearchCV on a training set and find the best alpha. You then evaluate that model on a held-out test set and get R\u00b2 = 0.83. Why might this score be slightly inflated?",
        "options": [
            "The test set is systematically easier to predict than the training set due to random sampling.",
            "The best alpha was selected because it happened to perform well on the specific CV folds, partially capturing noise.",
            "GridSearchCV uses a different scoring metric than the one used for the final test evaluation.",
            "The model was retrained on the full training set, which artificially boosts test performance."
        ],
        "answer": "The best alpha was selected because it happened to perform well on the specific CV folds, partially capturing noise."
    },
    {
        "question": "In nested cross-validation with 5 outer folds and 5 inner folds, how many total model fits are performed when searching over 8 alpha values?",
        "options": [
            "40, because there are 5 outer folds times 8 alpha values, giving 40 total fits.",
            "200, because each outer fold runs the full inner GridSearch.",
            "25, because there are 5 outer folds times 5 inner folds, giving 25 total fits.",
            "13, because you add the 5 outer folds, 5 inner folds, and 3 unique alpha values together."
        ],
        "answer": "200, because each outer fold runs the full inner GridSearch."
    },
    {
        "question": "In the nested CV code, cross_val_score is called with a GridSearchCV object as its estimator. What happens during each outer fold?",
        "options": [
            "cross_val_score extracts the best hyperparameters from the GridSearchCV and fits a simple model with them.",
            "cross_val_score calls .fit() on the GridSearchCV, which runs the full inner search, then uses the refitted best model to predict on the outer test fold.",
            "cross_val_score skips the inner search and uses the default hyperparameters of the base estimator.",
            "cross_val_score averages the predictions of all hyperparameter combinations tested by the inner search."
        ],
        "answer": "cross_val_score calls .fit() on the GridSearchCV, which runs the full inner search, then uses the refitted best model to predict on the outer test fold."
    },
    {
        "question": "Your nested CV results show Ridge R\u00b2 = 0.7910 \u00b1 0.0266, while the single train/test split gave R\u00b2 = 0.8300. What does the \u00b1 0.0266 tell you?",
        "options": [
            "The model's performance varies by about 2.7 percentage points depending on which data ends up in each fold.",
            "The nested CV code has a bug that introduces random noise into the scores.",
            "The inner loop's hyperparameter search is unstable and needs more alpha values.",
            "The outer test folds contain outlier houses that distort the average score."
        ],
        "answer": "The model's performance varies by about 2.7 percentage points depending on which data ends up in each fold."
    },
    {
        "question": "Why does nested CV use the full dataset (X, y) rather than just X_train, y_train?",
        "options": [
            "Because nested CV replaces the train/test split; every data point serves as both training and test data across different folds.",
            "Because the outer loop needs more data than the inner loop to produce stable hyperparameter estimates.",
            "Because GridSearchCV requires the full dataset to properly scale features before splitting.",
            "Because using only the training set would cause the inner loop to overfit more severely."
        ],
        "answer": "Because nested CV replaces the train/test split; every data point serves as both training and test data across different folds."
    },
    {
        "question": "In nested CV, Outer Fold 1 selects alpha=1 as optimal, while Outer Fold 3 selects alpha=10. What does this tell you?",
        "options": [
            "The inner loop's random state was not fixed, causing non-reproducible results.",
            "The optimal regularization strength depends on which specific samples are in the training subset.",
            "The model is fundamentally unstable and should not be trusted for any predictions.",
            "One of the outer folds contains corrupted data that skews the hyperparameter search."
        ],
        "answer": "The optimal regularization strength depends on which specific samples are in the training subset."
    },
    {
        "question": "A learning curve shows training R\u00b2 = 0.95 and validation R\u00b2 = 0.70 at the largest training size, with a persistent gap between the two curves. What should you do?",
        "options": [
            "Add more training data, since the validation curve is still improving.",
            "Increase model complexity, since both scores are too low.",
            "Increase regularization or reduce model complexity, since the model is overfitting.",
            "Switch to a different scoring metric, since R\u00b2 is not appropriate for this dataset."
        ],
        "answer": "Increase regularization or reduce model complexity, since the model is overfitting."
    },
    {
        "question": "A learning curve shows both training and validation R\u00b2 converging at approximately 0.55, with the validation curve flatlining as training size increases. What does this indicate?",
        "options": [
            "The model is overfitting and would benefit from more training data.",
            "The model is underfitting; adding more data will not help, and a more complex model is needed.",
            "The model is well-calibrated and performing at its theoretical maximum.",
            "The cross-validation folds are too small to produce reliable validation scores."
        ],
        "answer": "The model is underfitting; adding more data will not help, and a more complex model is needed."
    },
    {
        "question": "In the learning curve code, train_sizes=np.linspace(0.1, 1.0, 5) is used. How many different training set sizes are evaluated?",
        "options": [
            "5, representing 10%, 32.5%, 55%, 77.5%, and 100% of the training data.",
            "10, because linspace generates values from 0.1 to 1.0 in steps of 0.1.",
            "1, because the model is only trained on the full dataset at the end.",
            "3, because the first and last values are excluded from the evaluation."
        ],
        "answer": "5, representing 10%, 32.5%, 55%, 77.5%, and 100% of the training data."
    },
    {
        "question": "You compare nested CV R\u00b2 scores for Ridge (0.7910), Lasso (0.7910), and ElasticNet (0.7909). All three are nearly identical. What does this suggest about the underlying data?",
        "options": [
            "The feature set is small enough that L1 and L2 penalties produce nearly equivalent coefficient shrinkage.",
            "The GridSearchCV grids for all three models were identical, forcing them to the same solution.",
            "The nested CV implementation has a bug that returns the same score for all models.",
            "The data contains no predictive signal, so all models converge to the same baseline."
        ],
        "answer": "The feature set is small enough that L1 and L2 penalties produce nearly equivalent coefficient shrinkage."
    },
    {
        "question": "Which statement best describes the relationship between the inner loop's best CV score and the outer loop's average score in nested CV?",
        "options": [
            "The inner loop's score is always higher because it averages over smaller training subsets.",
            "The outer loop's score is always higher because it evaluates on larger test folds.",
            "The two scores are always identical because they use the same scoring metric.",
            "The inner loop's score is always lower because it uses more folds than the outer loop."
        ],
        "answer": "The inner loop's score is always higher because it averages over smaller training subsets."
    },
    {
        "question": "You are building a production model for a client who needs a trustworthy performance estimate. Your dataset has 1,500 samples. Which approach should you use?",
        "options": [
            "A single 80/20 train/test split with GridSearchCV, because it is fast and simple.",
            "Nested cross-validation, because it provides an unbiased estimate with quantifiable variance.",
            "Leave-one-out cross-validation without hyperparameter tuning, because it uses all data for training.",
            "A single model fit on all 1,500 samples with no validation, because the dataset is too small to split."
        ],
        "answer": "Nested cross-validation, because it provides an unbiased estimate with quantifiable variance."
    },
    {
        "question": "In the bias-variance tradeoff plot for Ridge regression, the CV R\u00b2 peaks at alpha=1. If you were to add alpha=0.5 and alpha=2 to your grid, what would you expect?",
        "options": [
            "Alpha=0.5 would have a higher CV R\u00b2 because lower alpha always means better generalization.",
            "Alpha=2 would have a higher CV R\u00b2 because higher alpha always means better generalization.",
            "Both would likely have lower CV R\u00b2 than alpha=1, since alpha=1 is already at the peak.",
            "One would be higher and one lower, but you cannot predict which without running the experiment."
        ],
        "answer": "One would be higher and one lower, but you cannot predict which without running the experiment."
    },
    {
        "question": "A student claims: 'Nested CV is unnecessary if I have a separate test set that I never touch during tuning.' What is the strongest counterargument?",
        "options": [
            "The student is correct; a separate test set completely eliminates the need for nested CV.",
            "A single test set gives only one performance estimate with no measure of its variability across different data splits.",
            "Nested CV is always faster than a train/test split, so there is no reason not to use it.",
            "The test set score is always lower than the nested CV score, making nested CV the more conservative choice."
        ],
        "answer": "A single test set gives only one performance estimate with no measure of its variability across different data splits."
    },
    {
        "question": "When using learning_curve() with cv=5, how is the validation score computed at each training size?",
        "options": [
            "The model is trained once on the specified fraction of data and scored on the remaining data.",
            "For each training size, 5 models are trained on different folds and the validation scores are averaged.",
            "The model is trained on the full dataset and scored on 5 different random subsets of the test set.",
            "The training and validation scores are computed on the same data to measure overfitting."
        ],
        "answer": "For each training size, 5 models are trained on different folds and the validation scores are averaged."
    },
    {
        "question": "You notice that the learning curve's validation score is still climbing at the largest training size (100% of training data). What is the most actionable next step?",
        "options": [
            "Reduce model complexity, since the model is clearly overfitting at all training sizes.",
            "Collect more training data, since the model has not yet plateaued and could benefit from additional examples.",
            "Switch to a simpler model, since the current model is too complex for the available data.",
            "Increase the number of cross-validation folds to get a more accurate validation score."
        ],
        "answer": "Collect more training data, since the model has not yet plateaued and could benefit from additional examples."
    },
    {
        "question": "In the notebook, the 'Features Kept' column counts coefficients with absolute value > 1e-5. Why is this threshold used instead of checking for exact zeros?",
        "options": [
            "Because Ridge regression never produces exact zeros; coefficients are shrunk toward zero but rarely reach it.",
            "Because floating-point arithmetic makes exact zero comparisons unreliable for all models.",
            "Because Lasso always produces coefficients slightly above zero due to numerical imprecision.",
            "Because the 1e-5 threshold is the default regularization strength in scikit-learn."
        ],
        "answer": "Because Ridge regression never produces exact zeros; coefficients are shrunk toward zero but rarely reach it."
    },
    {
        "question": "You run nested CV with 5 outer folds and 3 inner folds, searching 12 hyperparameter combinations. Your manager asks how long it will take compared to a single GridSearchCV run (5 folds, 12 combinations). What do you tell them?",
        "options": [
            "It will take about the same time, since the inner loop reuses computations from the outer loop.",
            "It will take roughly 5 times longer, because the outer loop runs the full inner GridSearch once per outer fold.",
            "It will take 15 times longer, because you multiply the outer folds, inner folds, and combinations.",
            "It will take less time, because nested CV uses smaller training sets in each fold."
        ],
        "answer": "It will take roughly 5 times longer, because the outer loop runs the full inner GridSearch once per outer fold."
    },
    {
        "question": "The notebook states that nested CV scores are 'typically lower' than train/test split scores. Under what circumstance might nested CV produce a higher score?",
        "options": [
            "When the single test split happened to contain unusually difficult-to-predict samples by chance.",
            "When the inner loop uses more folds than the outer loop, inflating the score.",
            "When the model is severely underfitting, making all evaluation methods produce identical scores.",
            "This can never happen; nested CV always produces a lower score by definition."
        ],
        "answer": "When the single test split happened to contain unusually difficult-to-predict samples by chance."
    },
    {
        "question": "In the GridSearchCV output, 'Fitting 5 folds for each of 20 candidates, totalling 100 fits' appears for ElasticNet. What are the 20 candidates?",
        "options": [
            "20 different random splits of the training data into train and validation sets.",
            "The Cartesian product of 4 alpha values and 5 l1_ratio values, giving 20 unique hyperparameter combinations.",
            "20 different model architectures, each with a different number of features.",
            "20 iterations of the ElasticNet optimization algorithm before convergence."
        ],
        "answer": "The Cartesian product of 4 alpha values and 5 l1_ratio values, giving 20 unique hyperparameter combinations."
    },
    {
        "question": "You fit a learning curve and observe that the training score starts at 0.99 with 10% of the data and drops to 0.79 with 100% of the data, while the validation score rises from 0.50 to 0.78. What is happening?",
        "options": [
            "The model is overfitting on small training sets and gradually improving its generalization as more data is added.",
            "The model is underfitting at all training sizes and would benefit from a more complex architecture.",
            "The cross-validation folds are too small at low training sizes, producing unreliable scores.",
            "The learning rate is too high, causing the model to diverge as more data is introduced."
        ],
        "answer": "The model is overfitting on small training sets and gradually improving its generalization as more data is added."
    },
    {
        "question": "A colleague argues that nested CV is 'wasteful' because it trains hundreds of models but only reports a single average score. How would you respond?",
        "options": [
            "They are correct; a single train/test split with GridSearchCV is always sufficient for reporting results.",
            "The average score is the deliverable, but the standard deviation across outer folds is equally valuable information about model stability.",
            "Nested CV produces a separate final model for each outer fold, giving you five models to deploy in production.",
            "The computational cost of nested CV is negligible on modern hardware, so the objection is moot."
        ],
        "answer": "The average score is the deliverable, but the standard deviation across outer folds is equally valuable information about model stability."
    },
    {
        "question": "After running nested CV, you want to deploy a final model for production predictions. Which model should you deploy?",
        "options": [
            "The average of all models produced across the outer folds, since averaging reduces variance.",
            "The model from the outer fold with the highest test score, since it performed best.",
            "A fresh model trained on all data using the most common hyperparameters from the inner loops.",
            "The model from the inner loop with the lowest validation error, since it was most accurate."
        ],
        "answer": "A fresh model trained on all data using the most common hyperparameters from the inner loops."
    },
]

# --- QUIZ ENGINE ---
class QuizApp:
    def __init__(self, data):
        self.data = data
        self.current_index = 0
        self.score = 0
        self.user_answers = []

        # UI Elements
        self.header = widgets.HTML(value="<h2>Grid Search, Nested CV, and Learning Curves Quiz</h2>")
        self.progress = widgets.HTML(value="")
        self.question_text = widgets.HTML(value="")
        self.options = widgets.RadioButtons(options=[], layout={'width': 'max-content'})
        self.submit_btn = widgets.Button(description="Submit Answer", button_style='primary')
        self.next_btn = widgets.Button(description="Next Question", button_style='info', disabled=True)
        self.feedback = widgets.HTML(value="")
        self.output = widgets.Output()

        self.submit_btn.on_click(self.handle_submit)
        self.next_btn.on_click(self.handle_next)

        self.render_question()

    def render_question(self):
        clear_output(wait=True)
        q = self.data[self.current_index]
        self.progress.value = f"<b>Question {self.current_index + 1} of {len(self.data)}</b>"
        self.question_text.value = f"<p style='font-size:14px;'>{q['question']}</p>"
        self.options.options = q['options']
        self.options.value = None
        self.submit_btn.disabled = False
        self.next_btn.disabled = True
        self.feedback.value = ""

        display(self.header, self.progress, self.question_text, self.options,
                widgets.HBox([self.submit_btn, self.next_btn]), self.feedback)

    def handle_submit(self, b):
        if self.options.value is None:
            self.feedback.value = "<span style='color:red;'>Please select an option!</span>"
            return

        correct_answer = self.data[self.current_index]['answer']
        if self.options.value == correct_answer:
            self.score += 1
            self.feedback.value = "<span style='color:green; font-weight:bold;'>\u2705 Correct!</span>"
        else:
            self.feedback.value = f"<span style='color:red; font-weight:bold;'>\u274c Incorrect.</span><br>The correct answer was: <b>{correct_answer}</b>"

        self.submit_btn.disabled = True
        self.next_btn.disabled = False

    def handle_next(self, b):
        self.current_index += 1
        if self.current_index < len(self.data):
            self.render_question()
        else:
            self.show_results()

    def show_results(self):
        clear_output()
        percent = (self.score / len(self.data)) * 100
        result_html = f"""
        <div style="border: 2px solid #0366d6; padding: 20px; border-radius: 10px;">
            <h2>Quiz Complete!</h2>
            <p style='font-size:16px;'>Your Final Score: <b>{self.score} / {len(self.data)}</b></p>
            <p style='font-size:20px; color:#0366d6;'><b>Percentage: {percent:.1f}%</b></p>
        </div>
        """
        display(widgets.HTML(result_html))

# Start the app
QuizApp(quiz_data)
```
