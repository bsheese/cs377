```python
# @title
import ipywidgets as widgets
from IPython.display import display, clear_output

# --- QUIZ DATA ---
quiz_data = [
    {
        "question": "What is the primary objective of introducing regularization into a regression model?",
        "options": ["To find the absolute global minimum of the error function on training data.",
                    "To penalize model complexity and discourage excessively large coefficients.",
                    "To increase the weight of the most important features to improve fit.",
                    "To eliminate the need for an intercept in the regression equation."],
        "answer": "To penalize model complexity and discourage excessively large coefficients."
    },
    {
        "question": "In the context of standard OLS, why are 'massive weights' considered a problem?",
        "options": ["They make the model computationally too expensive to run.",
                    "They cause the model to ignore random noise in the dataset.",
                    "They make the model unstable, where tiny input changes cause massive prediction swings.",
                    "They lead to an underfit model with high bias."],
        "answer": "They make the model unstable, where tiny input changes cause massive prediction swings."
    },
    {
        "question": "How does regularization 'compromise' during the training process?",
        "options": ["It balances minimizing error with keeping feature weights small.",
                    "It balances training speed with the number of features used.",
                    "It balances the mean squared error with the R-squared value.",
                    "It compromises between forward and backward selection."],
        "answer": "It balances minimizing error with keeping feature weights small."
    },
    {
        "question": "Compared to stepwise selection (Forward/Backward), regularization is described as 'nuanced' because:",
        "options": ["It can only remove features that have zero correlation with the target.",
                    "It is a binary process that decides if a feature is 100% in or out.",
                    "It can shrink coefficients to very small numbers without abruptly muting them.",
                    "It requires fewer observations than features to work."],
        "answer": "It can shrink coefficients to very small numbers without abruptly muting them."
    },
    {
        "question": "Why is stepwise selection considered 'unstable' compared to regularized models?",
        "options": ["Because it always results in a model with the highest possible MSE.",
                    "Because small changes in training data can lead to a completely different set of chosen features.",
                    "Because it can only be used with polynomial features.",
                    "Because it is mathematically incapable of reaching an intercept."],
        "answer": "Because small changes in training data can lead to a completely different set of chosen features."
    },
    {
        "question": "Regarding computation, why did the Ames selection take 90 minutes with OLS but seconds with Regularization?",
        "options": ["OLS selection requires training and comparing thousands of separate models sequentially.",
                    "Regularization uses a simpler version of the dataset with fewer rows.",
                    "OLS requires the computer to calculate the log of every feature.",
                    "Regularization skips the cross-validation step entirely."],
        "answer": "OLS selection requires training and comparing thousands of separate models sequentially."
    },
    {
        "question": "What happens in Ridge Regression (L2) when Alpha is set to exactly 0?",
        "options": ["All coefficients are forced to zero.",
                    "The model becomes a LASSO regression.",
                    "The model acts exactly like a standard OLS linear regression.",
                    "The penalty becomes infinitely large."],
        "answer": "The model acts exactly like a standard OLS linear regression."
    },
    {
        "question": "Which statement best describes the coefficient shrinkage behavior of Ridge Regression?",
        "options": ["It forces unimportant features to be exactly 0.0.",
                    "It shrinks coefficients asymptotically toward zero but never forces them to exactly 0.0.",
                    "It only shrinks the intercept while leaving slopes untouched.",
                    "It increases coefficients for features that are highly correlated."],
        "answer": "It shrinks coefficients asymptotically toward zero but never forces them to exactly 0.0."
    },
    {
        "question": "Why is Ridge Regression (L2) particularly effective at handling multicollinearity?",
        "options": ["It picks the feature with the highest p-value and drops the rest.",
                    "It combines correlated features into a single new feature.",
                    "It distributes 'credit' among correlated variables, shrinking their weights together.",
                    "It transforms all correlated variables into categorical integers."],
        "answer": "It distributes 'credit' among correlated variables, shrinking their weights together."
    },
    {
        "question": "The 'L2' in L2 regularization refers to which mathematical concept?",
        "options": ["Taking the square root of the number of features.",
                    "The L2 norm, which calculates the penalty based on the squared value of coefficients.",
                    "Having exactly two target variables in the model.",
                    "The second derivative of the loss function."],
        "answer": "The L2 norm, which calculates the penalty based on the squared value of coefficients."
    },
    {
        "question": "If you have a domain where every variable is conceptually important and you don't want to lose any, which method is best?",
        "options": ["LASSO (L1)", "Backward Selection", "Ridge (L2)", "Forward Selection"],
        "answer": "Ridge (L2)"
    },
    {
        "question": "What is a major limitation of Ridge Regression in terms of model 'decluttering'?",
        "options": ["It cannot perform automatic feature selection/removal.",
                    "It requires the target variable to be normally distributed.",
                    "It is only compatible with categorical data.",
                    "It significantly increases the variance of the model."],
        "answer": "It cannot perform automatic feature selection/removal."
    },
    {
        "question": "Why is 'Feature Scaling' (like StandardScaler) mandatory for both Ridge and Lasso?",
        "options": ["Because the penalty is applied to the coefficient size; unscaled data would unfairly penalize features with large units.",
                    "Because the algorithm cannot handle negative numbers in the raw data.",
                    "Because it ensures that the R-squared value remains between 0 and 1.",
                    "Because regularization only works on data that is normally distributed."],
        "answer": "Because the penalty is applied to the coefficient size; unscaled data would unfairly penalize features with large units."
    },
    {
        "question": "How does regularization affect the 'Interpretability' of the resulting slopes (coefficients)?",
        "options": ["It makes them more accurate representations of real-world physical changes.",
        "It introduces mathematical bias, meaning slopes can no longer be interpreted as literal unit increases.",
        "It makes the slopes exactly equal to the correlation coefficient.",
        "It has no effect on the interpretation of the slopes."],
        "answer": "It introduces mathematical bias, meaning slopes can no longer be interpreted as literal unit increases."
    },
    {
        "question": "What does the acronym LASSO stand for?",
        "options": ["Linear Absolute Selection and Standard Operator",
                    "Least Absolute Shrinkage and Selection Operator",
                    "Logarithmic Aligned Squared Standard Optimization",
                    "Least Adaptive Selection and Shrinkage Optimizer"],
        "answer": "Least Absolute Shrinkage and Selection Operator"
    },
    {
        "question": "What is the key mathematical difference between Lasso and Ridge penalties?",
        "options": ["Lasso uses squared values (L2); Ridge uses absolute values (L1).",
                    "Lasso uses absolute values (L1); Ridge uses squared values (L2).",
                    "Lasso penalizes the intercept; Ridge does not.",
                    "Lasso uses a log scale; Ridge uses a linear scale."],
        "answer": "Lasso uses absolute values (L1); Ridge uses squared values (L2)."
    },
    {
        "question": "Lasso is often preferred over Ridge when the goal is:",
        "options": ["To ensure no variable ever reaches 0.0.",
                    "To keep the most complex model possible.",
                    "To achieve automatic feature selection and create a simpler model.",
                    "To eliminate the need for cross-validation."],
        "answer": "To achieve automatic feature selection and create a simpler model."
    },
    {
        "question": "If Lasso encounters two highly correlated features, how does it typically behave?",
        "options": ["It keeps both and gives them equal weights.",
                    "It tends to arbitrarily pick one to keep and drops the other to zero.",
                    "It averages the two features together.",
                    "It drops both features to avoid multicollinearity."],
        "answer": "It tends to arbitrarily pick one to keep and drops the other to zero."
    },
    {
        "question": "In the Bias-Variance tradeoff, regularization intentionally:",
        "options": ["Increases variance to reduce bias.",
                    "Trades a bit of training accuracy (bias) for better generalization (reduced variance).",
                    "Eliminates both bias and variance simultaneously.",
                    "Decreases bias to ensure the training error is 0."],
        "answer": "Trades a bit of training accuracy (bias) for better generalization (reduced variance)."
    },
    {
        "question": "What is ElasticNet Regression?",
        "options": ["A method that uses only categorical data.",
                    "A regression that stretches the data points to fit a curve.",
                    "A mathematical combination of both Lasso (L1) and Ridge (L2) penalties.",
                    "A type of regression that does not require feature scaling."],
        "answer": "A mathematical combination of both Lasso (L1) and Ridge (L2) penalties."
    },
    {
        "question": "In ElasticNet, what does an 'l1_ratio' of 1.0 signify?",
        "options": ["The model is 100% Ridge.",
                    "The model is a 50/50 blend of Lasso and Ridge.",
                    "The model is 100% Lasso.",
                    "The model has no regularization penalty."],
        "answer": "The model is 100% Lasso."
    },
    {
        "question": "Why is ElasticNet useful in fields like genomics where there are 'more columns than rows' (p > n)?",
        "options": ["It can handle datasets where standard OLS and Lasso mathematically break down.",
                    "It automatically creates new rows of data using synthetic sampling.",
                    "It requires no hyperparameter tuning.",
                    "It is the only model that allows for negative R-squared values."],
        "answer": "It can handle datasets where standard OLS and Lasso mathematically break down."
    },
    {
        "question": "What is the primary computational 'cost' of using ElasticNet over Ridge or Lasso?",
        "options": ["It requires double the memory on the hard drive.",
                    "It requires tuning two parameters (Alpha and L1_ratio) simultaneously, often needing a 2D Grid Search.",
                    "It can only be run on small datasets with under 100 rows.",
                    "It makes the residual plots impossible to read."],
        "answer": "It requires tuning two parameters (Alpha and L1_ratio) simultaneously, often needing a 2D Grid Search."
    },
    {
        "question": "According to the text, if a researcher only cares about 'predictive accuracy' and not interpretability, they usually move on to:",
        "options": ["Simple Linear Regression", "Tree-based models (like Random Forests or XGBoost)", "Manual Feature Selection", "Ridge Regression with Alpha=0"],
        "answer": "Tree-based models (like Random Forests or XGBoost)"
    },
    {
        "question": "Regularization helps prevent a model from 'memorizing' random noise. This 'memorization' is also known as:",
        "options": ["Underfitting", "Overfitting", "Global Optimization", "Standardization"],
        "answer": "Overfitting"
    },
    {
        "question": "In the Ames modeling workflow, why was 'SalePrice' log-transformed (y = np.log)?",
        "options": ["To make the calculation of Alpha easier.",
                    "To handle right-skewness in the target variable.",
                    "Because Lasso only works on log-scaled data.",
                    "To increase the number of features in the model."],
        "answer": "To handle right-skewness in the target variable."
    },
    {
        "question": "Lasso Regression uses an L1 penalty. The '1' in L1 refers to:",
        "options": ["The fact that it only uses one feature.",
                    "Taking the absolute value to the power of 1.",
                    "Having only one alpha value.",
                    "Being the #1 preferred method for Ridge enthusiasts."],
        "answer": "Taking the absolute value to the power of 1."
    },
    {
        "question": "If Alpha is set to an 'overwhelmingly high' value in Lasso, what is the likely result?",
        "options": ["Every feature will have a massive coefficient.",
                    "The model will be identical to an OLS model.",
                    "The model will drop all predictors, leaving only the intercept.",
                    "The R-squared value will increase significantly."],
        "answer": "The model will drop all predictors, leaving only the intercept."
    },
    {
        "question": "When comparing Ridge and ElasticNet behavior on highly correlated groups of variables:",
        "options": ["Ridge kills the whole group; ElasticNet keeps one.",
                    "Ridge keeps the whole group; ElasticNet can keep the group but also zero out truly useless variables.",
                    "Both models behave exactly like Forward Selection.",
                    "Neither model can handle correlated variables."],
        "answer": "Ridge keeps the whole group; ElasticNet can keep the group but also zero out truly useless variables."
    },
    {
        "question": "Which step is 'CRITICAL' for linear models when performing One-Hot Encoding to prevent multicollinearity?",
        "options": ["Scaling the target variable.", "Setting 'drop_first=True'.", "Removing all numeric columns.", "Using cross-validation with 100 folds."],
        "answer": "Setting 'drop_first=True'."
    }
]

# --- QUIZ ENGINE ---
class QuizApp:
    def __init__(self, data):
        self.data = data
        self.current_index = 0
        self.score = 0
        self.user_answers = []

        # UI Elements
        self.header = widgets.HTML(value="<h2>Regularization Conceptual Quiz</h2>")
        self.progress = widgets.HTML(value="")
        self.question_text = widgets.HTML(value="")
        self.options = widgets.RadioButtons(options=[], layout={'width': 'max-content'})
        self.submit_btn = widgets.Button(description="Submit Answer", button_style='primary')
        self.next_btn = widgets.Button(description="Next Question", button_style='info', disabled=True)
        self.feedback = widgets.HTML(value="")
        self.output = widgets.Output()

        self.submit_btn.on_click(self.handle_submit)
        self.next_btn.on_click(self.handle_next)

        self.render_question()

    def render_question(self):
        clear_output(wait=True)
        q = self.data[self.current_index]
        self.progress.value = f"<b>Question {self.current_index + 1} of {len(self.data)}</b>"
        self.question_text.value = f"<p style='font-size:14px;'>{q['question']}</p>"
        self.options.options = q['options']
        self.options.value = None
        self.submit_btn.disabled = False
        self.next_btn.disabled = True
        self.feedback.value = ""

        display(self.header, self.progress, self.question_text, self.options,
                widgets.HBox([self.submit_btn, self.next_btn]), self.feedback)

    def handle_submit(self, b):
        if self.options.value is None:
            self.feedback.value = "<span style='color:red;'>Please select an option!</span>"
            return

        correct_answer = self.data[self.current_index]['answer']
        if self.options.value == correct_answer:
            self.score += 1
            self.feedback.value = "<span style='color:green; font-weight:bold;'>✅ Correct!</span>"
        else:
            self.feedback.value = f"<span style='color:red; font-weight:bold;'>❌ Incorrect.</span><br>The correct answer was: <b>{correct_answer}</b>"

        self.submit_btn.disabled = True
        self.next_btn.disabled = False

    def handle_next(self, b):
        self.current_index += 1
        if self.current_index < len(self.data):
            self.render_question()
        else:
            self.show_results()

    def show_results(self):
        clear_output()
        percent = (self.score / len(self.data)) * 100
        result_html = f"""
        <div style="border: 2px solid #0366d6; padding: 20px; border-radius: 10px;">
            <h2>Quiz Complete!</h2>
            <p style='font-size:16px;'>Your Final Score: <b>{self.score} / {len(self.data)}</b></p>
            <p style='font-size:20px; color:#0366d6;'><b>Percentage: {percent:.1f}%</b></p>
        </div>
        """
        display(widgets.HTML(result_html))

# Start the app
QuizApp(quiz_data)
```
