```python
# @title
import ipywidgets as widgets
from IPython.display import display, clear_output

# --- QUIZ DATA ---
quiz_data = [
    {
        "question": "Linear models required manual polynomial features to capture non-linear patterns. How do tree-based models handle non-linearity?",
        "options": [
            "They apply logarithmic transformations to all features automatically.",
            "They partition the data into regions and predict the average within each region.",
            "They add squared terms to the regression equation during training.",
            "They use gradient descent to find the optimal non-linear curve."
        ],
        "answer": "They partition the data into regions and predict the average within each region."
    },
    {
        "question": "In a regression tree, what value does each leaf node predict?",
        "options": [
            "The median of the target values of training samples in that leaf.",
            "The average of the target values of training samples in that leaf.",
            "The most frequent target value among training samples in that leaf.",
            "A weighted sum of coefficients applied to the features in that leaf."
        ],
        "answer": "The average of the target values of training samples in that leaf."
    },
    {
        "question": "How does a regression tree decide which feature and split point to use at each node?",
        "options": [
            "It randomly selects a feature and tries all possible split points on that feature.",
            "It selects the feature with the highest correlation to the target variable.",
            "It tries every possible split on every feature and picks the one that reduces error the most.",
            "It uses the feature that was most important in the parent node's split."
        ],
        "answer": "It tries every possible split on every feature and picks the one that reduces error the most."
    },
    {
        "question": "An unconstrained decision tree achieves R\u00b2 = 1.0 on training data but R\u00b2 = 0.80 on test data. What is happening?",
        "options": [
            "The model is underfitting because it has too few splits to capture the patterns.",
            "The model is overfitting by creating a separate leaf for nearly every training sample.",
            "The training data contains outliers that the model is correctly ignoring.",
            "The test set is fundamentally different from the training set due to data leakage."
        ],
        "answer": "The model is overfitting by creating a separate leaf for nearly every training sample."
    },
    {
        "question": "In the depth experiment, the test R\u00b2 peaked at depth 7 and then declined as depth increased. What does this pattern represent?",
        "options": [
            "The bias-variance tradeoff, where deeper trees reduce bias but increase variance.",
            "A computational limitation where deeper trees run out of memory.",
            "The learning rate decay, where deeper trees contribute less to the prediction.",
            "A data quality issue where deeper splits encounter noisier samples."
        ],
        "answer": "The bias-variance tradeoff, where deeper trees reduce bias but increase variance."
    },
    {
        "question": "Why does a Random Forest use bootstrap sampling (drawing with replacement) for each tree?",
        "options": [
            "To ensure every tree sees the exact same data but in a different order.",
            "To reduce the computational cost of training each individual tree.",
            "To give each tree a different perspective on the data so they make different errors.",
            "To eliminate the need for a separate test set during model evaluation."
        ],
        "answer": "To give each tree a different perspective on the data so they make different errors."
    },
    {
        "question": "In a Random Forest, what does max_features='sqrt' mean at each split?",
        "options": [
            "Each tree considers only the square root of the total features across the entire tree.",
            "Each tree considers a random subset of about \u221an features at each individual node.",
            "The forest builds exactly \u221an trees, where n is the number of features.",
            "Each tree uses the top \u221an most important features and ignores the rest."
        ],
        "answer": "Each tree considers a random subset of about \u221an features at each individual node."
    },
    {
        "question": "Why does feature randomness in Random Forests prevent every tree from splitting on the same dominant feature first?",
        "options": [
            "Because the dominant feature is removed from the dataset before training begins.",
            "Because each tree is forced to consider only a random subset of features at each node.",
            "Because the dominant feature's values are randomly shuffled before each split.",
            "Because the algorithm penalizes trees that use the same feature twice."
        ],
        "answer": "Because each tree is forced to consider only a random subset of features at each node."
    },
    {
        "question": "The Random Forest selected max_depth=None as optimal. Why is unlimited depth acceptable in a forest but catastrophic for a single tree?",
        "options": [
            "Because the forest uses a different splitting criterion that prevents overfitting.",
            "Because the averaging process across diverse trees cancels out individual overfitting.",
            "Because each tree in the forest sees fewer features, making deep trees simpler.",
            "Because the forest automatically prunes trees that show signs of overfitting."
        ],
        "answer": "Because the averaging process across diverse trees cancels out individual overfitting."
    },
    {
        "question": "What is the primary objective of a Random Forest from a bias-variance perspective?",
        "options": [
            "Reducing bias by building deeper and deeper trees sequentially.",
            "Reducing variance by averaging many diverse, uncorrelated predictions.",
            "Reducing both bias and variance equally through random feature selection.",
            "Eliminating bias entirely by using an infinite number of trees."
        ],
        "answer": "Reducing variance by averaging many diverse, uncorrelated predictions."
    },
    {
        "question": "In gradient boosting, what does each new tree predict?",
        "options": [
            "The original target variable (e.g., house price) directly.",
            "The residuals (errors) left by the ensemble of all previous trees.",
            "The average of all previous trees' predictions.",
            "A weighted combination of the target variable and the previous residuals."
        ],
        "answer": "The residuals (errors) left by the ensemble of all previous trees."
    },
    {
        "question": "Gradient boosting starts with an initial prediction equal to the average target value. What is the role of the first tree?",
        "options": [
            "It predicts the target variable directly, ignoring the initial prediction.",
            "It predicts the residuals from the initial prediction, correcting the average.",
            "It replaces the initial prediction entirely with its own predictions.",
            "It determines which features are most important for all subsequent trees."
        ],
        "answer": "It predicts the residuals from the initial prediction, correcting the average."
    },
    {
        "question": "What is the key difference between Random Forests and Gradient Boosting in how they build their ensembles?",
        "options": [
            "Random Forests build trees sequentially; Gradient Boosting builds them in parallel.",
            "Random Forests average independent trees; Gradient Boosting corrects errors sequentially.",
            "Random Forests use deep trees; Gradient Boosting uses only single-split trees.",
            "Random Forests reduce bias; Gradient Boosting reduces variance."
        ],
        "answer": "Random Forests average independent trees; Gradient Boosting corrects errors sequentially."
    },
    {
        "question": "Why does HistGradientBoostingRegressor use histogram binning?",
        "options": [
            "To convert categorical features into numeric codes automatically.",
            "To group continuous data into discrete bins, speeding up the split-finding process.",
            "To reduce the number of features by combining correlated ones into bins.",
            "To handle missing values by placing them into a separate histogram bin."
        ],
        "answer": "To group continuous data into discrete bins, speeding up the split-finding process."
    },
    {
        "question": "The raw-data HGB model achieved R\u00b2 = 0.9449, nearly identical to the cleaned-data version. What does this demonstrate?",
        "options": [
            "That data cleaning is never necessary when using tree-based models.",
            "That tree-based models are robust to messy data and don't require extensive preprocessing.",
            "That the Ames dataset was already clean enough that no cleaning was needed.",
            "That HGB automatically performs the same cleaning steps as the notebook's cleaning function."
        ],
        "answer": "That tree-based models are robust to messy data and don't require extensive preprocessing."
    },
    {
        "question": "How does XGBoost's built-in regularization differ from Ridge and Lasso regularization?",
        "options": [
            "XGBoost applies penalties to leaf weights and tree complexity, not to linear coefficients.",
            "XGBoost only uses L1 regularization, while Ridge and Lasso use both L1 and L2.",
            "XGBoost applies regularization after training, while Ridge and Lasso apply it during training.",
            "XGBoost does not use regularization; it relies solely on tree depth limits."
        ],
        "answer": "XGBoost applies penalties to leaf weights and tree complexity, not to linear coefficients."
    },
    {
        "question": "Boosting is inherently sequential, making it hard to parallelize. How does XGBoost achieve speed improvements?",
        "options": [
            "It builds multiple trees in parallel and averages their predictions.",
            "It parallelizes the internal work of building each tree, such as feature sorting.",
            "It skips the residual calculation step for trees after the first 50.",
            "It uses a single CPU core but with a more efficient optimization algorithm."
        ],
        "answer": "It parallelizes the internal work of building each tree, such as feature sorting."
    },
    {
        "question": "How does XGBoost handle missing values at each split?",
        "options": [
            "It imputes missing values with the median before training begins.",
            "It removes any sample with a missing value from the training set.",
            "It tests sending missing values left or right and learns which direction reduces error.",
            "It assigns missing values to a separate leaf node that predicts the global average."
        ],
        "answer": "It tests sending missing values left or right and learns which direction reduces error."
    },
    {
        "question": "Feature importance in tree models measures how much each split reduced prediction error. What is a key limitation of this metric?",
        "options": [
            "It cannot handle categorical features, so their importance is always zero.",
            "It tells you what the model prioritizes but not how the feature affects the target.",
            "It only works for the top 5 features and ignores all remaining features.",
            "It requires the model to be retrained after importance scores are calculated."
        ],
        "answer": "It tells you what the model prioritizes but not how the feature affects the target."
    },
    {
        "question": "In the feature importance plot, Overall Qual accounts for 45.76% of the model's learning. What does this percentage mean?",
        "options": [
            "Overall Qual alone can predict 45.76% of house prices without any other features.",
            "Splits on Overall Qual were responsible for 45.76% of the total error reduction across all trees.",
            "45.76% of the trees in the ensemble used Overall Qual as their first split.",
            "Overall Qual has a coefficient of 0.4576 in the underlying linear approximation."
        ],
        "answer": "Splits on Overall Qual were responsible for 45.76% of the total error reduction across all trees."
    },
    {
        "question": "Gr Liv Area has a VIF of 118 (extreme multicollinearity) but only 2.98% feature importance. Why?",
        "options": [
            "Living area is genuinely unimportant for predicting house prices in Ames.",
            "XGBoost chose correlated features like Overall Qual for splits, leaving Gr Liv Area with little credit.",
            "The VIF calculation is only relevant for linear models and has no bearing on tree models.",
            "The feature importance calculation is biased against continuous features."
        ],
        "answer": "XGBoost chose correlated features like Overall Qual for splits, leaving Gr Liv Area with little credit."
    },
    {
        "question": "In the model comparison, XGBoost's single-split score was 0.9418 but its nested CV estimate was 0.9177. What explains this gap?",
        "options": [
            "The nested CV used a different scoring metric that is inherently more conservative.",
            "The single-split score was inflated because the same data influenced both tuning and evaluation.",
            "The nested CV used fewer trees, resulting in a less powerful model.",
            "The test set in the single-split approach contained easier-to-predict houses by chance."
        ],
        "answer": "The single-split score was inflated because the same data influenced both tuning and evaluation."
    },
    {
        "question": "The nested CV standard deviation for XGBoost was \u00b10.0195. What does this tell you?",
        "options": [
            "The model's performance varies by about 2 percentage points across different data splits.",
            "The hyperparameter search found 19.5% of the possible combinations to be optimal.",
            "The model is overfitting by approximately 1.95 percentage points on each fold.",
            "The inner loop's best score was 1.95 percentage points higher than the outer loop's score."
        ],
        "answer": "The model's performance varies by about 2 percentage points across different data splits."
    },
    {
        "question": "The notebook recommends starting with a Random Forest for most real-world problems. Why?",
        "options": [
            "Random Forests always produce the highest accuracy of any model.",
            "Random Forests are fast, robust, and hard to mess up compared to tuning boosting models.",
            "Random Forests are the only tree-based method that handles missing values natively.",
            "Random Forests provide the most interpretable feature importance scores."
        ],
        "answer": "Random Forests are fast, robust, and hard to mess up compared to tuning boosting models."
    },
    {
        "question": "What tradeoff do you accept when moving from linear regression to XGBoost?",
        "options": [
            "You gain interpretability but lose the ability to handle non-linear patterns.",
            "You gain predictive accuracy but lose the ability to explain individual coefficient effects.",
            "You gain speed but lose the ability to handle categorical features.",
            "You gain robustness to outliers but lose the ability to use cross-validation."
        ],
        "answer": "You gain predictive accuracy but lose the ability to explain individual coefficient effects."
    },
    {
        "question": "In the nested CV for XGBoost, the inner loop used 3 folds and the outer loop used 5 folds with 12 hyperparameter combinations. How many total model fits occurred?",
        "options": [
            "60, because 5 outer folds times 12 hyperparameter combinations equals 60.",
            "180, because 5 outer folds each run a full inner search of 12 combinations times 3 inner folds.",
            "15, because 5 outer folds times 3 inner folds equals 15 total evaluations.",
            "36, because 12 hyperparameter combinations times 3 inner folds equals 36."
        ],
        "answer": "180, because 5 outer folds each run a full inner search of 12 combinations times 3 inner folds."
    },
    {
        "question": "Central Air achieved 7.86% feature importance despite being a binary feature. Why might a binary feature score highly?",
        "options": [
            "Binary features are always more important than continuous features in tree models.",
            "The presence or absence of central air is a major price differentiator in the Ames market.",
            "Binary features are used more frequently as split points because they are easier to compute.",
            "The model automatically upweights binary features to compensate for their simplicity."
        ],
        "answer": "The presence or absence of central air is a major price differentiator in the Ames market."
    },
    {
        "question": "The notebook mentions that partial dependence plots or SHAP values are the 'natural next step' after feature importance. What problem do they solve?",
        "options": [
            "They identify which features should be removed from the model entirely.",
            "They reveal the direction and shape of a feature's effect on the prediction.",
            "They calculate the exact dollar value contribution of each feature.",
            "They determine the optimal hyperparameters for the model."
        ],
        "answer": "They reveal the direction and shape of a feature's effect on the prediction."
    },
    {
        "question": "Decision trees convert categorical features to numeric codes using .cat.codes. What risk does this introduce?",
        "options": [
            "The model treats categories as having an ordinal relationship that may not exist.",
            "The model loses all information about the categorical feature during conversion.",
            "The model becomes unable to handle new categories that appear in the test set.",
            "The model requires one-hot encoding to work correctly with numeric codes."
        ],
        "answer": "The model treats categories as having an ordinal relationship that may not exist."
    },
    {
        "question": "The notebook's final model comparison shows Decision Tree (0.83) < Random Forest (0.88) < HGB (0.94) < XGBoost (0.94). What does this progression illustrate?",
        "options": [
            "That more complex models always produce better results regardless of the dataset.",
            "That each ensemble method builds on the previous one's weaknesses to improve accuracy.",
            "That the differences between models are negligible and any tree model would suffice.",
            "That the hyperparameter grids were progressively larger, giving later models an unfair advantage."
        ],
        "answer": "That each ensemble method builds on the previous one's weaknesses to improve accuracy."
    },
]

# --- QUIZ ENGINE ---
class QuizApp:
    def __init__(self, data):
        self.data = data
        self.current_index = 0
        self.score = 0
        self.user_answers = []

        # UI Elements
        self.header = widgets.HTML(value="<h2>Tree-Based Methods Quiz</h2>")
        self.progress = widgets.HTML(value="")
        self.question_text = widgets.HTML(value="")
        self.options = widgets.RadioButtons(options=[], layout={'width': 'max-content'})
        self.submit_btn = widgets.Button(description="Submit Answer", button_style='primary')
        self.next_btn = widgets.Button(description="Next Question", button_style='info', disabled=True)
        self.feedback = widgets.HTML(value="")
        self.output = widgets.Output()

        self.submit_btn.on_click(self.handle_submit)
        self.next_btn.on_click(self.handle_next)

        self.render_question()

    def render_question(self):
        clear_output(wait=True)
        q = self.data[self.current_index]
        self.progress.value = f"<b>Question {self.current_index + 1} of {len(self.data)}</b>"
        self.question_text.value = f"<p style='font-size:14px;'>{q['question']}</p>"
        self.options.options = q['options']
        self.options.value = None
        self.submit_btn.disabled = False
        self.next_btn.disabled = True
        self.feedback.value = ""

        display(self.header, self.progress, self.question_text, self.options,
                widgets.HBox([self.submit_btn, self.next_btn]), self.feedback)

    def handle_submit(self, b):
        if self.options.value is None:
            self.feedback.value = "<span style='color:red;'>Please select an option!</span>"
            return

        correct_answer = self.data[self.current_index]['answer']
        if self.options.value == correct_answer:
            self.score += 1
            self.feedback.value = "<span style='color:green; font-weight:bold;'>\u2705 Correct!</span>"
        else:
            self.feedback.value = f"<span style='color:red; font-weight:bold;'>\u274c Incorrect.</span><br>The correct answer was: <b>{correct_answer}</b>"

        self.submit_btn.disabled = True
        self.next_btn.disabled = False

    def handle_next(self, b):
        self.current_index += 1
        if self.current_index < len(self.data):
            self.render_question()
        else:
            self.show_results()

    def show_results(self):
        clear_output()
        percent = (self.score / len(self.data)) * 100
        result_html = f"""
        <div style="border: 2px solid #0366d6; padding: 20px; border-radius: 10px;">
            <h2>Quiz Complete!</h2>
            <p style='font-size:16px;'>Your Final Score: <b>{self.score} / {len(self.data)}</b></p>
            <p style='font-size:20px; color:#0366d6;'><b>Percentage: {percent:.1f}%</b></p>
        </div>
        """
        display(widgets.HTML(result_html))

# Start the app
QuizApp(quiz_data)
```
