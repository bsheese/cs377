```python
import ipywidgets as widgets
from IPython.display import display, clear_output

all_quizzes = {
    "17_0_1: Spread & TSS": [
        {
            "question": "Why does the sum of raw deviations from the mean always equal zero?",
            "options": [
                "Deviations are undefined for values below the mean",
                "Positive and negative deviations cancel because the mean is the balance point of the data",
                "The mean is always zero, so all deviations are also zero",
                "Raw deviations are always positive when computed correctly"
            ],
            "answer": "Positive and negative deviations cancel because the mean is the balance point of the data"
        },
        {
            "question": "Why is squaring deviations preferred over taking their absolute values?",
            "options": [
                "Squaring is computationally faster than absolute value",
                "Squaring penalizes large errors more heavily and produces a differentiable function",
                "Absolute value cannot prevent the cancellation problem",
                "Squared deviations are always in the same units as the data"
            ],
            "answer": "Squaring penalizes large errors more heavily and produces a differentiable function"
        },
        {
            "question": "TSS for penguin body mass is computed in g². Why is standard deviation a better summary of spread?",
            "options": [
                "Standard deviation is always smaller than variance",
                "Taking the square root of variance returns the spread to the original units",
                "Variance cannot be compared across datasets; standard deviation can",
                "TSS grows with sample size; standard deviation does not"
            ],
            "answer": "Taking the square root of variance returns the spread to the original units"
        },
        {
            "question": "The mean is described as the 'baseline prediction that minimizes squared error.' What does this mean?",
            "options": [
                "Predicting the mean for every observation minimizes the sum of squared residuals",
                "The mean is the only statistic that equals zero when squared",
                "Baseline predictions are always equal to the sample mean by definition",
                "The mean minimizes prediction error only for normally distributed data"
            ],
            "answer": "Predicting the mean for every observation minimizes the sum of squared residuals"
        },
        {
            "question": "If a dataset has TSS = 0, what must be true?",
            "options": [
                "All values in the dataset are equal to each other",
                "The dataset has exactly one observation",
                "The mean of the dataset is exactly zero",
                "The standard deviation is undefined because of division by zero"
            ],
            "answer": "All values in the dataset are equal to each other"
        }
    ],
    "17_0_2: Visualizing Associations": [
        {
            "question": "Anscombe's Quartet has four datasets with nearly identical means, variances, and correlations. What does this demonstrate?",
            "options": [
                "Summary statistics fully capture the structure of any dataset",
                "Summary statistics can be identical while the underlying patterns are completely different",
                "Correlation is the most important summary statistic for bivariate data",
                "All linear regression models are equally accurate for the same summary statistics"
            ],
            "answer": "Summary statistics can be identical while the underlying patterns are completely different"
        },
        {
            "question": "Anscombe Dataset II is a perfect parabola. Fitting a linear model to it produces R² ≈ 0.67. What is the problem?",
            "options": [
                "R² = 0.67 is too low for any model to be useful",
                "The linear model misses the curved pattern; predictions will be systematically wrong outside the middle range",
                "A parabolic relationship cannot be detected without more than 11 data points",
                "The model is correct; R² = 0.67 means 67% of variance is explained by the line"
            ],
            "answer": "The linear model misses the curved pattern; predictions will be systematically wrong outside the middle range"
        },
        {
            "question": "What is the difference between a y-outlier and a high-leverage point?",
            "options": [
                "A y-outlier has an unusual x-value; a high-leverage point has an unusual y-value",
                "A y-outlier has an unusual y-value given x; a high-leverage point has an extreme x-value",
                "They describe the same type of point using different terminology",
                "y-outliers affect the slope; high-leverage points only affect the intercept"
            ],
            "answer": "A y-outlier has an unusual y-value given x; a high-leverage point has an extreme x-value"
        },
        {
            "question": "Dataset IV in Anscombe's Quartet has all x-values near 8 except one point at x=19. Why is this single point dangerous?",
            "options": [
                "It is a y-outlier that inflates the residual sum of squares",
                "It is a high-leverage point that can pull the slope toward it with little resistance",
                "The model ignores points far from the mean x-value",
                "It causes the intercept to equal zero regardless of other data"
            ],
            "answer": "It is a high-leverage point that can pull the slope toward it with little resistance"
        },
        {
            "question": "The Datasaurus Dozen shows 13 datasets — including a dinosaur shape — all with identical summary statistics. What is the key takeaway?",
            "options": [
                "Automated ML pipelines should always compute summary statistics before visualizing",
                "Visualization is essential; summary statistics alone cannot reveal the true data structure",
                "The Datasaurus proves that all datasets follow the same underlying distribution",
                "Summary statistics are only reliable when computed on more than 100 data points"
            ],
            "answer": "Visualization is essential; summary statistics alone cannot reveal the true data structure"
        }
    ],
    "17_0_3: Covariance & Correlation": [
        {
            "question": "In the quadrant trick, why do upper-right points contribute positively to covariance?",
            "options": [
                "Upper-right points have both x and y above the mean, so (x-x̄)(y-ȳ) = (+)(+) = +",
                "Upper-right points have the largest absolute values and dominate the calculation",
                "Positive quadrants are defined as any region above the x-axis",
                "Covariance counts the number of points per quadrant, weighting upper-right as positive"
            ],
            "answer": "Upper-right points have both x and y above the mean, so (x-x̄)(y-ȳ) = (+)(+) = +"
        },
        {
            "question": "Covariance between flipper length (mm) and mass (g) is 9,796 mm·g. In kg, it becomes 9.796. Is the relationship stronger in kg?",
            "options": [
                "Yes — a smaller covariance value indicates a weaker relationship",
                "No — the relationship strength is unchanged; only the units differ",
                "Yes — covariance in kg is always more accurate than in grams",
                "No — covariance cannot be meaningfully compared across unit systems"
            ],
            "answer": "No — the relationship strength is unchanged; only the units differ"
        },
        {
            "question": "Pearson's r = 0 for a dataset where y = x². Why doesn't this mean x and y are unrelated?",
            "options": [
                "r can only be computed for datasets with more than 30 observations",
                "r measures only linear association; a perfect quadratic relationship is non-linear",
                "y = x² would produce r = -1 because the curve is concave",
                "r cannot equal exactly 0 for any real dataset"
            ],
            "answer": "r measures only linear association; a perfect quadratic relationship is non-linear"
        },
        {
            "question": "Dividing covariance by the product of standard deviations (σ_x · σ_y) produces a unitless r. Why does this division cancel units?",
            "options": [
                "Dividing by standard deviations converts the data to z-scores in both dimensions",
                "Standard deviations are always equal to 1 in normalized data",
                "Multiplying two standard deviations produces the same units as the covariance, canceling them",
                "The product σ_x · σ_y equals the variance, which removes all units"
            ],
            "answer": "Multiplying two standard deviations produces the same units as the covariance, canceling them"
        },
        {
            "question": "r = 0.87 between flipper length and body mass. Does this mean flipper length causes heavier penguins?",
            "options": [
                "Yes — strong correlation always indicates a causal relationship",
                "No — correlation measures association; causation requires additional experimental evidence",
                "Only if the p-value for r is less than 0.05",
                "Yes — biological measurements are always causal"
            ],
            "answer": "No — correlation measures association; causation requires additional experimental evidence"
        }
    ],
    "17_0_4: Residuals": [
        {
            "question": "A residual is e_i = y_i - ŷ_i. What does a large positive residual indicate?",
            "options": [
                "The model overpredicted — it placed the line above the actual point",
                "The model underpredicted — the actual value is well above the line",
                "The point is a high-leverage observation that should be removed",
                "The prediction is exactly correct for that observation"
            ],
            "answer": "The model underpredicted — the actual value is well above the line"
        },
        {
            "question": "Any line passing through (x̄, ȳ) has residuals that sum to zero. Why can't you use this sum as a model quality metric?",
            "options": [
                "A sum of zero means the model predicted every point exactly",
                "Many terrible lines also have residuals summing to zero; positive and negative errors cancel",
                "Residuals from lines not through (x̄, ȳ) always sum to zero too",
                "The sum is always zero only for perfectly linear relationships"
            ],
            "answer": "Many terrible lines also have residuals summing to zero; positive and negative errors cancel"
        },
        {
            "question": "Residuals are measured vertically (in y-units) rather than perpendicularly to the line. Why?",
            "options": [
                "Perpendicular distances are undefined for lines with non-integer slopes",
                "We want to measure prediction error in the target variable's units, not geometric distance",
                "Vertical residuals are always smaller and easier to minimize",
                "Perpendicular residuals would make the resulting coefficients non-interpretable"
            ],
            "answer": "We want to measure prediction error in the target variable's units, not geometric distance"
        },
        {
            "question": "A residual plot shows a U-shaped curve (not random scatter). What does this tell you about the model?",
            "options": [
                "The model has high variance but low bias",
                "The model systematically misses a non-linear pattern in the data",
                "The residuals are normally distributed, which is desirable",
                "There are influential outliers that should be removed before refitting"
            ],
            "answer": "The model systematically misses a non-linear pattern in the data"
        }
    ],
    "17_0_5: OLS & R-squared": [
        {
            "question": "Plotting RSS vs. candidate slopes produces a U-shaped bowl with one minimum. Why exactly one minimum?",
            "options": [
                "RSS is a convex quadratic function of the slope, guaranteeing a single global minimum",
                "The dataset is too small to produce multiple local minima",
                "Grid search always converges to one answer regardless of the landscape shape",
                "The minimum is guaranteed only when the data passes through the origin"
            ],
            "answer": "RSS is a convex quadratic function of the slope, guaranteeing a single global minimum"
        },
        {
            "question": "The optimal slope formula is m = r · (σ_y / σ_x). If r = 0 (no correlation), what is the optimal slope?",
            "options": [
                "0 — the best prediction is always the mean of y, regardless of x",
                "σ_y / σ_x — the ratio of standard deviations is the slope",
                "Undefined — slope cannot be computed when correlation is zero",
                "1 — a unit slope is the safest default when r is unknown"
            ],
            "answer": "0 — the best prediction is always the mean of y, regardless of x"
        },
        {
            "question": "R² = 0.50 for the Ames Housing model. In plain English, what does this mean?",
            "options": [
                "The model predicts house prices with 50% accuracy",
                "The model explains 50% of the variance in price that existed before fitting the line",
                "50% of predictions are within one standard deviation of the true value",
                "The model's slope is exactly half the true relationship slope"
            ],
            "answer": "The model explains 50% of the variance in price that existed before fitting the line"
        },
        {
            "question": "R² is defined as 1 - RSS/TSS. What does R² = 0 tell you about the model?",
            "options": [
                "RSS = 0, meaning the model perfectly fits all data points",
                "RSS = TSS, meaning the model is no better than predicting the mean",
                "TSS = 0, meaning all y-values are identical",
                "The model has failed to converge and should be retrained"
            ],
            "answer": "RSS = TSS, meaning the model is no better than predicting the mean"
        },
        {
            "question": "For simple linear regression, R² = r². If r = 0.70, what is R²?",
            "options": [
                "0.70 — the correlation and R² are equal for linear models",
                "0.49 — R² equals the squared correlation",
                "0.35 — R² equals half the correlation for bivariate data",
                "1.40 — R² is two times the correlation in simple regression"
            ],
            "answer": "0.49 — R² equals the squared correlation"
        },
        {
            "question": "Can R² ever be negative? What would a negative R² mean?",
            "options": [
                "No — R² is bounded between 0 and 1 by mathematical definition",
                "Yes — the model performs worse than simply predicting the mean on test data",
                "Yes — it indicates that the slope is in the wrong direction",
                "No — RSS can never exceed TSS for a fitted model"
            ],
            "answer": "Yes — the model performs worse than simply predicting the mean on test data"
        }
    ]
}

class QuizApp:
    def __init__(self, quiz_name, data):
        self.quiz_name = quiz_name
        self.data = data
        self.current_index = 0
        self.score = 0

        self.header = widgets.HTML(value="<h2>" + quiz_name + "</h2>")
        self.progress = widgets.HTML(value="")
        self.question_text = widgets.HTML(value="")
        self.options = widgets.RadioButtons(options=[], layout={"width": "max-content"})
        self.submit_btn = widgets.Button(description="Submit Answer", button_style="primary")
        self.next_btn = widgets.Button(description="Next Question", button_style="info", disabled=True)
        self.back_btn = widgets.Button(description="Back to Menu", button_style="warning")
        self.feedback = widgets.HTML(value="")

        self.submit_btn.on_click(self.handle_submit)
        self.next_btn.on_click(self.handle_next)
        self.back_btn.on_click(self.handle_back)
        self.render_question()

    def render_question(self):
        clear_output(wait=True)
        q = self.data[self.current_index]
        self.progress.value = "<b>Question " + str(self.current_index + 1) + " of " + str(len(self.data)) + "</b>"
        self.question_text.value = "<p style='font-size:14px;'>" + q["question"] + "</p>"
        self.options.options = q["options"]
        self.options.value = None
        self.submit_btn.disabled = False
        self.next_btn.disabled = True
        self.feedback.value = ""
        display(self.header, self.progress, self.question_text, self.options,
                widgets.HBox([self.submit_btn, self.next_btn]), self.feedback)

    def handle_submit(self, b):
        if self.options.value is None:
            self.feedback.value = "<span style='color:red;'>Please select an option!</span>"
            return
        correct = self.data[self.current_index]["answer"]
        if self.options.value == correct:
            self.score += 1
            self.feedback.value = "<span style='color:green; font-weight:bold;'>Correct!</span>"
        else:
            self.feedback.value = "<span style='color:red; font-weight:bold;'>Incorrect.</span><br>The correct answer was: <b>" + correct + "</b>"
        self.submit_btn.disabled = True
        self.next_btn.disabled = False

    def handle_next(self, b):
        self.current_index += 1
        if self.current_index < len(self.data):
            self.render_question()
        else:
            self.show_results()

    def handle_back(self, b):
        show_menu()

    def show_results(self):
        clear_output()
        pct = (self.score / len(self.data)) * 100
        html = """<div style="border: 2px solid #0366d6; padding: 20px; border-radius: 10px;">
            <h2>Quiz Complete!</h2>
            <p style="font-size:16px;">Your Final Score: <b>""" + str(self.score) + " / " + str(len(self.data)) + """</b></p>
            <p style="font-size:20px; color:#0366d6;"><b>Percentage: """ + "{:.1f}".format(pct) + "%</b></p></div>"""
        display(widgets.HTML(html), self.back_btn)

def show_menu():
    clear_output(wait=True)
    header = widgets.HTML(value="<h2>17_0 Regression Preliminaries — Practice Quizzes</h2>")
    instruction = widgets.HTML(value="<p>Select a notebook quiz below to begin.</p>")
    quiz_selector = widgets.Dropdown(options=list(all_quizzes.keys()), description="Select Quiz:", layout={"width": "50%"})
    start_btn = widgets.Button(description="Start Quiz", button_style="success")

    def on_start(b):
        QuizApp(quiz_selector.value, all_quizzes[quiz_selector.value])

    start_btn.on_click(on_start)
    display(header, instruction, quiz_selector, start_btn)

show_menu()
```
