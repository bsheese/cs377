# Beyond Additive Models
## Case Study: Interaction Effects in Medical Costs

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols

# 1. Load the data directly from the source
url = "https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/insurance.csv"
df = pd.read_csv(url)

# 2. Visualization: The "Aha!" Moment
# This plot shows that the 'slope' for smokers is much steeper than for non-smokers
sns.set_style('whitegrid')
sns.lmplot(x="bmi", y="charges", hue="smoker", data=df, aspect=1.5, markers=["o", "x"], palette="Set1")
plt.title("Visualizing the Interaction: BMI vs Medical Charges")
plt.show()

# 3. Model 1: Additive (Assuming BMI and Smoking are independent effects)
# Formula: charges = Intercept + BMI + Smoker
model_additive = ols('charges ~ bmi + smoker', data=df).fit()

# 4. Model 2: Interaction (Allowing the effect of BMI to change based on Smoking)
# The '*' operator includes bmi, smoker, AND the interaction term bmi:smoker
model_interaction = ols('charges ~ bmi * smoker', data=df).fit()

# 5. Compare the results
print("--- ADDITIVE MODEL (No Interaction) ---")
print(f"R-squared: {model_additive.rsquared:.4f}")
print(f"AIC:       {model_additive.aic:.2f}")

print("\n--- INTERACTION MODEL ---")
print(f"R-squared: {model_interaction.rsquared:.4f}")
print(f"AIC:       {model_interaction.aic:.2f}")

print("\n--- INTERACTION MODEL COEFFICIENTS ---")
print(model_interaction.params)

# 6. Teaching Note / Interpretation:
# The 'bmi:smoker[T.yes]' coefficient tells you how much the SLOPE increases 
# for smokers specifically.
```

### What to expect when you run this:
1.  **R-squared Jump:** The $R^2$ increases from roughly **0.66** in the additive model to **0.74** in the interaction model — a solid jump for adding a single term (and AIC drops, agreeing that the extra term earns its keep).
2.  **The Slopes:**
    *   The coefficient for `bmi` is about **83**: for **non-smokers**, each additional BMI point is associated with roughly \$83 of extra annual charges — almost nothing on this scale.
    *   The interaction coefficient `bmi:smoker[T.yes]` is about **1,390**: for **smokers**, each BMI point adds an *additional* \$1,390 on top of the base slope — making the smoker slope (≈ \$1,470 per BMI point) nearly 18× steeper than the non-smoker slope.
3.  **The Plot:** `sns.lmplot` shows two lines: one nearly flat (non-smokers) and one steeply climbing (smokers). The single additive model is forced to average these two regimes; the interaction model fits each one.

### Why this is a great teaching moment:
In the **Ames Housing** data, students often think "more variables = better model." This exercise shows them that **how variables relate to each other** (feature engineering/interactions) is often more powerful than simply adding more columns.
