```python
import ipywidgets as widgets
from IPython.display import display, clear_output

all_quizzes = {
    "17_3_1: Interaction Terms": [
        {
            "question": "The lmplot with hue='smoker' shows two lines with dramatically different slopes. What does this specifically indicate?",
            "options": [
                "Smoking is a confounding variable that should be excluded from the model",
                "The effect of BMI on charges differs between smokers and non-smokers — an interaction exists",
                "The data has heteroscedasticity that a transformation would fix",
                "Smokers and non-smokers should be modeled in completely separate datasets"
            ],
            "answer": "The effect of BMI on charges differs between smokers and non-smokers — an interaction exists"
        },
        {
            "question": "In the additive model 'charges ~ bmi + smoker', what does 'independent effects' mean?",
            "options": [
                "BMI and smoking are statistically uncorrelated in the dataset",
                "BMI's effect on charges is the same whether someone smokes or not",
                "Each predictor is independently significant in the model summary",
                "The model was fit on a random independent sample of patients"
            ],
            "answer": "BMI's effect on charges is the same whether someone smokes or not"
        },
        {
            "question": "In 'charges ~ bmi * smoker', three coefficients are produced. What does the 'bmi:smoker[T.yes]' coefficient represent?",
            "options": [
                "The base cost difference between smokers and non-smokers at BMI=0",
                "The additional per-BMI-unit cost increase for smokers above the non-smoker rate",
                "The total cost per BMI unit regardless of smoking status",
                "The probability that a high-BMI patient is a smoker"
            ],
            "answer": "The additional per-BMI-unit cost increase for smokers above the non-smoker rate"
        },
        {
            "question": "The non-smoker BMI coefficient is ~83 and the interaction coefficient is ~1,390. What is the total cost per BMI unit for a smoker?",
            "options": [
                "83 — the interaction term replaces the main effect for smokers",
                "1,390 — the main effect is irrelevant when an interaction is present",
                "~1,473 — the main effect plus the interaction coefficient",
                "737 — the average of the main effect and the interaction term"
            ],
            "answer": "~1,473 — the main effect plus the interaction coefficient"
        },
        {
            "question": "Why is 'bmi * smoker' preferable to manually writing 'bmi + smoker + bmi:smoker' in the formula?",
            "options": [
                "The * syntax runs faster because it skips the main effects calculation",
                "The * syntax ensures main effects are always included with the interaction term",
                "Manually specifying terms causes statsmodels to double-count the main effects",
                "The * syntax automatically centers BMI to reduce multicollinearity"
            ],
            "answer": "The * syntax ensures main effects are always included with the interaction term"
        },
        {
            "question": "R² increases from ~0.66 (additive) to ~0.74 (interaction). Is R² improvement alone sufficient justification for the interaction model?",
            "options": [
                "Yes — any R² improvement justifies adding the interaction term",
                "No — R² always increases with more terms; AIC and domain logic are also needed",
                "Yes — only R² measures out-of-sample predictive accuracy",
                "No — R² improvements only matter when they exceed 50 percentage points"
            ],
            "answer": "No — R² always increases with more terms; AIC and domain logic are also needed"
        },
        {
            "question": "If the 'bmi:smoker[T.yes]' interaction coefficient were exactly zero, what would this tell you?",
            "options": [
                "BMI is unimportant for predicting charges in this model",
                "The BMI slope is the same for smokers and non-smokers — no interaction exists",
                "Smoking has no effect on charges when controlling for BMI",
                "The interaction term was not properly encoded in the formula"
            ],
            "answer": "The BMI slope is the same for smokers and non-smokers — no interaction exists"
        },
        {
            "question": "AIC (Akaike Information Criterion) penalizes model complexity. If both models have similar AIC, what does this suggest?",
            "options": [
                "The interaction model is always preferred when AIC values are equal",
                "The complexity of the interaction term does not pay off sufficiently in fit improvement",
                "AIC cannot distinguish between additive and interaction models",
                "Both models will produce identical predictions on test data"
            ],
            "answer": "The complexity of the interaction term does not pay off sufficiently in fit improvement"
        },
        {
            "question": "A student writes 'charges ~ bmi:smoker' without including main effects. What is the problem with this model?",
            "options": [
                "The interaction operator : is not valid without the * operator",
                "The model has no term for smokers who have BMI=0, making the intercept uninterpretable",
                "Statsmodels automatically adds main effects when interaction terms are specified",
                "The model would overfit because interactions always require large sample sizes"
            ],
            "answer": "The model has no term for smokers who have BMI=0, making the intercept uninterpretable"
        },
        {
            "question": "An insurance company builds an additive model (no interaction) for pricing. Why could this lead to unfair pricing?",
            "options": [
                "Additive models always overcharge non-smokers and undercharge smokers",
                "The model would charge smokers and non-smokers the same rate per BMI unit, ignoring the true cost difference",
                "Additive models cannot include categorical variables like smoker status",
                "The model would ignore BMI entirely when smoker is in the model"
            ],
            "answer": "The model would charge smokers and non-smokers the same rate per BMI unit, ignoring the true cost difference"
        }
    ]
}

class QuizApp:
    def __init__(self, quiz_name, data):
        self.quiz_name = quiz_name
        self.data = data
        self.current_index = 0
        self.score = 0

        self.header = widgets.HTML(value="<h2>" + quiz_name + "</h2>")
        self.progress = widgets.HTML(value="")
        self.question_text = widgets.HTML(value="")
        self.options = widgets.RadioButtons(options=[], layout={"width": "max-content"})
        self.submit_btn = widgets.Button(description="Submit Answer", button_style="primary")
        self.next_btn = widgets.Button(description="Next Question", button_style="info", disabled=True)
        self.back_btn = widgets.Button(description="Back to Menu", button_style="warning")
        self.feedback = widgets.HTML(value="")

        self.submit_btn.on_click(self.handle_submit)
        self.next_btn.on_click(self.handle_next)
        self.back_btn.on_click(self.handle_back)
        self.render_question()

    def render_question(self):
        clear_output(wait=True)
        q = self.data[self.current_index]
        self.progress.value = "<b>Question " + str(self.current_index + 1) + " of " + str(len(self.data)) + "</b>"
        self.question_text.value = "<p style='font-size:14px;'>" + q["question"] + "</p>"
        self.options.options = q["options"]
        self.options.value = None
        self.submit_btn.disabled = False
        self.next_btn.disabled = True
        self.feedback.value = ""
        display(self.header, self.progress, self.question_text, self.options,
                widgets.HBox([self.submit_btn, self.next_btn]), self.feedback)

    def handle_submit(self, b):
        if self.options.value is None:
            self.feedback.value = "<span style='color:red;'>Please select an option!</span>"
            return
        correct = self.data[self.current_index]["answer"]
        if self.options.value == correct:
            self.score += 1
            self.feedback.value = "<span style='color:green; font-weight:bold;'>Correct!</span>"
        else:
            self.feedback.value = "<span style='color:red; font-weight:bold;'>Incorrect.</span><br>The correct answer was: <b>" + correct + "</b>"
        self.submit_btn.disabled = True
        self.next_btn.disabled = False

    def handle_next(self, b):
        self.current_index += 1
        if self.current_index < len(self.data):
            self.render_question()
        else:
            self.show_results()

    def handle_back(self, b):
        show_menu()

    def show_results(self):
        clear_output()
        pct = (self.score / len(self.data)) * 100
        html = """<div style="border: 2px solid #0366d6; padding: 20px; border-radius: 10px;">
            <h2>Quiz Complete!</h2>
            <p style="font-size:16px;">Your Final Score: <b>""" + str(self.score) + " / " + str(len(self.data)) + """</b></p>
            <p style="font-size:20px; color:#0366d6;"><b>Percentage: """ + "{:.1f}".format(pct) + "%</b></p></div>"""
        display(widgets.HTML(html), self.back_btn)

def show_menu():
    clear_output(wait=True)
    header = widgets.HTML(value="<h2>17_3 Interaction Terms — Practice Quiz</h2>")
    instruction = widgets.HTML(value="<p>Select a notebook quiz below to begin.</p>")
    quiz_selector = widgets.Dropdown(options=list(all_quizzes.keys()), description="Select Quiz:", layout={"width": "50%"})
    start_btn = widgets.Button(description="Start Quiz", button_style="success")

    def on_start(b):
        QuizApp(quiz_selector.value, all_quizzes[quiz_selector.value])

    start_btn.on_click(on_start)
    display(header, instruction, quiz_selector, start_btn)

show_menu()
```
