```python
# @title
import ipywidgets as widgets
from IPython.display import display, clear_output

# --- ALL QUIZ DATA ---
all_quizzes = {
    "17_2_4_1: Data Cleaning": [
        {
            "question": "In data cleaning, what is the primary conceptual reason for removing 'extreme outliers' (like very large houses sold for very low prices) before building a linear model?",
            "options": [
                "To increase the size of the dataset",
                "To prevent a few unusual observations from disproportionately skewing the regression line",
                "To ensure the model can only predict luxury home prices",
                "Because linear regression cannot mathematically process large numbers"
            ],
            "answer": "To prevent a few unusual observations from disproportionately skewing the regression line"
        },
        {
            "question": "Why is it conceptually important to remove unique identifiers like 'PID' or 'Order' from a machine learning featureset?",
            "options": [
                "They are too large to store in memory",
                "They are strings and linear regression requires numbers",
                "They provide no generalizable information and could lead to 'memorizing' specific rows",
                "They are always perfectly correlated with the target variable"
            ],
            "answer": "They provide no generalizable information and could lead to 'memorizing' specific rows"
        },
        {
            "question": "What is the statistical problem with a 'monotonic' feature (a column where every single value is the same)?",
            "options": [
                "It has zero variance and therefore provides no predictive information",
                "It will cause the model to crash due to 'division by zero'",
                "It makes the R-squared value too high to be realistic",
                "It is considered a 'yolked' variable"
            ],
            "answer": "It has zero variance and therefore provides no predictive information"
        },
        {
            "question": "The notebook describes 'yolked features' (e.g., Pool Quality and Pool Area). What is the primary concern when these two features are included in a model together?",
            "options": [
                "They represent redundant information where one value is structurally dependent on the other",
                "They are always missing from the dataset at the same time",
                "They are measured in different units",
                "They cause the target variable to become right-skewed"
            ],
            "answer": "They represent redundant information where one value is structurally dependent on the other"
        },
        {
            "question": "If a house has 'None' for a basement category and '0' for basement square footage, what does this tell us about the relationship between these features?",
            "options": [
                "The data is corrupted",
                "The categorical 'None' is a structural prerequisite for the numerical '0'",
                "The two variables are independent",
                "One is an outlier of the other"
            ],
            "answer": "The categorical 'None' is a structural prerequisite for the numerical '0'"
        },
        {
            "question": "Why would a data scientist choose to drop a feature like 'Pool Area' if 99% of the entries are zero?",
            "options": [
                "Because zeros cannot be used in a linear regression",
                "To reduce 'noise' and complexity in the model for a very small gain in predictive power",
                "Because the model only works for houses with pools",
                "To ensure the dataset is perfectly balanced"
            ],
            "answer": "To reduce 'noise' and complexity in the model for a very small gain in predictive power"
        },
        {
            "question": "In the context of 'Garage Yr Blt', why is it dangerous to simply replace missing values with '0'?",
            "options": [
                "Because a year of '0' acts as a massive outlier that distorts the temporal logic of the model",
                "Because Python does not recognize 0 as a valid year",
                "Because it will make the house appear newer than it actually is",
                "Because 0 is reserved for categorical data only"
            ],
            "answer": "Because a year of '0' acts as a massive outlier that distorts the temporal logic of the model"
        },
        {
            "question": "What is the conceptual goal of 'collapsing' a categorical variable with 10 categories into just 'Category A' and 'Other'?",
            "options": [
                "To increase the number of features in the model",
                "To simplify the model and prevent 'sparsity' (too many columns with mostly zeros)",
                "To make the data more difficult to interpret",
                "To ensure the model is non-linear"
            ],
            "answer": "To simplify the model and prevent 'sparsity' (too many columns with mostly zeros)"
        },
        {
            "question": "Why do we check for 'unbalanced' categoricals (e.g., where 95% of houses are on a 'Paved Street')?",
            "options": [
                "Because we want to find the most popular type of house",
                "Because variables with very little variance provide almost no 'signal' for the model to learn from",
                "Because linear regression requires a 50/50 split in all categories",
                "To identify which houses are the most expensive"
            ],
            "answer": "Because variables with very little variance provide almost no 'signal' for the model to learn from"
        },
        {
            "question": "Consider the code: `df['garage_attached'] = np.where(df['Garage Type'] == 'Attchd', 1, 0)`. What is the logical purpose of this transformation?",
            "options": [
                "To delete the 'Garage Type' column",
                "To create a binary indicator (dummy variable) focusing on one specific attribute",
                "To calculate the average price of attached garages",
                "To normalize the garage area"
            ],
            "answer": "To create a binary indicator (dummy variable) focusing on one specific attribute"
        },
        {
            "question": "What is the primary benefit of converting binary string columns (like 'Yes'/'No') into the 'boolean' or 'bool' data type?",
            "options": [
                "It makes the file size larger",
                "It clearly communicates to the model (and the programmer) that the data is a logical toggle",
                "It allows the model to handle missing values automatically",
                "It prevents the data from being used in a heatmap"
            ],
            "answer": "It clearly communicates to the model (and the programmer) that the data is a logical toggle"
        },
        {
            "question": "In the code `df.dropna(how='all')`, what is the specific condition required to remove a row?",
            "options": [
                "The row must have at least one missing value",
                "The row must be entirely composed of missing values",
                "The row must be a duplicate of another row",
                "The row must have a missing Sale Price"
            ],
            "answer": "The row must be entirely composed of missing values"
        },
        {
            "question": "When dealing with 'unbalanced' numeric data (e.g., a feature that is 85% zero), why might we convert it to a binary '0 or 1' instead of leaving it as a continuous number?",
            "options": [
                "To make the model more complex",
                "Because the presence/absence of the feature is likely more informative than the specific amount",
                "Because linear regression cannot handle the number 0",
                "To ensure the mean and median are the same"
            ],
            "answer": "Because the presence/absence of the feature is likely more informative than the specific amount"
        },
        {
            "question": "What is the conceptual logic behind using 'median replacement' for missing numeric data instead of 'mean replacement'?",
            "options": [
                "The median is always higher than the mean",
                "The median is more robust to outliers and better represents the 'typical' value in skewed data",
                "The mean is mathematically impossible to calculate with missing data",
                "The median is only used for categorical data"
            ],
            "answer": "The median is more robust to outliers and better represents the 'typical' value in skewed data"
        },
        {
            "question": "If we planned to use a more complex model (like a Neural Network) instead of Multiple Linear Regression, how might our data cleaning strategy change?",
            "options": [
                "We would probably remove even more data",
                "We might leave in more features and perform less aggressive 'simplification'",
                "We would only use categorical data",
                "We would never remove outliers"
            ],
            "answer": "We might leave in more features and perform less aggressive 'simplification'"
        },
        {
            "question": "What is the risk of having too many 'one-hot encoded' columns in a linear model?",
            "options": [
                "The model will become too simple",
                "The model may suffer from the 'curse of dimensionality,' making it harder to find a stable solution",
                "The model will automatically delete the target variable",
                "The R-squared will always be negative"
            ],
            "answer": "The model may suffer from the 'curse of dimensionality,' making it harder to find a stable solution"
        },
        {
            "question": "Why is the `safe_drop` function used instead of just calling `df.drop()`?",
            "options": [
                "It makes the code run significantly faster",
                "It prevents the code from crashing if you run the same cell twice or if a column is already gone",
                "It automatically saves a copy of the dropped data to the hard drive",
                "It only drops rows, never columns"
            ],
            "answer": "It prevents the code from crashing if you run the same cell twice or if a column is already gone"
        },
        {
            "question": "The notebook mentions 'Structural Multicollinearity' when discussing yolked variables. Conceptually, what is multicollinearity?",
            "options": [
                "When the target variable is normally distributed",
                "When independent variables (features) are highly correlated with each other",
                "When the model has too many rows",
                "When the errors are random"
            ],
            "answer": "When independent variables (features) are highly correlated with each other"
        },
        {
            "question": "When the code uses `df.select_dtypes(include=np.number)`, what is it attempting to isolate?",
            "options": [
                "Only the target variable",
                "Columns that contain continuous or discrete quantitative data",
                "Columns that are categorical in nature",
                "Columns that have no missing values"
            ],
            "answer": "Columns that contain continuous or discrete quantitative data"
        },
        {
            "question": "Why is 'Year Sold' or 'Month Sold' sometimes considered an 'unbalanced' or 'uninformative' categorical variable in a housing dataset?",
            "options": [
                "Because people only buy houses in the summer",
                "Because if the distribution is nearly even or lacks a strong trend, it adds complexity without predictive gain",
                "Because years are numbers and cannot be categories",
                "Because the model already knows the current year"
            ],
            "answer": "Because if the distribution is nearly even or lacks a strong trend, it adds complexity without predictive gain"
        },
        {
            "question": "What does the phrase 'Data cleaning is dictated by the model choice' mean in this context?",
            "options": [
                "All models require the exact same cleaning steps",
                "The assumptions and limitations of Multiple Linear Regression require specific data shapes (like linearity and lack of noise)",
                "The data is cleaned by the computer automatically once a model is chosen",
                "You should only clean data after the model is finished"
            ],
            "answer": "The assumptions and limitations of Multiple Linear Regression require specific data shapes (like linearity and lack of noise)"
        },
        {
            "question": "When collapsing 'Foundation' into 'PConc', 'CBlock', and 'Other', why group the less frequent foundations into 'Other'?",
            "options": [
                "Because they are probably errors in the data",
                "To ensure the model has enough observations in each category to find a reliable coefficient",
                "To make the houses in the 'Other' category more expensive",
                "Because linear regression can only handle three categories maximum"
            ],
            "answer": "To ensure the model has enough observations in each category to find a reliable coefficient"
        },
        {
            "question": "What is the purpose of using `value_counts(normalize=True)` during the cleaning process?",
            "options": [
                "To see the raw count of every value",
                "To see the proportion (percentage) of the data belonging to each category",
                "To find the mean of a categorical variable",
                "To identify and remove duplicates"
            ],
            "answer": "To see the proportion (percentage) of the data belonging to each category"
        },
        {
            "question": "If a feature is 90% 'Standard Circuit' and 10% other values, why would a modeler consider dropping it?",
            "options": [
                "Standard circuit breakers are not important for house prices",
                "The lack of variation means the feature likely won't help the model differentiate between expensive and cheap houses",
                "Because 10% is too much missing data",
                "Because categorical data must be 100% consistent"
            ],
            "answer": "The lack of variation means the feature likely won't help the model differentiate between expensive and cheap houses"
        },
        {
            "question": "In the context of the Ames dataset, what does 'Gr Liv Area' represent conceptually?",
            "options": [
                "Green Living Area (eco-friendly space)",
                "Gross Living Area (above-grade square footage)",
                "Ground Level Area only",
                "Garage and Living Area combined"
            ],
            "answer": "Gross Living Area (above-grade square footage)"
        },
        {
            "question": "Why is it important to look at the `.info()` output's 'Non-Null Count'?",
            "options": [
                "To determine how many features are in the dataset",
                "To identify which columns have missing data that require imputation or dropping",
                "To see the memory usage of the dataframe",
                "To check the data types of the target variable"
            ],
            "answer": "To identify which columns have missing data that require imputation or dropping"
        },
        {
            "question": "The notebook describes dropping 'Misc Feature' and 'Misc Val.' What is the trade-off being made here?",
            "options": [
                "Trading accuracy for speed",
                "Trading potential niche information for a simpler, less 'noisy' model",
                "Trading categorical data for numerical data",
                "Trading training data for test data"
            ],
            "answer": "Trading potential niche information for a simpler, less 'noisy' model"
        },
        {
            "question": "Why is it standard practice to perform outlier removal *before* calculating medians for imputation?",
            "options": [
                "Outliers can skew the median (though less than the mean), so removing them first ensures a more representative 'typical' value",
                "The median cannot be calculated if outliers are present",
                "To increase the number of missing values",
                "Because the author recommended it"
            ],
            "answer": "Outliers can skew the median (though less than the mean), so removing them first ensures a more representative 'typical' value"
        },
        {
            "question": "What is the primary risk of 'over-cleaning' your data (dropping too many columns)?",
            "options": [
                "The model will become too complex to understand",
                "You might remove a feature that actually has a strong, unique relationship with the target variable",
                "The computer will run out of memory",
                "The dataset will become too large"
            ],
            "answer": "You might remove a feature that actually has a strong, unique relationship with the target variable"
        },
        {
            "question": "What does the code `df.loc[~df['Foundation'].isin(['PConc', 'CBlock']), 'Foundation'] = 'Other'` do?",
            "options": [
                "It deletes all foundations that are 'PConc' or 'CBlock'",
                "It labels any foundation type that is NOT 'PConc' or 'CBlock' as 'Other'",
                "It creates a new dataframe containing only 'Other' foundations",
                "It checks if 'Foundation' is a numeric column"
            ],
            "answer": "It labels any foundation type that is NOT 'PConc' or 'CBlock' as 'Other'"
        },
    ],
    "17_2_4_2: Forward/Backward Selection": [
        {
            "question": "What is the primary purpose of 'One-Hot Encoding' in the context of the Ames dataset?",
            "options": [
                "To normalize numeric features",
                "To convert categorical text into numeric 1s and 0s",
                "To handle outliers in the target variable",
                "To reduce the number of features"
            ],
            "answer": "To convert categorical text into numeric 1s and 0s"
        },
        {
            "question": "Why is 'drop_first=True' critical when creating dummy variables for linear models?",
            "options": [
                "It prevents the 'Dummy Variable Trap' (perfect multicollinearity)",
                "It makes the model run faster",
                "It automatically log-transformed the data",
                "It removes rows with missing values"
            ],
            "answer": "It prevents the 'Dummy Variable Trap' (perfect multicollinearity)"
        },
        {
            "question": "In the provided code, why is a 'Pipeline' used during cross-validation?",
            "options": [
                "To speed up the OLS algorithm",
                "To prevent data leakage by ensuring scaling is only fit on training folds",
                "To allow the use of categorical data without encoding",
                "To automatically calculate VIF scores"
            ],
            "answer": "To prevent data leakage by ensuring scaling is only fit on training folds"
        },
        {
            "question": "What does Scikit-Learn's 'SequentialFeatureSelector' do in 'forward' mode?",
            "options": [
                "Removes the worst features one by one",
                "Starts empty and adds the best feature at each step",
                "Checks all possible combinations of features simultaneously",
                "Only selects features with a VIF below 5"
            ],
            "answer": "Starts empty and adds the best feature at each step"
        },
        {
            "question": "Why would you use only 2 or 3 cross-validation folds instead of 10 during model development?",
            "options": [
                "It is more accurate for large datasets",
                "To reduce computation time during expensive feature selection",
                "Because OLS cannot handle more than 5 folds",
                "To ensure the R-squared value is higher"
            ],
            "answer": "To reduce computation time during expensive feature selection"
        },
        {
            "question": "If 'Bedroom AbvGr' has a negative coefficient while 'Gr Liv Area' is held constant, what is the interpretation?",
            "options": [
                "Houses with more bedrooms are always cheaper",
                "More bedrooms in the same square footage leads to smaller, less desirable rooms",
                "The model is fundamentally broken",
                "Bedrooms have no impact on house price"
            ],
            "answer": "More bedrooms in the same square footage leads to smaller, less desirable rooms"
        },
        {
            "question": "In a 'Residuals vs. Predicted Values' plot, what does a 'fan' or 'funnel' shape indicate?",
            "options": [
                "The model is perfect",
                "Heteroscedasticity (non-constant error variance)",
                "The features are highly correlated",
                "The target variable is normally distributed"
            ],
            "answer": "Heteroscedasticity (non-constant error variance)"
        },
        {
            "question": "What is the primary goal of log-transforming a right-skewed target variable?",
            "options": [
                "To make the coefficients larger",
                "To handle non-linear features",
                "To squash extreme values and make the distribution more 'normal'",
                "To eliminate the need for a constant intercept"
            ],
            "answer": "To squash extreme values and make the distribution more 'normal'"
        },
        {
            "question": "In the context of OLS, why can we use a squared term ($X^2$) and still call it a 'linear' model?",
            "options": [
                "Because the resulting graph is a straight line",
                "Because the model is linear in its parameters (coefficients)",
                "Because $X^2$ is always a straight line",
                "Because we use a linear scaler"
            ],
            "answer": "Because the model is linear in its parameters (coefficients)"
        },
        {
            "question": "Adding a squared term ($X^2$) to a model allows the regression line to:",
            "options": [
                "Stay perfectly horizontal",
                "Bend exactly once (forming a parabola)",
                "Become a step-function",
                "Ignore all outliers"
            ],
            "answer": "Bend exactly once (forming a parabola)"
        },
        {
            "question": "What does a negative coefficient for a squared term (e.g., $-X^2$) typically represent in real estate data?",
            "options": [
                "Exponential growth",
                "Diminishing returns (curve flattens out as $X$ increases)",
                "A perfectly linear relationship",
                "Data corruption"
            ],
            "answer": "Diminishing returns (curve flattens out as $X$ increases)"
        },
        {
            "question": "Why is it dangerous to use high-degree polynomials (e.g., $X^5$ or $X^{10}$)?",
            "options": [
                "The model becomes too simple",
                "It leads to overfitting and a 'wiggly' line that predicts poorly on new data",
                "OLS cannot solve equations above degree 2",
                "The R-squared will always be 0"
            ],
            "answer": "It leads to overfitting and a 'wiggly' line that predicts poorly on new data"
        },
        {
            "question": "What is 'structural multicollinearity' in polynomial regression?",
            "options": [
                "Correlation caused by the inherent nature of the raw data",
                "Correlation between a base feature ($X$) and its derived square ($X^2$)",
                "Correlation between the target and the features",
                "Missing data in the categorical columns"
            ],
            "answer": "Correlation between a base feature ($X$) and its derived square ($X^2$)"
        },
        {
            "question": "What is a common technique to reduce structural multicollinearity before squaring a feature?",
            "options": [
                "Log-transforming the feature",
                "Centering or standardizing the feature",
                "Deleting the original linear feature",
                "Removing the intercept"
            ],
            "answer": "Centering or standardizing the feature"
        },
        {
            "question": "What does the Variance Inflation Factor (VIF) measure?",
            "options": [
                "The accuracy of the model",
                "How much the variance of a coefficient is 'inflated' due to correlation with other predictors",
                "The number of outliers in the data",
                "The speed of the forward selection process"
            ],
            "answer": "How much the variance of a coefficient is 'inflated' due to correlation with other predictors"
        },
        {
            "question": "Which VIF score is generally considered a 'red flag' indicating problematic multicollinearity?",
            "options": [
                "VIF = 0",
                "VIF = 1",
                "VIF > 5 or 10",
                "VIF < 2"
            ],
            "answer": "VIF > 5 or 10"
        },
        {
            "question": "If 'Overall Qual' and 'Overall Qual Squared' are both positive, the curve represents:",
            "options": [
                "Diminishing returns",
                "Accelerating/Exponential-like growth",
                "A straight horizontal line",
                "A downward trend"
            ],
            "answer": "Accelerating/Exponential-like growth"
        },
        {
            "question": "In the Ames analysis, 'Gr Liv Area' had a VIF of ~118. What does this suggest?",
            "options": [
                "It is the only important variable",
                "It is highly redundant with other square-footage variables",
                "The data was collected incorrectly",
                "It is perfectly independent of other variables"
            ],
            "answer": "It is highly redundant with other square-footage variables"
        },
        {
            "question": "What is the 'ideal' VIF score for a completely independent feature?",
            "options": [
                "0",
                "1",
                "10",
                "-1"
            ],
            "answer": "1"
        },
        {
            "question": "When interpreting a regression heatmap, a value of 1.0 between two independent variables suggests:",
            "options": [
                "Perfect independence",
                "Perfect multicollinearity",
                "A negative relationship",
                "Normal distribution"
            ],
            "answer": "Perfect multicollinearity"
        },
        {
            "question": "If forward selection excludes a polynomial term you expected to be important, why might that happen?",
            "options": [
                "Forward selection doesn't support math",
                "Other features already capture that predictive power",
                "The R-squared is already too high",
                "Polynomials can only be used with backward selection"
            ],
            "answer": "Other features already capture that predictive power"
        },
        {
            "question": "Why is 'StandardScaler' applied inside the Pipeline rather than to the whole dataset at once?",
            "options": [
                "To make the data look better",
                "To prevent data leakage from the test/validation folds into the training process",
                "To make sure the target variable is scaled",
                "Because Scikit-Learn requires it"
            ],
            "answer": "To prevent data leakage from the test/validation folds into the training process"
        },
        {
            "question": "A 'hockey stick' curve in house prices (upward bend) suggests that:",
            "options": [
                "Quality increases don't matter",
                "High-end luxury finishes command an exponential premium",
                "The house is losing value over time",
                "The model has high bias"
            ],
            "answer": "High-end luxury finishes command an exponential premium"
        },
        {
            "question": "In the 'run_modeling_workflow' function, what is 'random_state=42' used for?",
            "options": [
                "To improve the R-squared",
                "To ensure the train-test split is reproducible",
                "To shuffle the columns",
                "To identify outliers"
            ],
            "answer": "To ensure the train-test split is reproducible"
        },
        {
            "question": "If your model has high R-squared on training data but low R-squared on test data, it is:",
            "options": [
                "Underfitting",
                "Overfitting",
                "Perfectly Balanced",
                "Biased"
            ],
            "answer": "Overfitting"
        },
        {
            "question": "Which metric does the provided code use to evaluate the error of the model?",
            "options": [
                "R-squared",
                "Mean Squared Error (MSE)",
                "Mean Absolute Error (MAE)",
                "VIF"
            ],
            "answer": "Mean Squared Error (MSE)"
        },
        {
            "question": "Backward selection starts with:",
            "options": [
                "No features",
                "All features",
                "A random set of features",
                "Only categorical features"
            ],
            "answer": "All features"
        },
        {
            "question": "Why is 'Gr Liv Area' often the strongest predictor in the Ames dataset?",
            "options": [
                "Because it's a log-transform",
                "Because square footage is a primary driver of home value",
                "Because it has the lowest VIF",
                "Because it was one-hot encoded"
            ],
            "answer": "Because square footage is a primary driver of home value"
        },
        {
            "question": "When a residual plot shows random scatter around the zero line, it suggests:",
            "options": [
                "The model is failing",
                "The linear assumptions are appropriate",
                "There is high multicollinearity",
                "The model is overfitting"
            ],
            "answer": "The linear assumptions are appropriate"
        },
        {
            "question": "What is the primary advantage of 'Forward Selection' over 'All Possible Subsets'?",
            "options": [
                "It is more accurate",
                "It is computationally much faster",
                "It ignores multicollinearity",
                "It doesn't require a target variable"
            ],
            "answer": "It is computationally much faster"
        },
    ],
    "17_2_4_3: Regularization": [
        {
            "question": "What is the primary objective of introducing regularization into a regression model?",
            "options": [
                "To find the absolute global minimum of the error function on training data.",
                "To penalize model complexity and discourage excessively large coefficients.",
                "To increase the weight of the most important features to improve fit.",
                "To eliminate the need for an intercept in the regression equation."
            ],
            "answer": "To penalize model complexity and discourage excessively large coefficients."
        },
        {
            "question": "In the context of standard OLS, why are 'massive weights' considered a problem?",
            "options": [
                "They make the model computationally too expensive to run.",
                "They cause the model to ignore random noise in the dataset.",
                "They make the model unstable, where tiny input changes cause massive prediction swings.",
                "They lead to an underfit model with high bias."
            ],
            "answer": "They make the model unstable, where tiny input changes cause massive prediction swings."
        },
        {
            "question": "How does regularization 'compromise' during the training process?",
            "options": [
                "It balances minimizing error with keeping feature weights small.",
                "It balances training speed with the number of features used.",
                "It balances the mean squared error with the R-squared value.",
                "It compromises between forward and backward selection."
            ],
            "answer": "It balances minimizing error with keeping feature weights small."
        },
        {
            "question": "Compared to stepwise selection (Forward/Backward), regularization is described as 'nuanced' because:",
            "options": [
                "It can only remove features that have zero correlation with the target.",
                "It is a binary process that decides if a feature is 100% in or out.",
                "It can shrink coefficients to very small numbers without abruptly muting them.",
                "It requires fewer observations than features to work."
            ],
            "answer": "It can shrink coefficients to very small numbers without abruptly muting them."
        },
        {
            "question": "Why is stepwise selection considered 'unstable' compared to regularized models?",
            "options": [
                "Because it always results in a model with the highest possible MSE.",
                "Because small changes in training data can lead to a completely different set of chosen features.",
                "Because it can only be used with polynomial features.",
                "Because it is mathematically incapable of reaching an intercept."
            ],
            "answer": "Because small changes in training data can lead to a completely different set of chosen features."
        },
        {
            "question": "Regarding computation, why did the Ames selection take 90 minutes with OLS but seconds with Regularization?",
            "options": [
                "OLS selection requires training and comparing thousands of separate models sequentially.",
                "Regularization uses a simpler version of the dataset with fewer rows.",
                "OLS requires the computer to calculate the log of every feature.",
                "Regularization skips the cross-validation step entirely."
            ],
            "answer": "OLS selection requires training and comparing thousands of separate models sequentially."
        },
        {
            "question": "What happens in Ridge Regression (L2) when Alpha is set to exactly 0?",
            "options": [
                "All coefficients are forced to zero.",
                "The model becomes a LASSO regression.",
                "The model acts exactly like a standard OLS linear regression.",
                "The penalty becomes infinitely large."
            ],
            "answer": "The model acts exactly like a standard OLS linear regression."
        },
        {
            "question": "Which statement best describes the coefficient shrinkage behavior of Ridge Regression?",
            "options": [
                "It forces unimportant features to be exactly 0.0.",
                "It shrinks coefficients asymptotically toward zero but never forces them to exactly 0.0.",
                "It only shrinks the intercept while leaving slopes untouched.",
                "It increases coefficients for features that are highly correlated."
            ],
            "answer": "It shrinks coefficients asymptotically toward zero but never forces them to exactly 0.0."
        },
        {
            "question": "Why is Ridge Regression (L2) particularly effective at handling multicollinearity?",
            "options": [
                "It picks the feature with the highest p-value and drops the rest.",
                "It combines correlated features into a single new feature.",
                "It distributes 'credit' among correlated variables, shrinking their weights together.",
                "It transforms all correlated variables into categorical integers."
            ],
            "answer": "It distributes 'credit' among correlated variables, shrinking their weights together."
        },
        {
            "question": "The 'L2' in L2 regularization refers to which mathematical concept?",
            "options": [
                "Taking the square root of the number of features.",
                "The L2 norm, which calculates the penalty based on the squared value of coefficients.",
                "Having exactly two target variables in the model.",
                "The second derivative of the loss function."
            ],
            "answer": "The L2 norm, which calculates the penalty based on the squared value of coefficients."
        },
        {
            "question": "If you have a domain where every variable is conceptually important and you don't want to lose any, which method is best?",
            "options": [
                "LASSO (L1)",
                "Backward Selection",
                "Ridge (L2)",
                "Forward Selection"
            ],
            "answer": "Ridge (L2)"
        },
        {
            "question": "What is a major limitation of Ridge Regression in terms of model 'decluttering'?",
            "options": [
                "It cannot perform automatic feature selection/removal.",
                "It requires the target variable to be normally distributed.",
                "It is only compatible with categorical data.",
                "It significantly increases the variance of the model."
            ],
            "answer": "It cannot perform automatic feature selection/removal."
        },
        {
            "question": "Why is 'Feature Scaling' (like StandardScaler) mandatory for both Ridge and Lasso?",
            "options": [
                "Because the penalty is applied to the coefficient size; unscaled data would unfairly penalize features with large units.",
                "Because the algorithm cannot handle negative numbers in the raw data.",
                "Because it ensures that the R-squared value remains between 0 and 1.",
                "Because regularization only works on data that is normally distributed."
            ],
            "answer": "Because the penalty is applied to the coefficient size; unscaled data would unfairly penalize features with large units."
        },
        {
            "question": "How does regularization affect the 'Interpretability' of the resulting slopes (coefficients)?",
            "options": [
                "It makes them more accurate representations of real-world physical changes.",
                "It introduces mathematical bias, meaning slopes can no longer be interpreted as literal unit increases.",
                "It makes the slopes exactly equal to the correlation coefficient.",
                "It has no effect on the interpretation of the slopes."
            ],
            "answer": "It introduces mathematical bias, meaning slopes can no longer be interpreted as literal unit increases."
        },
        {
            "question": "What does the acronym LASSO stand for?",
            "options": [
                "Linear Absolute Selection and Standard Operator",
                "Least Absolute Shrinkage and Selection Operator",
                "Logarithmic Aligned Squared Standard Optimization",
                "Least Adaptive Selection and Shrinkage Optimizer"
            ],
            "answer": "Least Absolute Shrinkage and Selection Operator"
        },
        {
            "question": "What is the key mathematical difference between Lasso and Ridge penalties?",
            "options": [
                "Lasso uses squared values (L2); Ridge uses absolute values (L1).",
                "Lasso uses absolute values (L1); Ridge uses squared values (L2).",
                "Lasso penalizes the intercept; Ridge does not.",
                "Lasso uses a log scale; Ridge uses a linear scale."
            ],
            "answer": "Lasso uses absolute values (L1); Ridge uses squared values (L2)."
        },
        {
            "question": "Lasso is often preferred over Ridge when the goal is:",
            "options": [
                "To ensure no variable ever reaches 0.0.",
                "To keep the most complex model possible.",
                "To achieve automatic feature selection and create a simpler model.",
                "To eliminate the need for cross-validation."
            ],
            "answer": "To achieve automatic feature selection and create a simpler model."
        },
        {
            "question": "If Lasso encounters two highly correlated features, how does it typically behave?",
            "options": [
                "It keeps both and gives them equal weights.",
                "It tends to arbitrarily pick one to keep and drops the other to zero.",
                "It averages the two features together.",
                "It drops both features to avoid multicollinearity."
            ],
            "answer": "It tends to arbitrarily pick one to keep and drops the other to zero."
        },
        {
            "question": "In the Bias-Variance tradeoff, regularization intentionally:",
            "options": [
                "Increases variance to reduce bias.",
                "Trades a bit of training accuracy (bias) for better generalization (reduced variance).",
                "Eliminates both bias and variance simultaneously.",
                "Decreases bias to ensure the training error is 0."
            ],
            "answer": "Trades a bit of training accuracy (bias) for better generalization (reduced variance)."
        },
        {
            "question": "What is ElasticNet Regression?",
            "options": [
                "A method that uses only categorical data.",
                "A regression that stretches the data points to fit a curve.",
                "A mathematical combination of both Lasso (L1) and Ridge (L2) penalties.",
                "A type of regression that does not require feature scaling."
            ],
            "answer": "A mathematical combination of both Lasso (L1) and Ridge (L2) penalties."
        },
        {
            "question": "In ElasticNet, what does an 'l1_ratio' of 1.0 signify?",
            "options": [
                "The model is 100% Ridge.",
                "The model is a 50/50 blend of Lasso and Ridge.",
                "The model is 100% Lasso.",
                "The model has no regularization penalty."
            ],
            "answer": "The model is 100% Lasso."
        },
        {
            "question": "Why is ElasticNet useful in fields like genomics where there are 'more columns than rows' (p > n)?",
            "options": [
                "It can handle datasets where standard OLS and Lasso mathematically break down.",
                "It automatically creates new rows of data using synthetic sampling.",
                "It requires no hyperparameter tuning.",
                "It is the only model that allows for negative R-squared values."
            ],
            "answer": "It can handle datasets where standard OLS and Lasso mathematically break down."
        },
        {
            "question": "What is the primary computational 'cost' of using ElasticNet over Ridge or Lasso?",
            "options": [
                "It requires double the memory on the hard drive.",
                "It requires tuning two parameters (Alpha and L1_ratio) simultaneously, often needing a 2D Grid Search.",
                "It can only be run on small datasets with under 100 rows.",
                "It makes the residual plots impossible to read."
            ],
            "answer": "It requires tuning two parameters (Alpha and L1_ratio) simultaneously, often needing a 2D Grid Search."
        },
        {
            "question": "According to the text, if a researcher only cares about 'predictive accuracy' and not interpretability, they usually move on to:",
            "options": [
                "Simple Linear Regression",
                "Tree-based models (like Random Forests or XGBoost)",
                "Manual Feature Selection",
                "Ridge Regression with Alpha=0"
            ],
            "answer": "Tree-based models (like Random Forests or XGBoost)"
        },
        {
            "question": "Regularization helps prevent a model from 'memorizing' random noise. This 'memorization' is also known as:",
            "options": [
                "Underfitting",
                "Overfitting",
                "Global Optimization",
                "Standardization"
            ],
            "answer": "Overfitting"
        },
        {
            "question": "In the Ames modeling workflow, why was 'SalePrice' log-transformed (y = np.log)?",
            "options": [
                "To make the calculation of Alpha easier.",
                "To handle right-skewness in the target variable.",
                "Because Lasso only works on log-scaled data.",
                "To increase the number of features in the model."
            ],
            "answer": "To handle right-skewness in the target variable."
        },
        {
            "question": "Lasso Regression uses an L1 penalty. The '1' in L1 refers to:",
            "options": [
                "The fact that it only uses one feature.",
                "Taking the absolute value to the power of 1.",
                "Having only one alpha value.",
                "Being the #1 preferred method for Ridge enthusiasts."
            ],
            "answer": "Taking the absolute value to the power of 1."
        },
        {
            "question": "If Alpha is set to an 'overwhelmingly high' value in Lasso, what is the likely result?",
            "options": [
                "Every feature will have a massive coefficient.",
                "The model will be identical to an OLS model.",
                "The model will drop all predictors, leaving only the intercept.",
                "The R-squared value will increase significantly."
            ],
            "answer": "The model will drop all predictors, leaving only the intercept."
        },
        {
            "question": "When comparing Ridge and ElasticNet behavior on highly correlated groups of variables:",
            "options": [
                "Ridge kills the whole group; ElasticNet keeps one.",
                "Ridge keeps the whole group; ElasticNet can keep the group but also zero out truly useless variables.",
                "Both models behave exactly like Forward Selection.",
                "Neither model can handle correlated variables."
            ],
            "answer": "Ridge keeps the whole group; ElasticNet can keep the group but also zero out truly useless variables."
        },
        {
            "question": "Which step is 'CRITICAL' for linear models when performing One-Hot Encoding to prevent multicollinearity?",
            "options": [
                "Scaling the target variable.",
                "Setting 'drop_first=True'.",
                "Removing all numeric columns.",
                "Using cross-validation with 100 folds."
            ],
            "answer": "Setting 'drop_first=True'."
        },
    ],
    "17_2_4_4: Hyperparameter Tuning": [
        {
            "question": "In the manual alpha tuning loop, you plot both training R\u00b2 and cross-validation R\u00b2 against alpha. As alpha increases from 0.01 to 1000, what pattern do you expect to see in the two curves?",
            "options": [
                "Training R\u00b2 rises steadily while CV R\u00b2 falls steadily, reflecting increasing model complexity.",
                "Training R\u00b2 falls steadily while CV R\u00b2 rises then falls, peaking at the optimal alpha.",
                "Both curves rise in parallel, since regularization always improves model fit.",
                "Training R\u00b2 stays flat while CV R\u00b2 oscillates randomly around a constant value."
            ],
            "answer": "Training R\u00b2 falls steadily while CV R\u00b2 rises then falls, peaking at the optimal alpha."
        },
        {
            "question": "A colleague runs GridSearchCV with 5-fold CV on a training set of 2,000 samples, testing 10 alpha values. How many total model fits does GridSearchCV perform?",
            "options": [
                "10, because it only fits one model per alpha value.",
                "50, because it fits 10 alphas across 5 folds.",
                "2,000, because it fits one model per training sample.",
                "15, because it adds the number of alphas and folds together."
            ],
            "answer": "50, because it fits 10 alphas across 5 folds."
        },
        {
            "question": "You use GridSearchCV with refit=True (the default). After the search completes, what model does grid_search.best_estimator_ return?",
            "options": [
                "The model trained on the single best-performing fold during the search.",
                "The model retrained on the entire training set using the best hyperparameters found.",
                "An ensemble of all models tested during the search, weighted by their CV scores.",
                "The model from the fold with the lowest validation error, without retraining."
            ],
            "answer": "The model retrained on the entire training set using the best hyperparameters found."
        },
        {
            "question": "Why does the Pipeline prefix hyperparameter keys with 'model__' (e.g., 'model__alpha') when passing them to GridSearchCV?",
            "options": [
                "It is a naming convention that GridSearchCV ignores; any prefix would work.",
                "It tells GridSearchCV which step in the pipeline the parameter belongs to.",
                "It prevents GridSearchCV from trying to tune the scaler's parameters.",
                "It signals to scikit-learn that the parameter should be tuned on a logarithmic scale."
            ],
            "answer": "It tells GridSearchCV which step in the pipeline the parameter belongs to."
        },
        {
            "question": "You fit a StandardScaler on the full dataset, then pass the scaled data to cross_val_score with cv=5. What is the consequence?",
            "options": [
                "Nothing changes; scaling before CV is mathematically equivalent to scaling within each fold.",
                "The scaler's mean and variance incorporate information from every fold, leaking into each training split.",
                "Cross-validation automatically detects the pre-scaled data and adjusts its fold boundaries.",
                "The model trains faster because the scaling computation is performed only once instead of five times."
            ],
            "answer": "The scaler's mean and variance incorporate information from every fold, leaking into each training split."
        },
        {
            "question": "In ElasticNet, the l1_ratio parameter controls the mix between L1 and L2 penalties. What does l1_ratio=0.7 mean?",
            "options": [
                "70% of the penalty comes from the L1 (Lasso) term and 30% from the L2 (Ridge) term.",
                "70% of the features will be dropped to exactly zero by the model.",
                "The model will behave 70% like Ridge and 30% like Lasso.",
                "The overall penalty strength is reduced to 70% of the specified alpha value."
            ],
            "answer": "70% of the penalty comes from the L1 (Lasso) term and 30% from the L2 (Ridge) term."
        },
        {
            "question": "After tuning three regularized models (Ridge, Lasso, ElasticNet) with GridSearchCV, all three achieve nearly identical test R\u00b2 scores. What is the most likely explanation?",
            "options": [
                "The hyperparameter grids were too narrow to distinguish between the models.",
                "The remaining features are weakly informative, so all regularization approaches converge to similar solutions.",
                "GridSearchCV always returns the same model regardless of the algorithm specified.",
                "The test set is too small to distinguish between different modeling approaches."
            ],
            "answer": "The remaining features are weakly informative, so all regularization approaches converge to similar solutions."
        },
        {
            "question": "You run GridSearchCV on a training set and find the best alpha. You then evaluate that model on a held-out test set and get R\u00b2 = 0.83. Why might this score be slightly inflated?",
            "options": [
                "The test set is systematically easier to predict than the training set due to random sampling.",
                "The best alpha was selected because it happened to perform well on the specific CV folds, partially capturing noise.",
                "GridSearchCV uses a different scoring metric than the one used for the final test evaluation.",
                "The model was retrained on the full training set, which artificially boosts test performance."
            ],
            "answer": "The best alpha was selected because it happened to perform well on the specific CV folds, partially capturing noise."
        },
        {
            "question": "In nested cross-validation with 5 outer folds and 5 inner folds, how many total model fits are performed when searching over 8 alpha values?",
            "options": [
                "40, because there are 5 outer folds times 8 alpha values, giving 40 total fits.",
                "200, because each outer fold runs the full inner GridSearch.",
                "25, because there are 5 outer folds times 5 inner folds, giving 25 total fits.",
                "13, because you add the 5 outer folds, 5 inner folds, and 3 unique alpha values together."
            ],
            "answer": "200, because each outer fold runs the full inner GridSearch."
        },
        {
            "question": "In the nested CV code, cross_val_score is called with a GridSearchCV object as its estimator. What happens during each outer fold?",
            "options": [
                "cross_val_score extracts the best hyperparameters from the GridSearchCV and fits a simple model with them.",
                "cross_val_score calls .fit() on the GridSearchCV, which runs the full inner search, then uses the refitted best model to predict on the outer test fold.",
                "cross_val_score skips the inner search and uses the default hyperparameters of the base estimator.",
                "cross_val_score averages the predictions of all hyperparameter combinations tested by the inner search."
            ],
            "answer": "cross_val_score calls .fit() on the GridSearchCV, which runs the full inner search, then uses the refitted best model to predict on the outer test fold."
        },
        {
            "question": "Your nested CV results show Ridge R\u00b2 = 0.7910 \u00b1 0.0266, while the single train/test split gave R\u00b2 = 0.8300. What does the \u00b1 0.0266 tell you?",
            "options": [
                "The model's performance varies by about 2.7 percentage points depending on which data ends up in each fold.",
                "The nested CV code has a bug that introduces random noise into the scores.",
                "The inner loop's hyperparameter search is unstable and needs more alpha values.",
                "The outer test folds contain outlier houses that distort the average score."
            ],
            "answer": "The model's performance varies by about 2.7 percentage points depending on which data ends up in each fold."
        },
        {
            "question": "Why does nested CV use the full dataset (X, y) rather than just X_train, y_train?",
            "options": [
                "Because nested CV replaces the train/test split; every data point serves as both training and test data across different folds.",
                "Because the outer loop needs more data than the inner loop to produce stable hyperparameter estimates.",
                "Because GridSearchCV requires the full dataset to properly scale features before splitting.",
                "Because using only the training set would cause the inner loop to overfit more severely."
            ],
            "answer": "Because nested CV replaces the train/test split; every data point serves as both training and test data across different folds."
        },
        {
            "question": "In nested CV, Outer Fold 1 selects alpha=1 as optimal, while Outer Fold 3 selects alpha=10. What does this tell you?",
            "options": [
                "The inner loop's random state was not fixed, causing non-reproducible results.",
                "The optimal regularization strength depends on which specific samples are in the training subset.",
                "The model is fundamentally unstable and should not be trusted for any predictions.",
                "One of the outer folds contains corrupted data that skews the hyperparameter search."
            ],
            "answer": "The optimal regularization strength depends on which specific samples are in the training subset."
        },
        {
            "question": "A learning curve shows training R\u00b2 = 0.95 and validation R\u00b2 = 0.70 at the largest training size, with a persistent gap between the two curves. What should you do?",
            "options": [
                "Add more training data, since the validation curve is still improving.",
                "Increase model complexity, since both scores are too low.",
                "Increase regularization or reduce model complexity, since the model is overfitting.",
                "Switch to a different scoring metric, since R\u00b2 is not appropriate for this dataset."
            ],
            "answer": "Increase regularization or reduce model complexity, since the model is overfitting."
        },
        {
            "question": "A learning curve shows both training and validation R\u00b2 converging at approximately 0.55, with the validation curve flatlining as training size increases. What does this indicate?",
            "options": [
                "The model is overfitting and would benefit from more training data.",
                "The model is underfitting; adding more data will not help, and a more complex model is needed.",
                "The model is well-calibrated and performing at its theoretical maximum.",
                "The cross-validation folds are too small to produce reliable validation scores."
            ],
            "answer": "The model is underfitting; adding more data will not help, and a more complex model is needed."
        },
        {
            "question": "In the learning curve code, train_sizes=np.linspace(0.1, 1.0, 5) is used. How many different training set sizes are evaluated?",
            "options": [
                "5, representing 10%, 32.5%, 55%, 77.5%, and 100% of the training data.",
                "10, because linspace generates values from 0.1 to 1.0 in steps of 0.1.",
                "1, because the model is only trained on the full dataset at the end.",
                "3, because the first and last values are excluded from the evaluation."
            ],
            "answer": "5, representing 10%, 32.5%, 55%, 77.5%, and 100% of the training data."
        },
        {
            "question": "You compare nested CV R\u00b2 scores for Ridge (0.7910), Lasso (0.7910), and ElasticNet (0.7909). All three are nearly identical. What does this suggest about the underlying data?",
            "options": [
                "The feature set is small enough that L1 and L2 penalties produce nearly equivalent coefficient shrinkage.",
                "The GridSearchCV grids for all three models were identical, forcing them to the same solution.",
                "The nested CV implementation has a bug that returns the same score for all models.",
                "The data contains no predictive signal, so all models converge to the same baseline."
            ],
            "answer": "The feature set is small enough that L1 and L2 penalties produce nearly equivalent coefficient shrinkage."
        },
        {
            "question": "Which statement best describes the relationship between the inner loop's best CV score and the outer loop's average score in nested CV?",
            "options": [
                "The inner loop's score is always higher because it averages over smaller training subsets.",
                "The outer loop's score is always higher because it evaluates on larger test folds.",
                "The two scores are always identical because they use the same scoring metric.",
                "The inner loop's score is always lower because it uses more folds than the outer loop."
            ],
            "answer": "The inner loop's score is always higher because it averages over smaller training subsets."
        },
        {
            "question": "You are building a production model for a client who needs a trustworthy performance estimate. Your dataset has 1,500 samples. Which approach should you use?",
            "options": [
                "A single 80/20 train/test split with GridSearchCV, because it is fast and simple.",
                "Nested cross-validation, because it provides an unbiased estimate with quantifiable variance.",
                "Leave-one-out cross-validation without hyperparameter tuning, because it uses all data for training.",
                "A single model fit on all 1,500 samples with no validation, because the dataset is too small to split."
            ],
            "answer": "Nested cross-validation, because it provides an unbiased estimate with quantifiable variance."
        },
        {
            "question": "In the bias-variance tradeoff plot for Ridge regression, the CV R\u00b2 peaks at alpha=1. If you were to add alpha=0.5 and alpha=2 to your grid, what would you expect?",
            "options": [
                "Alpha=0.5 would have a higher CV R\u00b2 because lower alpha always means better generalization.",
                "Alpha=2 would have a higher CV R\u00b2 because higher alpha always means better generalization.",
                "Both would likely have lower CV R\u00b2 than alpha=1, since alpha=1 is already at the peak.",
                "One would be higher and one lower, but you cannot predict which without running the experiment."
            ],
            "answer": "One would be higher and one lower, but you cannot predict which without running the experiment."
        },
        {
            "question": "A student claims: 'Nested CV is unnecessary if I have a separate test set that I never touch during tuning.' What is the strongest counterargument?",
            "options": [
                "The student is correct; a separate test set completely eliminates the need for nested CV.",
                "A single test set gives only one performance estimate with no measure of its variability across different data splits.",
                "Nested CV is always faster than a train/test split, so there is no reason not to use it.",
                "The test set score is always lower than the nested CV score, making nested CV the more conservative choice."
            ],
            "answer": "A single test set gives only one performance estimate with no measure of its variability across different data splits."
        },
        {
            "question": "When using learning_curve() with cv=5, how is the validation score computed at each training size?",
            "options": [
                "The model is trained once on the specified fraction of data and scored on the remaining data.",
                "For each training size, 5 models are trained on different folds and the validation scores are averaged.",
                "The model is trained on the full dataset and scored on 5 different random subsets of the test set.",
                "The training and validation scores are computed on the same data to measure overfitting."
            ],
            "answer": "For each training size, 5 models are trained on different folds and the validation scores are averaged."
        },
        {
            "question": "You notice that the learning curve's validation score is still climbing at the largest training size (100% of training data). What is the most actionable next step?",
            "options": [
                "Reduce model complexity, since the model is clearly overfitting at all training sizes.",
                "Collect more training data, since the model has not yet plateaued and could benefit from additional examples.",
                "Switch to a simpler model, since the current model is too complex for the available data.",
                "Increase the number of cross-validation folds to get a more accurate validation score."
            ],
            "answer": "Collect more training data, since the model has not yet plateaued and could benefit from additional examples."
        },
        {
            "question": "In the notebook, the 'Features Kept' column counts coefficients with absolute value > 1e-5. Why is this threshold used instead of checking for exact zeros?",
            "options": [
                "Because Ridge regression never produces exact zeros; coefficients are shrunk toward zero but rarely reach it.",
                "Because floating-point arithmetic makes exact zero comparisons unreliable for all models.",
                "Because Lasso always produces coefficients slightly above zero due to numerical imprecision.",
                "Because the 1e-5 threshold is the default regularization strength in scikit-learn."
            ],
            "answer": "Because Ridge regression never produces exact zeros; coefficients are shrunk toward zero but rarely reach it."
        },
        {
            "question": "You run nested CV with 5 outer folds and 3 inner folds, searching 12 hyperparameter combinations. Your manager asks how long it will take compared to a single GridSearchCV run (5 folds, 12 combinations). What do you tell them?",
            "options": [
                "It will take about the same time, since the inner loop reuses computations from the outer loop.",
                "It will take roughly 5 times longer, because the outer loop runs the full inner GridSearch once per outer fold.",
                "It will take 15 times longer, because you multiply the outer folds, inner folds, and combinations.",
                "It will take less time, because nested CV uses smaller training sets in each fold."
            ],
            "answer": "It will take roughly 5 times longer, because the outer loop runs the full inner GridSearch once per outer fold."
        },
        {
            "question": "The notebook states that nested CV scores are 'typically lower' than train/test split scores. Under what circumstance might nested CV produce a higher score?",
            "options": [
                "When the single test split happened to contain unusually difficult-to-predict samples by chance.",
                "When the inner loop uses more folds than the outer loop, inflating the score.",
                "When the model is severely underfitting, making all evaluation methods produce identical scores.",
                "This can never happen; nested CV always produces a lower score by definition."
            ],
            "answer": "When the single test split happened to contain unusually difficult-to-predict samples by chance."
        },
        {
            "question": "In the GridSearchCV output, 'Fitting 5 folds for each of 20 candidates, totalling 100 fits' appears for ElasticNet. What are the 20 candidates?",
            "options": [
                "20 different random splits of the training data into train and validation sets.",
                "The Cartesian product of 4 alpha values and 5 l1_ratio values, giving 20 unique hyperparameter combinations.",
                "20 different model architectures, each with a different number of features.",
                "20 iterations of the ElasticNet optimization algorithm before convergence."
            ],
            "answer": "The Cartesian product of 4 alpha values and 5 l1_ratio values, giving 20 unique hyperparameter combinations."
        },
        {
            "question": "You fit a learning curve and observe that the training score starts at 0.99 with 10% of the data and drops to 0.79 with 100% of the data, while the validation score rises from 0.50 to 0.78. What is happening?",
            "options": [
                "The model is overfitting on small training sets and gradually improving its generalization as more data is added.",
                "The model is underfitting at all training sizes and would benefit from a more complex architecture.",
                "The cross-validation folds are too small at low training sizes, producing unreliable scores.",
                "The learning rate is too high, causing the model to diverge as more data is introduced."
            ],
            "answer": "The model is overfitting on small training sets and gradually improving its generalization as more data is added."
        },
        {
            "question": "A colleague argues that nested CV is 'wasteful' because it trains hundreds of models but only reports a single average score. How would you respond?",
            "options": [
                "They are correct; a single train/test split with GridSearchCV is always sufficient for reporting results.",
                "The average score is the deliverable, but the standard deviation across outer folds is equally valuable information about model stability.",
                "Nested CV produces a separate final model for each outer fold, giving you five models to deploy in production.",
                "The computational cost of nested CV is negligible on modern hardware, so the objection is moot."
            ],
            "answer": "The average score is the deliverable, but the standard deviation across outer folds is equally valuable information about model stability."
        },
        {
            "question": "After running nested CV, you want to deploy a final model for production predictions. Which model should you deploy?",
            "options": [
                "The average of all models produced across the outer folds, since averaging reduces variance.",
                "The model from the outer fold with the highest test score, since it performed best.",
                "A fresh model trained on all data using the most common hyperparameters from the inner loops.",
                "The model from the inner loop with the lowest validation error, since it was most accurate."
            ],
            "answer": "A fresh model trained on all data using the most common hyperparameters from the inner loops."
        },
    ],
    "17_2_4_5: Tree-Based Methods": [
        {
            "question": "Linear models required manual polynomial features to capture non-linear patterns. How do tree-based models handle non-linearity?",
            "options": [
                "They apply logarithmic transformations to all features automatically.",
                "They partition the data into regions and predict the average within each region.",
                "They add squared terms to the regression equation during training.",
                "They use gradient descent to find the optimal non-linear curve."
            ],
            "answer": "They partition the data into regions and predict the average within each region."
        },
        {
            "question": "In a regression tree, what value does each leaf node predict?",
            "options": [
                "The median of the target values of training samples in that leaf.",
                "The average of the target values of training samples in that leaf.",
                "The most frequent target value among training samples in that leaf.",
                "A weighted sum of coefficients applied to the features in that leaf."
            ],
            "answer": "The average of the target values of training samples in that leaf."
        },
        {
            "question": "How does a regression tree decide which feature and split point to use at each node?",
            "options": [
                "It randomly selects a feature and tries all possible split points on that feature.",
                "It selects the feature with the highest correlation to the target variable.",
                "It tries every possible split on every feature and picks the one that reduces error the most.",
                "It uses the feature that was most important in the parent node's split."
            ],
            "answer": "It tries every possible split on every feature and picks the one that reduces error the most."
        },
        {
            "question": "An unconstrained decision tree achieves R\u00b2 = 1.0 on training data but R\u00b2 = 0.80 on test data. What is happening?",
            "options": [
                "The model is underfitting because it has too few splits to capture the patterns.",
                "The model is overfitting by creating a separate leaf for nearly every training sample.",
                "The training data contains outliers that the model is correctly ignoring.",
                "The test set is fundamentally different from the training set due to data leakage."
            ],
            "answer": "The model is overfitting by creating a separate leaf for nearly every training sample."
        },
        {
            "question": "In the depth experiment, the test R\u00b2 peaked at depth 7 and then declined as depth increased. What does this pattern represent?",
            "options": [
                "The bias-variance tradeoff, where deeper trees reduce bias but increase variance.",
                "A computational limitation where deeper trees run out of memory.",
                "The learning rate decay, where deeper trees contribute less to the prediction.",
                "A data quality issue where deeper splits encounter noisier samples."
            ],
            "answer": "The bias-variance tradeoff, where deeper trees reduce bias but increase variance."
        },
        {
            "question": "Why does a Random Forest use bootstrap sampling (drawing with replacement) for each tree?",
            "options": [
                "To ensure every tree sees the exact same data but in a different order.",
                "To reduce the computational cost of training each individual tree.",
                "To give each tree a different perspective on the data so they make different errors.",
                "To eliminate the need for a separate test set during model evaluation."
            ],
            "answer": "To give each tree a different perspective on the data so they make different errors."
        },
        {
            "question": "In a Random Forest, what does max_features='sqrt' mean at each split?",
            "options": [
                "Each tree considers only the square root of the total features across the entire tree.",
                "Each tree considers a random subset of about \u221an features at each individual node.",
                "The forest builds exactly \u221an trees, where n is the number of features.",
                "Each tree uses the top \u221an most important features and ignores the rest."
            ],
            "answer": "Each tree considers a random subset of about \u221an features at each individual node."
        },
        {
            "question": "Why does feature randomness in Random Forests prevent every tree from splitting on the same dominant feature first?",
            "options": [
                "Because the dominant feature is removed from the dataset before training begins.",
                "Because each tree is forced to consider only a random subset of features at each node.",
                "Because the dominant feature's values are randomly shuffled before each split.",
                "Because the algorithm penalizes trees that use the same feature twice."
            ],
            "answer": "Because each tree is forced to consider only a random subset of features at each node."
        },
        {
            "question": "The Random Forest selected max_depth=None as optimal. Why is unlimited depth acceptable in a forest but catastrophic for a single tree?",
            "options": [
                "Because the forest uses a different splitting criterion that prevents overfitting.",
                "Because the averaging process across diverse trees cancels out individual overfitting.",
                "Because each tree in the forest sees fewer features, making deep trees simpler.",
                "Because the forest automatically prunes trees that show signs of overfitting."
            ],
            "answer": "Because the averaging process across diverse trees cancels out individual overfitting."
        },
        {
            "question": "What is the primary objective of a Random Forest from a bias-variance perspective?",
            "options": [
                "Reducing bias by building deeper and deeper trees sequentially.",
                "Reducing variance by averaging many diverse, uncorrelated predictions.",
                "Reducing both bias and variance equally through random feature selection.",
                "Eliminating bias entirely by using an infinite number of trees."
            ],
            "answer": "Reducing variance by averaging many diverse, uncorrelated predictions."
        },
        {
            "question": "In gradient boosting, what does each new tree predict?",
            "options": [
                "The original target variable (e.g., house price) directly.",
                "The residuals (errors) left by the ensemble of all previous trees.",
                "The average of all previous trees' predictions.",
                "A weighted combination of the target variable and the previous residuals."
            ],
            "answer": "The residuals (errors) left by the ensemble of all previous trees."
        },
        {
            "question": "Gradient boosting starts with an initial prediction equal to the average target value. What is the role of the first tree?",
            "options": [
                "It predicts the target variable directly, ignoring the initial prediction.",
                "It predicts the residuals from the initial prediction, correcting the average.",
                "It replaces the initial prediction entirely with its own predictions.",
                "It determines which features are most important for all subsequent trees."
            ],
            "answer": "It predicts the residuals from the initial prediction, correcting the average."
        },
        {
            "question": "What is the key difference between Random Forests and Gradient Boosting in how they build their ensembles?",
            "options": [
                "Random Forests build trees sequentially; Gradient Boosting builds them in parallel.",
                "Random Forests average independent trees; Gradient Boosting corrects errors sequentially.",
                "Random Forests use deep trees; Gradient Boosting uses only single-split trees.",
                "Random Forests reduce bias; Gradient Boosting reduces variance."
            ],
            "answer": "Random Forests average independent trees; Gradient Boosting corrects errors sequentially."
        },
        {
            "question": "Why does HistGradientBoostingRegressor use histogram binning?",
            "options": [
                "To convert categorical features into numeric codes automatically.",
                "To group continuous data into discrete bins, speeding up the split-finding process.",
                "To reduce the number of features by combining correlated ones into bins.",
                "To handle missing values by placing them into a separate histogram bin."
            ],
            "answer": "To group continuous data into discrete bins, speeding up the split-finding process."
        },
        {
            "question": "The raw-data HGB model achieved R\u00b2 = 0.9449, nearly identical to the cleaned-data version. What does this demonstrate?",
            "options": [
                "That data cleaning is never necessary when using tree-based models.",
                "That tree-based models are robust to messy data and don't require extensive preprocessing.",
                "That the Ames dataset was already clean enough that no cleaning was needed.",
                "That HGB automatically performs the same cleaning steps as the notebook's cleaning function."
            ],
            "answer": "That tree-based models are robust to messy data and don't require extensive preprocessing."
        },
        {
            "question": "How does XGBoost's built-in regularization differ from Ridge and Lasso regularization?",
            "options": [
                "XGBoost applies penalties to leaf weights and tree complexity, not to linear coefficients.",
                "XGBoost only uses L1 regularization, while Ridge and Lasso use both L1 and L2.",
                "XGBoost applies regularization after training, while Ridge and Lasso apply it during training.",
                "XGBoost does not use regularization; it relies solely on tree depth limits."
            ],
            "answer": "XGBoost applies penalties to leaf weights and tree complexity, not to linear coefficients."
        },
        {
            "question": "Boosting is inherently sequential, making it hard to parallelize. How does XGBoost achieve speed improvements?",
            "options": [
                "It builds multiple trees in parallel and averages their predictions.",
                "It parallelizes the internal work of building each tree, such as feature sorting.",
                "It skips the residual calculation step for trees after the first 50.",
                "It uses a single CPU core but with a more efficient optimization algorithm."
            ],
            "answer": "It parallelizes the internal work of building each tree, such as feature sorting."
        },
        {
            "question": "How does XGBoost handle missing values at each split?",
            "options": [
                "It imputes missing values with the median before training begins.",
                "It removes any sample with a missing value from the training set.",
                "It tests sending missing values left or right and learns which direction reduces error.",
                "It assigns missing values to a separate leaf node that predicts the global average."
            ],
            "answer": "It tests sending missing values left or right and learns which direction reduces error."
        },
        {
            "question": "Feature importance in tree models measures how much each split reduced prediction error. What is a key limitation of this metric?",
            "options": [
                "It cannot handle categorical features, so their importance is always zero.",
                "It tells you what the model prioritizes but not how the feature affects the target.",
                "It only works for the top 5 features and ignores all remaining features.",
                "It requires the model to be retrained after importance scores are calculated."
            ],
            "answer": "It tells you what the model prioritizes but not how the feature affects the target."
        },
        {
            "question": "In the feature importance plot, Overall Qual accounts for 45.76% of the model's learning. What does this percentage mean?",
            "options": [
                "Overall Qual alone can predict 45.76% of house prices without any other features.",
                "Splits on Overall Qual were responsible for 45.76% of the total error reduction across all trees.",
                "45.76% of the trees in the ensemble used Overall Qual as their first split.",
                "Overall Qual has a coefficient of 0.4576 in the underlying linear approximation."
            ],
            "answer": "Splits on Overall Qual were responsible for 45.76% of the total error reduction across all trees."
        },
        {
            "question": "Gr Liv Area has a VIF of 118 (extreme multicollinearity) but only 2.98% feature importance. Why?",
            "options": [
                "Living area is genuinely unimportant for predicting house prices in Ames.",
                "XGBoost chose correlated features like Overall Qual for splits, leaving Gr Liv Area with little credit.",
                "The VIF calculation is only relevant for linear models and has no bearing on tree models.",
                "The feature importance calculation is biased against continuous features."
            ],
            "answer": "XGBoost chose correlated features like Overall Qual for splits, leaving Gr Liv Area with little credit."
        },
        {
            "question": "In the model comparison, XGBoost's single-split score was 0.9418 but its nested CV estimate was 0.9177. What explains this gap?",
            "options": [
                "The nested CV used a different scoring metric that is inherently more conservative.",
                "The single-split score was inflated because the same data influenced both tuning and evaluation.",
                "The nested CV used fewer trees, resulting in a less powerful model.",
                "The test set in the single-split approach contained easier-to-predict houses by chance."
            ],
            "answer": "The single-split score was inflated because the same data influenced both tuning and evaluation."
        },
        {
            "question": "The nested CV standard deviation for XGBoost was \u00b10.0195. What does this tell you?",
            "options": [
                "The model's performance varies by about 2 percentage points across different data splits.",
                "The hyperparameter search found 19.5% of the possible combinations to be optimal.",
                "The model is overfitting by approximately 1.95 percentage points on each fold.",
                "The inner loop's best score was 1.95 percentage points higher than the outer loop's score."
            ],
            "answer": "The model's performance varies by about 2 percentage points across different data splits."
        },
        {
            "question": "The notebook recommends starting with a Random Forest for most real-world problems. Why?",
            "options": [
                "Random Forests always produce the highest accuracy of any model.",
                "Random Forests are fast, robust, and hard to mess up compared to tuning boosting models.",
                "Random Forests are the only tree-based method that handles missing values natively.",
                "Random Forests provide the most interpretable feature importance scores."
            ],
            "answer": "Random Forests are fast, robust, and hard to mess up compared to tuning boosting models."
        },
        {
            "question": "What tradeoff do you accept when moving from linear regression to XGBoost?",
            "options": [
                "You gain interpretability but lose the ability to handle non-linear patterns.",
                "You gain predictive accuracy but lose the ability to explain individual coefficient effects.",
                "You gain speed but lose the ability to handle categorical features.",
                "You gain robustness to outliers but lose the ability to use cross-validation."
            ],
            "answer": "You gain predictive accuracy but lose the ability to explain individual coefficient effects."
        },
        {
            "question": "In the nested CV for XGBoost, the inner loop used 3 folds and the outer loop used 5 folds with 12 hyperparameter combinations. How many total model fits occurred?",
            "options": [
                "60, because 5 outer folds times 12 hyperparameter combinations equals 60.",
                "180, because 5 outer folds each run a full inner search of 12 combinations times 3 inner folds.",
                "15, because 5 outer folds times 3 inner folds equals 15 total evaluations.",
                "36, because 12 hyperparameter combinations times 3 inner folds equals 36."
            ],
            "answer": "180, because 5 outer folds each run a full inner search of 12 combinations times 3 inner folds."
        },
        {
            "question": "Central Air achieved 7.86% feature importance despite being a binary feature. Why might a binary feature score highly?",
            "options": [
                "Binary features are always more important than continuous features in tree models.",
                "The presence or absence of central air is a major price differentiator in the Ames market.",
                "Binary features are used more frequently as split points because they are easier to compute.",
                "The model automatically upweights binary features to compensate for their simplicity."
            ],
            "answer": "The presence or absence of central air is a major price differentiator in the Ames market."
        },
        {
            "question": "The notebook mentions that",
            "options": [
                "They identify which features should be removed from the model entirely.",
                "They reveal the direction and shape of a feature's effect on the prediction.",
                "They calculate the exact dollar value contribution of each feature.",
                "They determine the optimal hyperparameters for the model."
            ],
            "answer": "They reveal the direction and shape of a feature's effect on the prediction."
        },
        {
            "question": "Decision trees convert categorical features to numeric codes using .cat.codes. What risk does this introduce?",
            "options": [
                "The model treats categories as having an ordinal relationship that may not exist.",
                "The model loses all information about the categorical feature during conversion.",
                "The model becomes unable to handle new categories that appear in the test set.",
                "The model requires one-hot encoding to work correctly with numeric codes."
            ],
            "answer": "The model treats categories as having an ordinal relationship that may not exist."
        },
        {
            "question": "The notebook's final model comparison shows Decision Tree (0.83) < Random Forest (0.88) < HGB (0.94) < XGBoost (0.94). What does this progression illustrate?",
            "options": [
                "That more complex models always produce better results regardless of the dataset.",
                "That each ensemble method builds on the previous one's weaknesses to improve accuracy.",
                "That the differences between models are negligible and any tree model would suffice.",
                "That the hyperparameter grids were progressively larger, giving later models an unfair advantage."
            ],
            "answer": "That each ensemble method builds on the previous one's weaknesses to improve accuracy."
        },
    ],
}


# --- QUIZ APP ---
class QuizApp:
    def __init__(self, quiz_name, data):
        self.quiz_name = quiz_name
        self.data = data
        self.current_index = 0
        self.score = 0
        self.user_answers = []

        # UI Elements
        self.header = widgets.HTML(value="<h2>" + quiz_name + "</h2>")
        self.progress = widgets.HTML(value="")
        self.question_text = widgets.HTML(value="")
        self.options = widgets.RadioButtons(options=[], layout={"width": "max-content"})
        self.submit_btn = widgets.Button(description="Submit Answer", button_style="primary")
        self.next_btn = widgets.Button(description="Next Question", button_style="info", disabled=True)
        self.back_btn = widgets.Button(description="Back to Menu", button_style="warning")
        self.feedback = widgets.HTML(value="")

        self.submit_btn.on_click(self.handle_submit)
        self.next_btn.on_click(self.handle_next)
        self.back_btn.on_click(self.handle_back)

        self.render_question()

    def render_question(self):
        clear_output(wait=True)
        q = self.data[self.current_index]
        self.progress.value = "<b>Question " + str(self.current_index + 1) + " of " + str(len(self.data)) + "</b>"
        self.question_text.value = "<p style='font-size:14px;'>" + q["question"] + "</p>"
        self.options.options = q["options"]
        self.options.value = None
        self.submit_btn.disabled = False
        self.next_btn.disabled = True
        self.feedback.value = ""

        display(self.header, self.progress, self.question_text, self.options,
                widgets.HBox([self.submit_btn, self.next_btn]), self.feedback)

    def handle_submit(self, b):
        if self.options.value is None:
            self.feedback.value = "<span style='color:red;'>Please select an option!</span>"
            return

        correct_answer = self.data[self.current_index]["answer"]
        if self.options.value == correct_answer:
            self.score += 1
            self.feedback.value = "<span style='color:green; font-weight:bold;'>Correct!</span>"
        else:
            self.feedback.value = "<span style='color:red; font-weight:bold;'>Incorrect.</span><br>The correct answer was: <b>" + correct_answer + "</b>"

        self.submit_btn.disabled = True
        self.next_btn.disabled = False

    def handle_next(self, b):
        self.current_index += 1
        if self.current_index < len(self.data):
            self.render_question()
        else:
            self.show_results()

    def handle_back(self, b):
        show_menu()

    def show_results(self):
        clear_output()
        percent = (self.score / len(self.data)) * 100
        result_html = """
        <div style="border: 2px solid #0366d6; padding: 20px; border-radius: 10px;">
            <h2>Quiz Complete!</h2>
            <p style="font-size:16px;">Your Final Score: <b>" + str(self.score) + " / " + str(len(self.data)) + "</b></p>
            <p style="font-size:20px; color:#0366d6;"><b>Percentage: " + "{:.1f}".format(percent) + "%</b></p>
        </div>
        """
        display(widgets.HTML(result_html), self.back_btn)


# --- MENU ---
def show_menu():
    clear_output(wait=True)
    header = widgets.HTML(value="<h2>17_2 Regression Cross-Validation -- Practice Quizzes</h2>")
    instruction = widgets.HTML(value="<p>Select a notebook quiz below to begin.</p>")
    quiz_selector = widgets.Dropdown(
        options=list(all_quizzes.keys()),
        description="Select Quiz:",
        layout={"width": "50%"}
    )
    start_btn = widgets.Button(description="Start Quiz", button_style="success")

    def on_start(b):
        selected = quiz_selector.value
        QuizApp(selected, all_quizzes[selected])

    start_btn.on_click(on_start)
    display(header, instruction, quiz_selector, start_btn)


# Start the app
show_menu()
```
