# 17.1 Simple Linear Regression Practice Quiz
This notebook contains practice questions covering the concepts of Simple Linear Regression, Correlation, Residuals, and Model Assessment.


### Question
What is the formal equation for a simple linear regression model?

**Options:**
- y = b0 + b1x + e
- y = b0 + b1x^2 + e
- y = b0 * b1x + e
- y = b0 + b1/x + e

<details>
<summary>Click to see Answer</summary>

**Correct Answer: y = b0 + b1x + e**

In simple linear regression, the relationship between the predictor (x) and outcome (y) is modeled as a straight line with an intercept (b0) and a slope (b1), plus an error term (e).
</details>


### Question

**Options:**
- The slope
- The outcome variable
- The intercept
- The error term

<details>
<summary>Click to see Answer</summary>

**Correct Answer: The intercept**

b0 is the intercept, which is the expected value of y when x is equal to 0.
</details>


### Question
What does Pearson's correlation coefficient (r) measure?

**Options:**
- The strength and direction of any relationship
- The strength and direction of a linear relationship
- The slope of the regression line
- The proportion of variance explained

<details>
<summary>Click to see Answer</summary>

Correlation (r) specifically measures how well two variables follow a linear trend. It does not capture non-linear relationships well.
</details>


### Question

**Options:**
- 0 to 1
- -1 to 1
- -infinity to infinity
- 0 to infinity

<details>
<summary>Click to see Answer</summary>

**Correct Answer: -1 to 1**

</details>


### Question
What is a residual in linear regression?

**Options:**
- The sum of squared errors
- The difference between observed and predicted values
- The slope of the regression line
- The mean of the outcome variable

<details>
<summary>Click to see Answer</summary>

**Correct Answer: The difference between observed and predicted values**

A residual (e) is defined as e = y - y_hat, where y is the actual observed value and y_hat is the value predicted by the model.
</details>


### Question
What does R-squared (R2) represent?

**Options:**
- The slope of the line
- The correlation coefficient
- The proportion of variation in the outcome explained by the model
- The mean squared error

<details>
<summary>Click to see Answer</summary>

**Correct Answer: The proportion of variation in the outcome explained by the model**

R-squared, also known as the coefficient of determination, tells us what percentage of the total variability in the dependent variable is accounted for by the linear model.
</details>


### Question
If r = 0.8, what is the value of R-squared?

**Options:**
- 0.8
- 0.4
- 0.64
- 1.6

<details>
<summary>Click to see Answer</summary>

**Correct Answer: 0.64**

For simple linear regression, R-squared is the square of the correlation coefficient: R2 = r^2. 0.8^2 = 0.64.
</details>

### Question

**Options:**
- A linear relationship
- Non-constant variance (heteroscedasticity)
- An influential point
- A perfect fit

<details>
<summary>Click to see Answer</summary>

A funnel or fan shape in residuals indicates that the spread of the errors changes as the predictor variable changes, violating the assumption of constant variance.
</details>

### Question
What is the identity for Sum of Squares?

**Options:**
- SST = SSR + SSE
- SSR = SST + SSE
- SSE = SST + SSR
- SST = SSR - SSE

<details>
<summary>Click to see Answer</summary>

**Correct Answer: SST = SSR + SSE**

Total variability (SST) is decomposed into variability explained by regression (SSR) and unexplained residual variability (SSE).
</details>

### Question
In an indicator variable regression (binary predictor), what does the intercept represent?

**Options:**
- The mean of the group coded as 1
- The mean of the group coded as 0
- The difference between group means
- The overall mean of all data

<details>
<summary>Click to see Answer</summary>

**Correct Answer: The mean of the group coded as 0**

When x=0, the equation y = b0 + b1x simplifies to y = b0. Thus b0 is the average outcome for the reference group.
</details>


### Question
In an indicator variable regression, what does the slope (b1) represent?

**Options:**
- The mean of the group coded as 0
- The mean of the group coded as 1
- The difference in means between group 1 and group 0
- The standard deviation of group 1

<details>
<summary>Click to see Answer</summary>

**Correct Answer: The difference in means between group 1 and group 0**

The slope b1 represents the change in y for a 1-unit increase in x. Going from 0 to 1 is a 1-unit increase, so b1 is the difference in group averages.
</details>
