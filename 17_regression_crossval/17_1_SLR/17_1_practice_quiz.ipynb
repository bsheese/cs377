```python
import ipywidgets as widgets
from IPython.display import display, clear_output

all_quizzes = {
    "17_1_1: Two Paradigms": [
        {
            "question": "Why does Scikit-Learn require X to be 2D using .reshape(-1, 1) while Statsmodels accepts 1D arrays?",
            "options": [
                "Scikit-Learn is designed to handle multiple features uniformly, even with one predictor",
                "Statsmodels has better memory management for large arrays",
                "2D arrays are faster to process on modern computers",
                "1D arrays cannot store slope and intercept simultaneously"
            ],
            "answer": "Scikit-Learn is designed to handle multiple features uniformly, even with one predictor"
        },
        {
            "question": "Both Scikit-Learn and Statsmodels produce identical slope, intercept, and R² values. What does this convergence prove?",
            "options": [
                "Both libraries solve the same underlying mathematical problem",
                "Scikit-Learn copies code from Statsmodels internally",
                "Linear regression can only be computed one way",
                "The libraries use identical optimization algorithms"
            ],
            "answer": "Both libraries solve the same underlying mathematical problem"
        },
        {
            "question": "Why must we manually add an intercept with sm.add_constant(X) in Statsmodels but not in Scikit-Learn?",
            "options": [
                "Statsmodels treats the intercept as an explicit feature column in the X matrix",
                "Scikit-Learn does not believe intercepts are statistically important",
                "Statsmodels is incompatible with modern versions of NumPy",
                "Adding constants is faster in Statsmodels than elsewhere"
            ],
            "answer": "Statsmodels treats the intercept as an explicit feature column in the X matrix"
        },
        {
            "question": "A penguin with 0mm flipper length is predicted to weigh -5,872g. Why is this prediction nonsensical?",
            "options": [
                "Negative weight is physically impossible; extrapolating beyond data range is dangerous",
                "The model is fundamentally broken and unusable",
                "Penguins with small flippers are extremely rare and unmeasurable",
                "Statsmodels has a bug that produces negative predictions"
            ],
            "answer": "Negative weight is physically impossible; extrapolating beyond data range is dangerous"
        },
        {
            "question": "In the Palmer Penguins model, the intercept represents the expected body mass when flipper length is 0mm. Is this intercept meaningful?",
            "options": [
                "No—it's just a mathematical artifact outside the data range",
                "Yes—it represents the weight of a newborn penguin",
                "Yes—it is always the most important coefficient",
                "No—intercepts are never meaningful in linear regression"
            ],
            "answer": "No—it's just a mathematical artifact outside the data range"
        },
        {
            "question": "Why does the notebook verify the fitted slope against a closed-form formula?",
            "options": [
                "To prove both libraries implement the same mathematical solution correctly",
                "To ensure the computer's floating-point arithmetic is accurate",
                "To demonstrate that linear regression is more reliable than formulas",
                "To show which library is faster at computing slopes"
            ],
            "answer": "To prove both libraries implement the same mathematical solution correctly"
        },
        {
            "question": "Scikit-Learn's philosophy is 'prediction-focused' while Statsmodels is 'inference-focused.' What does this difference imply about priorities?",
            "options": [
                "Scikit-Learn emphasizes accuracy; Statsmodels emphasizes hypothesis testing and uncertainty",
                "Statsmodels makes predictions; Scikit-Learn only does theory",
                "Inference and prediction are unrelated concepts in practice",
                "Scikit-Learn cannot be used for statistical testing"
            ],
            "answer": "Scikit-Learn emphasizes accuracy; Statsmodels emphasizes hypothesis testing and uncertainty"
        },
        {
            "question": "When extracting the slope in Statsmodels, you use model.params[1]. Why is the slope at index 1, not index 0?",
            "options": [
                "Index 0 is the intercept; index 1 is the first (and only) predictor coefficient",
                "Statsmodels always starts indexing at 1, not 0",
                "Index 1 is reserved for slopes and index 0 for correlations",
                "The indexing is arbitrary and varies per Statsmodels version"
            ],
            "answer": "Index 0 is the intercept; index 1 is the first (and only) predictor coefficient"
        }
    ],
    "17_1_2: Statistical Significance": [
        {
            "question": "In bootstrap resampling, why must you resample (x, y) pairs together rather than x and y independently?",
            "options": [
                "Resampling pairs preserves the relationship between x and y observed in the data",
                "Independent resampling is faster but produces incorrect results",
                "Paired resampling ensures x and y have identical distributions",
                "The notebook author preferred this method for historical reasons"
            ],
            "answer": "Resampling pairs preserves the relationship between x and y observed in the data"
        },
        {
            "question": "The bootstrap SE (~1.49) nearly matches the Statsmodels SE (~1.54), even though they use completely different methods. What does this agreement suggest?",
            "options": [
                "Both methods are estimating the same underlying quantity correctly",
                "Bootstrap and parametric methods are mathematically identical",
                "The data was deliberately chosen to make them match",
                "Coincidence; you should not expect them to be similar"
            ],
            "answer": "Both methods are estimating the same underlying quantity correctly"
        },
        {
            "question": "A permutation test shuffles y-values randomly while keeping x fixed. What does this procedure simulate?",
            "options": [
                "A world where x and y have absolutely no relationship",
                "What would happen if you measured y incorrectly",
                "A bootstrapped sample from the true population",
                "The effect of adding measurement error to x"
            ],
            "answer": "A world where x and y have absolutely no relationship"
        },
        {
            "question": "In the permutation test with 5,000 shuffles, the observed slope of 50.15 never appeared. What is the empirical p-value?",
            "options": [
                "Less than 1/5000 = 0.0002 (effectively zero)",
                "Exactly 0.05 by definition",
                "50.15 / 5000 = 0.01003",
                "Cannot be computed without knowing the number of folds"
            ],
            "answer": "Less than 1/5000 = 0.0002 (effectively zero)"
        },
        {
            "question": "A p-value of < 0.001 means there's less than 0.1% chance of observing this slope if H₀ is true. Does this prove H₀ is false?",
            "options": [
                "No—small p-values indicate H₀ is implausible but don't prove it's false with 99.9% certainty",
                "Yes—p-values directly measure the probability that H₀ is true",
                "No—H₀ can only be tested using confidence intervals, not p-values",
                "Yes—p-values are equivalent to proof in statistics"
            ],
            "answer": "No—small p-values indicate H₀ is implausible but don't prove it's false with 99.9% certainty"
        },
        {
            "question": "Why do smaller sample sizes produce larger standard errors and wider confidence intervals?",
            "options": [
                "Less data means more uncertainty about the true population parameter",
                "Small samples have more outliers than large samples",
                "Standard error is inversely related to the number of predictors",
                "The formula for SE requires dividing by n-1"
            ],
            "answer": "Less data means more uncertainty about the true population parameter"
        },
        {
            "question": "In the coverage simulation, 95% of 1,000 intervals contained the true slope. Should you expect this to happen again with new data?",
            "options": [
                "Approximately yes—95% coverage reflects the proper calibration of the method",
                "No—the coverage will be different with new data",
                "Yes—coverage is guaranteed to be exactly 95% on all datasets",
                "Cannot answer without knowing the true parameter value"
            ],
            "answer": "Approximately yes—95% coverage reflects the proper calibration of the method"
        },
        {
            "question": "If you construct a 90% CI instead of 95%, the interval becomes wider or narrower?",
            "options": [
                "Narrower—less stringent confidence level allows a tighter range",
                "Wider—lower confidence requires more buffer",
                "Identical—confidence level doesn't affect interval width",
                "Cannot determine without computing the exact values"
            ],
            "answer": "Narrower—less stringent confidence level allows a tighter range"
        },
        {
            "question": "A regression has p < 0.001 (highly significant) but R² = 0.05 (weak effect). How do you reconcile these?",
            "options": [
                "A large sample makes even tiny slopes statistically significant",
                "The model has a hidden bug that inflates the p-value",
                "Significance and effect size are unrelated concepts",
                "High significance guarantees practical importance"
            ],
            "answer": "A large sample makes even tiny slopes statistically significant"
        },
        {
            "question": "The three CI methods (bootstrap percentile, formula, Statsmodels) all converge on [47.1, 53.2]. What does this suggest?",
            "options": [
                "All three methods are estimating the same underlying confidence interval correctly",
                "The three methods are mathematically identical",
                "The interval is guaranteed to contain the true parameter",
                "Bootstrap percentile intervals are the only valid method"
            ],
            "answer": "All three methods are estimating the same underlying confidence interval correctly"
        }
    ],
    "17_1_3: LINE Assumptions": [
        {
            "question": "In the Auto MPG data, the residual plot shows a U-shape. What does this reveal about the fit?",
            "options": [
                "The relationship is curved, not linear; displacement and mpg violate linearity",
                "The model has heteroscedasticity but is otherwise well-specified",
                "The data contains outliers that skew the results",
                "The residuals are perfectly normally distributed"
            ],
            "answer": "The relationship is curved, not linear; displacement and mpg violate linearity"
        },
        {
            "question": "A LOWESS smoother is overlaid on the residual plot. What does a curved LOWESS line indicate?",
            "options": [
                "Systematic deviation from zero suggests non-linear patterns in the data",
                "LOWESS always produces curved lines regardless of the data",
                "It shows that the model's errors are homoscedastic",
                "The residuals are perfectly symmetrically distributed"
            ],
            "answer": "Systematic deviation from zero suggests non-linear patterns in the data"
        },
        {
            "question": "What does a Durbin-Watson statistic of 2.0 indicate about the residuals?",
            "options": [
                "Values are independent; no autocorrelation detected",
                "Perfect positive autocorrelation between successive residuals",
                "The sample size is too small for valid testing",
                "Heteroscedasticity is present in the data"
            ],
            "answer": "Values are independent; no autocorrelation detected"
        },
        {
            "question": "Why is a Q-Q plot considered more informative than a histogram for detecting non-normality?",
            "options": [
                "Q-Q plots reveal deviations in the tails, where normality is most important",
                "Histograms cannot be used with regression residuals",
                "Q-Q plots automatically correct for non-normality",
                "Q-Q plots use the Jarque-Bera test internally"
            ],
            "answer": "Q-Q plots reveal deviations in the tails, where normality is most important"
        },
        {
            "question": "A 'funnel' shape in the residual plot (wider at one end) indicates what?",
            "options": [
                "Heteroscedasticity—error variance is not constant across fitted values",
                "The residuals are perfectly normally distributed",
                "The model has captured all nonlinear patterns",
                "Independence is violated due to time-series structure"
            ],
            "answer": "Heteroscedasticity—error variance is not constant across fitted values"
        },
        {
            "question": "The Central Limit Theorem makes inference 'robust' to mild non-normality at n > 30. What does robustness mean here?",
            "options": [
                "P-values and CIs remain approximately valid despite non-normal residuals",
                "Non-normality doesn't matter at all when n > 30",
                "The residuals automatically become normal in large samples",
                "Robustness means the model's predictions are unbiased"
            ],
            "answer": "P-values and CIs remain approximately valid despite non-normal residuals"
        },
        {
            "question": "Can you trust the regression slope even if LINE assumptions are violated?",
            "options": [
                "Yes—the slope estimate itself is unbiased, but SEs, CIs, and p-values are compromised",
                "No—violations invalidate everything about the model",
                "Only if you use robust standard errors",
                "Never, because OLS requires all assumptions to hold"
            ],
            "answer": "Yes—the slope estimate itself is unbiased, but SEs, CIs, and p-values are compromised"
        },
        {
            "question": "In time-series data, observations close in time tend to have similar values. Which LINE assumption might this violate?",
            "options": [
                "Independence—successive observations are correlated",
                "Linearity—time always creates nonlinear patterns",
                "Normality—time-series data is never normally distributed",
                "Equal Variance—time naturally creates heteroscedasticity"
            ],
            "answer": "Independence—successive observations are correlated"
        }
    ],
    "17_1_4: Influence & Leverage": [
        {
            "question": "Which combination makes a point truly dangerous for regression?",
            "options": [
                "High leverage AND large residual together",
                "High leverage with residual near zero",
                "Low leverage with large residual",
                "High correlation with the target"
            ],
            "answer": "High leverage AND large residual together"
        },
        {
            "question": "In the poisoned Ames dataset (10,000 sq ft, $500), why did removing this one point change the slope from 0.0997 to 0.1126?",
            "options": [
                "The point had high influence, pulling the regression line toward it",
                "Outliers always change the slope by the same magnitude",
                "The model's predictions were incorrect without retraining",
                "This change proves the model was fundamentally broken"
            ],
            "answer": "The point had high influence, pulling the regression line toward it"
        },
        {
            "question": "Cook's Distance combines leverage and residual disagreement. Why is this combination necessary?",
            "options": [
                "Only points with BOTH unusual x-value AND poor prediction are truly problematic",
                "Leverage and residuals are mathematically identical",
                "Cook's Distance is the only valid influence metric",
                "Combining them guarantees the point should be removed"
            ],
            "answer": "Only points with BOTH unusual x-value AND poor prediction are truly problematic"
        },
        {
            "question": "What is the primary difference between dropping influential data and using Huber regression?",
            "options": [
                "Huber down-weights outliers rather than deleting them entirely",
                "Huber regression only works with normally distributed residuals",
                "Dropping data is always preferable to using Huber",
                "Huber regression removes the need for any data cleaning"
            ],
            "answer": "Huber down-weights outliers rather than deleting them entirely"
        },
        {
            "question": "Is there an 'official' threshold for Cook's Distance above which you must remove a point?",
            "options": [
                "No—thresholds are guides; investigate outliers individually",
                "Yes—always remove points above 4/n",
                "Yes—always remove if Cook's D > 0.1",
                "No—cook's D is only meaningful for linear models"
            ],
            "answer": "No—thresholds are guides; investigate outliers individually"
        },
        {
            "question": "A data point is a correct entry (not an error) but has very high Cook's Distance. Should you drop it?",
            "options": [
                "Probably not—it's legitimate data that should be kept and explained",
                "Yes—influential points must always be removed",
                "Only if it improves model R-squared",
                "Yes—you must maintain equal sample sizes across groups"
            ],
            "answer": "Probably not—it's legitimate data that should be kept and explained"
        },
        {
            "question": "What is a key risk of iteratively removing points with high Cook's Distance?",
            "options": [
                "You might over-clean the data and remove legitimate variation",
                "Each removal makes the next point's Cook's D automatically increase",
                "You cannot use cross-validation after removing points",
                "The slope converges to zero after enough removals"
            ],
            "answer": "You might over-clean the data and remove legitimate variation"
        },
        {
            "question": "Instead of dropping an influential point, what are three alternative approaches?",
            "options": [
                "Use robust regression, report results with and without it, or investigate its cause",
                "Always use bootstrap or apply log transformations",
                "Increase the sample size until it's no longer extreme",
                "Switch to a different modeling technique like trees"
            ],
            "answer": "Use robust regression, report results with and without it, or investigate its cause"
        }
    ],
    "17_1_5: Transformations": [
        {
            "question": "The Gapminder data shows diminishing returns: each dollar of GDP predicts less life expectancy as GDP grows. What causes this curve?",
            "options": [
                "The derivative of log(x) is 1/x, steep when x is small, flat when x is large",
                "Wealthier countries report longer lifespans dishonestly",
                "Linear regression cannot handle relationship curvature",
                "The relationship becomes nonlinear past a certain GDP threshold"
            ],
            "answer": "The derivative of log(x) is 1/x, steep when x is small, flat when x is large"
        },
        {
            "question": "Why does a log transformation often fix BOTH linearity AND heteroscedasticity violations?",
            "options": [
                "Log compression affects both the mean structure and variance structure",
                "Transformation can only fix one violation at a time",
                "Heteroscedasticity doesn't exist in log-transformed data",
                "Linearity automatically implies equal variance"
            ],
            "answer": "Log compression affects both the mean structure and variance structure"
        },
        {
            "question": "In the Gapminder model, the coefficient for log(GDP) is 5.8. How do you interpret this slope?",
            "options": [
                "A 1% increase in GDP is associated with 5.8/100 ≈ 0.058 additional years of life",
                "Doubling GDP adds 5.8 years of life",
                "A log unit increase in GDP adds 5.8 years exactly",
                "GDP explains 5.8% of the variation in life expectancy"
            ],
            "answer": "A 1% increase in GDP is associated with 5.8/100 ≈ 0.058 additional years of life"
        },
        {
            "question": "In a log-level model (log(y) = β₀ + β₁*x), what does β₁ = 0.05 mean in plain English?",
            "options": [
                "A 1-unit increase in x corresponds to ~5% increase in y",
                "A 5% increase in y requires a 1-unit increase in x",
                "y doubles when x increases by 0.05 units",
                "x and y have a 5% correlation"
            ],
            "answer": "A 1-unit increase in x corresponds to ~5% increase in y"
        },
        {
            "question": "What does Box-Cox transformation do, and what does λ = 0 suggest?",
            "options": [
                "Finds optimal λ to maximize normality; λ=0 suggests a log transformation",
                "Applies the same transformation to all features equally",
                "Ensures the target variable has zero mean and unit variance",
                "Automatically removes outliers from the dataset"
            ],
            "answer": "Finds optimal λ to maximize normality; λ=0 suggests a log transformation"
        },
        {
            "question": "After log-transforming the target, your model achieves R² = 0.65 (vs. 0.46 before). Is this a real improvement?",
            "options": [
                "The R² values are incomparable across different y-scales; compare residual patterns instead",
                "Yes—higher R² always means better predictions",
                "No—log transformations artificially inflate R²",
                "Only if the improvement is greater than 20%"
            ],
            "answer": "The R² values are incomparable across different y-scales; compare residual patterns instead"
        },
        {
            "question": "After fitting a log-transformed model, predictions are in log-scale. How do you back-transform?",
            "options": [
                "Exponentiate the prediction; use Jensen's correction: exp(pred) * exp(σ²/2)",
                "Simply take the exponential of each prediction",
                "Divide the prediction by 2.718 (Euler's number)",
                "Back-transformation is impossible for log models"
            ],
            "answer": "Exponentiate the prediction; use Jensen's correction: exp(pred) * exp(σ²/2)"
        },
        {
            "question": "Why is Jensen's correction important when back-transforming log-level predictions?",
            "options": [
                "Without it, exp(predicted log value) underestimates the true expected value",
                "It ensures the back-transformed predictions are always positive",
                "It prevents the model from ever making negative predictions",
                "It is only necessary for Box-Cox transformations, not log"
            ],
            "answer": "Without it, exp(predicted log value) underestimates the true expected value"
        },
        {
            "question": "How do the three log models (level-log, log-level, log-log) differ in interpretation?",
            "options": [
                "They describe percentage vs. absolute changes in different combinations",
                "They are mathematically identical despite different notation",
                "Only one of them can be used at a time in practice",
                "They are chosen randomly; interpretation doesn't depend on choice"
            ],
            "answer": "They describe percentage vs. absolute changes in different combinations"
        },
        {
            "question": "Why might a transformation mask an underlying problem instead of fixing it?",
            "options": [
                "A transformation can hide that the functional form is genuinely non-linear",
                "Transformations always reveal underlying data problems",
                "Only inappropriate transformations can mask issues",
                "Masking issues is impossible with modern regression tools"
            ],
            "answer": "A transformation can hide that the functional form is genuinely non-linear"
        }
    ],
    "17_1_6: Generalization": [
        {
            "question": "Why is training error always optimistic about model performance?",
            "options": [
                "The model is fit to minimize training error; it hasn't seen test data",
                "Training error uses different metrics than test error",
                "Testing on new data always produces lower R² by definition",
                "Models are programmed to report optimistic training errors"
            ],
            "answer": "The model is fit to minimize training error; it hasn't seen test data"
        },
        {
            "question": "In the synthetic sin(x) experiment, what happens to test R² as polynomial degree increases from 1 to 20?",
            "options": [
                "Test R² rises initially, peaks at moderate degree, then falls (U-shaped pattern)",
                "Test R² continuously rises with degree",
                "Test R² stays constant regardless of complexity",
                "Test R² follows training error exactly"
            ],
            "answer": "Test R² rises initially, peaks at moderate degree, then falls (U-shaped pattern)"
        },
        {
            "question": "At degree 20 polynomial, train R² ≈ 0.98 and test R² ≈ -0.5. Why is negative test R² possible?",
            "options": [
                "Model predictions are worse than simply predicting the mean on test data",
                "Negative R² indicates the model is catastrophically broken",
                "The test set was corrupted or mislabeled",
                "Polynomial degrees above 10 always produce negative R²"
            ],
            "answer": "Model predictions are worse than simply predicting the mean on test data"
        },
        {
            "question": "What does a large gap between train R² and test R² primarily indicate?",
            "options": [
                "Overfitting—model has memorized training data noise",
                "Your test set is fundamentally different from training",
                "The model is underfitting and needs more features",
                "Data leakage from test set into training process"
            ],
            "answer": "Overfitting—model has memorized training data noise"
        },
        {
            "question": "In the bias-variance decomposition, what is the U-shaped curve describing?",
            "options": [
                "Test error vs. model complexity: bias decreases, variance increases, sum minimized at sweet spot",
                "R-squared vs. training set size, showing when enough data is collected",
                "Train vs. test error plotted against the number of iterations",
                "Feature importance scores ranked from highest to lowest"
            ],
            "answer": "Test error vs. model complexity: bias decreases, variance increases, sum minimized at sweet spot"
        },
        {
            "question": "Should you always use nested cross-validation instead of a single train/test split?",
            "options": [
                "Not always—simple splits suffice for quick iteration; nested CV for final reporting",
                "Always—nested CV is mandatory for any model",
                "Never—it wastes computation; a single split is statistically equivalent",
                "Only if you have fewer than 100 samples"
            ],
            "answer": "Not always—simple splits suffice for quick iteration; nested CV for final reporting"
        },
        {
            "question": "In a learning curve, what do the shaded bands around each line represent?",
            "options": [
                "Standard deviation or confidence interval of scores across folds",
                "The range of possible predictions for each sample",
                "Error bars showing measurement uncertainty",
                "Missing data that were excluded from training"
            ],
            "answer": "Standard deviation or confidence interval of scores across folds"
        },
        {
            "question": "If learning curve shows train R² = 0.95 and test R² = 0.70 at max training size, what should you try?",
            "options": [
                "Increase regularization or reduce model complexity",
                "Collect more training data to improve test performance",
                "Switch to a simpler model like linear regression",
                "Abandon the problem as unsolvable"
            ],
            "answer": "Increase regularization or reduce model complexity"
        },
        {
            "question": "If both train and test R² converge at 0.55 with more data, what does this indicate?",
            "options": [
                "Underfitting—the model is too simple; increasing complexity may help",
                "Overfitting—adding more data would improve test performance",
                "Perfect model fit at the maximum achievable performance",
                "The data is corrupted or of very poor quality"
            ],
            "answer": "Underfitting—the model is too simple; increasing complexity may help"
        }
    ]
}

class QuizApp:
    def __init__(self, quiz_name, data):
        self.quiz_name = quiz_name
        self.data = data
        self.current_index = 0
        self.score = 0
        self.user_answers = []

        self.header = widgets.HTML(value="<h2>" + quiz_name + "</h2>")
        self.progress = widgets.HTML(value="")
        self.question_text = widgets.HTML(value="")
        self.options = widgets.RadioButtons(options=[], layout={"width": "max-content"})
        self.submit_btn = widgets.Button(description="Submit Answer", button_style="primary")
        self.next_btn = widgets.Button(description="Next Question", button_style="info", disabled=True)
        self.back_btn = widgets.Button(description="Back to Menu", button_style="warning")
        self.feedback = widgets.HTML(value="")

        self.submit_btn.on_click(self.handle_submit)
        self.next_btn.on_click(self.handle_next)
        self.back_btn.on_click(self.handle_back)

        self.render_question()

    def render_question(self):
        clear_output(wait=True)
        q = self.data[self.current_index]
        self.progress.value = "<b>Question " + str(self.current_index + 1) + " of " + str(len(self.data)) + "</b>"
        self.question_text.value = "<p style='font-size:14px;'>" + q["question"] + "</p>"
        self.options.options = q["options"]
        self.options.value = None
        self.submit_btn.disabled = False
        self.next_btn.disabled = True
        self.feedback.value = ""

        display(self.header, self.progress, self.question_text, self.options,
                widgets.HBox([self.submit_btn, self.next_btn]), self.feedback)

    def handle_submit(self, b):
        if self.options.value is None:
            self.feedback.value = "<span style='color:red;'>Please select an option!</span>"
            return

        correct_answer = self.data[self.current_index]["answer"]
        if self.options.value == correct_answer:
            self.score += 1
            self.feedback.value = "<span style='color:green; font-weight:bold;'>Correct!</span>"
        else:
            self.feedback.value = "<span style='color:red; font-weight:bold;'>Incorrect.</span><br>The correct answer was: <b>" + correct_answer + "</b>"

        self.submit_btn.disabled = True
        self.next_btn.disabled = False

    def handle_next(self, b):
        self.current_index += 1
        if self.current_index < len(self.data):
            self.render_question()
        else:
            self.show_results()

    def handle_back(self, b):
        show_menu()

    def show_results(self):
        clear_output()
        percent = (self.score / len(self.data)) * 100
        result_html = """<div style="border: 2px solid #0366d6; padding: 20px; border-radius: 10px;">
            <h2>Quiz Complete!</h2>
            <p style="font-size:16px;">Your Final Score: <b>"" + str(self.score) + " / " + str(len(self.data)) + """</b></p>
            <p style="font-size:20px; color:#0366d6;"><b>Percentage: "" + "{:.1f}".format(percent) + "%</b></p>
        </div>"""
        display(widgets.HTML(result_html), self.back_btn)

def show_menu():
    clear_output(wait=True)
    header = widgets.HTML(value="<h2>17_1 Simple Linear Regression -- Practice Quizzes</h2>")
    instruction = widgets.HTML(value="<p>Select a notebook quiz below to begin.</p>")
    quiz_selector = widgets.Dropdown(
        options=list(all_quizzes.keys()),
        description="Select Quiz:",
        layout={"width": "50%"}
    )
    start_btn = widgets.Button(description="Start Quiz", button_style="success")

    def on_start(b):
        selected = quiz_selector.value
        QuizApp(selected, all_quizzes[selected])

    start_btn.on_click(on_start)
    display(header, instruction, quiz_selector, start_btn)

show_menu()
```
