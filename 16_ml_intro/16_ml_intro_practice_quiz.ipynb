```python
import ipywidgets as widgets
from IPython.display import display, clear_output

all_quizzes = {
    "16_1: Intro to ML": [
        {
            "question": "What is the fundamental difference between traditional programming and machine learning?",
            "options": [
                "Traditional programming learns rules from data; ML executes predefined rules",
                "ML learns rules from data; traditional programming executes predefined rules",
                "ML is faster; traditional programming is more accurate",
                "Traditional programming uses Python; ML requires specialized hardware"
            ],
            "answer": "ML learns rules from data; traditional programming executes predefined rules"
        },
        {
            "question": "In supervised learning, what are 'features' and what is the 'target'?",
            "options": [
                "Features are the labels we predict; the target is the input data",
                "Features are input variables used for prediction; the target is the output we predict",
                "Features are training examples; the target is the test set",
                "Features are model parameters; the target is the loss function"
            ],
            "answer": "Features are input variables used for prediction; the target is the output we predict"
        },
        {
            "question": "How does a classification problem differ from a regression problem?",
            "options": [
                "Classification predicts continuous values; regression predicts discrete categories",
                "Classification predicts discrete categories; regression predicts continuous values",
                "Classification requires more data than regression",
                "Regression always outperforms classification on structured data"
            ],
            "answer": "Classification predicts discrete categories; regression predicts continuous values"
        },
        {
            "question": "Why does a KNN classifier require StandardScaler before training?",
            "options": [
                "StandardScaler improves model accuracy by removing outliers",
                "KNN measures distance between points; unscaled features with large ranges dominate the distance",
                "KNN cannot process negative numbers without scaling",
                "StandardScaler converts categorical features to numeric values"
            ],
            "answer": "KNN measures distance between points; unscaled features with large ranges dominate the distance"
        },
        {
            "question": "You add a random noise feature to a KNN model. What happens to the decision boundary?",
            "options": [
                "The boundary becomes smoother and more accurate",
                "The boundary becomes noisier and less accurate because meaningless distance is added",
                "The boundary is unaffected because KNN ignores irrelevant features",
                "The model automatically detects and removes the noise feature"
            ],
            "answer": "The boundary becomes noisier and less accurate because meaningless distance is added"
        },
        {
            "question": "knn.predict_proba() returns probabilities for each class. When is this more useful than knn.predict()?",
            "options": [
                "When the model needs to run faster than predict()",
                "When you need confidence levels or want to adjust the decision threshold",
                "When the training data has missing values",
                "When features are categorical rather than continuous"
            ],
            "answer": "When you need confidence levels or want to adjust the decision threshold"
        },
        {
            "question": "A regression model trained on houses between 1,000–3,000 sqft predicts a 10,000 sqft house. Why is this risky?",
            "options": [
                "The model will refuse to make predictions outside its training range",
                "Extrapolating beyond the training range assumes the linear trend continues, which may be wrong",
                "Regression models cannot process numbers above 3,000",
                "The prediction will be identical to the mean of the training data"
            ],
            "answer": "Extrapolating beyond the training range assumes the linear trend continues, which may be wrong"
        },
        {
            "question": "Which of the following is an example of unsupervised learning?",
            "options": [
                "Predicting whether an email is spam using labeled spam/not-spam examples",
                "Grouping customers by purchasing patterns without any predefined categories",
                "Training an agent to play chess by rewarding winning moves",
                "Predicting tomorrow's temperature from historical weather data"
            ],
            "answer": "Grouping customers by purchasing patterns without any predefined categories"
        },
        {
            "question": "The 'garbage in, garbage out' principle refers to what?",
            "options": [
                "Models should be deleted if they produce incorrect predictions",
                "Poor quality training data produces unreliable models regardless of algorithm sophistication",
                "Code errors in data processing always cause model crashes",
                "Outliers must be removed before any model can be trained"
            ],
            "answer": "Poor quality training data produces unreliable models regardless of algorithm sophistication"
        }
    ],
    "16_3: NumPy Arrays": [
        {
            "question": "You assign 9.99 to position 0 of an integer NumPy array. The stored value becomes 9. Why?",
            "options": [
                "NumPy rounds all floats to the nearest integer",
                "NumPy truncates the decimal because the array enforces integer dtype",
                "9.99 is converted to its binary representation and rounded up",
                "NumPy raises a TypeError for mixed-type assignments"
            ],
            "answer": "NumPy truncates the decimal because the array enforces integer dtype"
        },
        {
            "question": "Why is `mylist + 4` a TypeError but `myarray + 4` works correctly?",
            "options": [
                "Arrays support element-wise arithmetic; lists interpret + as concatenation only",
                "Lists cannot store the number 4 without type conversion",
                "NumPy redefines + to mean append for all Python objects",
                "Lists require a for-loop; arrays do not use Python operators"
            ],
            "answer": "Arrays support element-wise arithmetic; lists interpret + as concatenation only"
        },
        {
            "question": "Why should you avoid building an array with repeated np.append() calls inside a loop?",
            "options": [
                "np.append() is deprecated in modern NumPy versions",
                "Each call copies the entire array, making the loop O(n²) overall",
                "np.append() modifies the original array and causes index errors",
                "Loops cannot call NumPy functions more than once"
            ],
            "answer": "Each call copies the entire array, making the loop O(n²) overall"
        },
        {
            "question": "You want songs with tempo > 120 AND energy > 0.6. Why must you use & instead of Python's 'and'?",
            "options": [
                "'and' is reserved for boolean scalars; & applies element-wise to boolean arrays",
                "& is faster than 'and' for large arrays",
                "'and' returns an integer array instead of a boolean array",
                "NumPy does not recognize Python keywords inside array expressions"
            ],
            "answer": "'and' is reserved for boolean scalars; & applies element-wise to boolean arrays"
        },
        {
            "question": "np.linspace(0, 1, 5) produces [0.0, 0.25, 0.5, 0.75, 1.0]. What does np.arange(0, 1, 0.25) produce?",
            "options": [
                "[0.0, 0.25, 0.5, 0.75, 1.0] — five elements including endpoint",
                "[0.0, 0.25, 0.5, 0.75] — four elements excluding endpoint",
                "[0.25, 0.5, 0.75, 1.0] — excludes start but includes endpoint",
                "[0.0, 0.25, 0.5, 0.75, 1.0, 1.25] — continues past 1.0 by one step"
            ],
            "answer": "[0.0, 0.25, 0.5, 0.75] — four elements excluding endpoint"
        },
        {
            "question": "arr.reshape(-1, 1) converts a 1D array to 2D. Why does scikit-learn require this shape?",
            "options": [
                "Scikit-learn uses 2D arrays to represent a feature matrix with rows=samples, cols=features",
                "Scikit-learn cannot process arrays with fewer than 2 dimensions",
                "2D arrays store more decimal precision than 1D arrays",
                "reshape(-1, 1) normalizes the data to unit variance"
            ],
            "answer": "Scikit-learn uses 2D arrays to represent a feature matrix with rows=samples, cols=features"
        },
        {
            "question": "np.random.seed(42) is called before generating training data. What would happen without it?",
            "options": [
                "The array would contain only zeros until explicitly populated",
                "The random data would change each run, making results non-reproducible",
                "NumPy would raise an error when generating random numbers",
                "The generated distribution would no longer be normal"
            ],
            "answer": "The random data would change each run, making results non-reproducible"
        },
        {
            "question": "An array has shape (100,). You reshape it to (-1, 5). What is the resulting shape?",
            "options": [
                "(5, 100) — the -1 is replaced by 100",
                "(20, 5) — NumPy infers 20 rows to make 100 elements fit",
                "(100, 5) — the original size is preserved as rows",
                "(-1, 5) — NumPy cannot infer the -1 dimension here"
            ],
            "answer": "(20, 5) — NumPy infers 20 rows to make 100 elements fit"
        }
    ],
    "16_4: Pandas Series": [
        {
            "question": "series.loc['apple':'cherry'] includes 'cherry' in the result. series.iloc[1:3] excludes position 3. Why?",
            "options": [
                ".loc uses inclusive slicing for label-based access; .iloc uses exclusive slicing like Python lists",
                ".loc and .iloc behave identically; the difference is a bug in this example",
                ".iloc includes the endpoint if the index is a string",
                ".loc excludes the endpoint when labels are alphabetically ordered"
            ],
            "answer": ".loc uses inclusive slicing for label-based access; .iloc uses exclusive slicing like Python lists"
        },
        {
            "question": "A Series has two entries with the label 'apple'. What does series.loc['apple'] return?",
            "options": [
                "Only the first 'apple' entry by default",
                "A Series containing both 'apple' entries",
                "A KeyError because duplicate labels are invalid",
                "The last 'apple' entry since it overwrites the first"
            ],
            "answer": "A Series containing both 'apple' entries"
        },
        {
            "question": "You write series_copy = series. Later, series_copy['apple'] = 999 also changes series. Why?",
            "options": [
                "Assignment creates an alias pointing to the same object, not an independent copy",
                "Pandas Series are immutable, so changes propagate automatically",
                "The value 999 overwrites the original Series because it's out of range",
                "Labels in Series always share memory across assignments"
            ],
            "answer": "Assignment creates an alias pointing to the same object, not an independent copy"
        },
        {
            "question": "series.sort_values() returns a new Series. Why does the notebook warn against inplace=True?",
            "options": [
                "inplace=True is deprecated and will be removed in future Pandas versions",
                "inplace=True can create subtle bugs when mixed with chained operations or views",
                "inplace=True is slower than returning a new Series",
                "inplace=True resets the index, which destroys the original label information"
            ],
            "answer": "inplace=True can create subtle bugs when mixed with chained operations or views"
        },
        {
            "question": "You want states with populations between 5 million and 15 million. Which code is correct?",
            "options": [
                "series.loc[series > 5e6 and series < 15e6]",
                "series.loc[(series > 5e6) & (series < 15e6)]",
                "series.loc[series > 5e6 | series < 15e6]",
                "series[(5e6 < series < 15e6)]"
            ],
            "answer": "series.loc[(series > 5e6) & (series < 15e6)]"
        },
        {
            "question": "series.str.split().str[0] extracts the first word from each string. Why is .str[0] needed instead of [0]?",
            "options": [
                ".str.split() returns a Series of lists; .str[0] extracts element 0 from each list element-wise",
                "[0] would return the entire first row of the Series",
                ".str[0] is faster because it uses vectorized operations",
                "[0] raises a TypeError when applied to string Series"
            ],
            "answer": ".str.split() returns a Series of lists; .str[0] extracts element 0 from each list element-wise"
        },
        {
            "question": "The cumulative sum of sorted state populations is used to find which states account for 50% of the US population. Why must you sort first?",
            "options": [
                "Unsorted data cannot be summed with .cumsum()",
                "Sorting by descending population lets you find the minimum number of largest states that reach the threshold",
                "Boolean masking requires sorted data to function correctly",
                ".cumsum() resets to zero if it encounters an unsorted value"
            ],
            "answer": "Sorting by descending population lets you find the minimum number of largest states that reach the threshold"
        }
    ],
    "16_5: Pandas DataFrames": [
        {
            "question": "Why must you use &, |, and ~ instead of Python's 'and', 'or', 'not' with DataFrame boolean masks?",
            "options": [
                "'and'/'or' are keywords reserved for if-statements and cannot appear in expressions",
                "'and'/'or' operate on the whole Series as a single truth value; &/| operate element-wise",
                "Pandas requires uppercase operators AND, OR, NOT for boolean operations",
                "'and'/'or' produce integers instead of boolean arrays"
            ],
            "answer": "'and'/'or' operate on the whole Series as a single truth value; &/| operate element-wise"
        },
        {
            "question": "df.loc[1973, 'artist'] returns a single value. df.loc[[1973], 'artist'] returns a Series. Why?",
            "options": [
                "Wrapping the row label in a list signals 'return a DataFrame/Series' rather than a scalar",
                "Double brackets always return two copies of the same value",
                "This is a Pandas bug that was fixed in newer versions",
                "The list syntax applies the selection to the column axis instead"
            ],
            "answer": "Wrapping the row label in a list signals 'return a DataFrame/Series' rather than a scalar"
        },
        {
            "question": "When copying multiple columns, the notebook uses .values: df[['new1','new2']] = df[['old1','old2']].values. Why is .values needed?",
            "options": [
                ".values prevents Pandas from trying to align on index labels, avoiding NaN results",
                ".values converts the DataFrame to a faster integer format",
                ".values is required when the columns have different data types",
                ".values drops duplicate index entries before assignment"
            ],
            "answer": ".values prevents Pandas from trying to align on index labels, avoiding NaN results"
        },
        {
            "question": "Searching for the Sage family with df['Name'].str.contains('Sage') also returns 'Sagesser'. How would you fix this?",
            "options": [
                "Use str.startswith('Sage ') or str.contains(' Sage') to require a word boundary",
                "Use str.find('Sage') == 0 to match only at the start of the string",
                "Append a wildcard: str.contains('Sage*')",
                "Filter by name length after applying contains"
            ],
            "answer": "Use str.startswith('Sage ') or str.contains(' Sage') to require a word boundary"
        },
        {
            "question": "df.drop(columns=['col1']) returns a new DataFrame. When would using inplace=True be preferable?",
            "options": [
                "When the DataFrame is too large to copy into memory",
                "When you want to permanently modify the DataFrame without creating an extra variable",
                "inplace=True is always preferable because it is faster",
                "Only when dropping rows, not columns"
            ],
            "answer": "When you want to permanently modify the DataFrame without creating an extra variable"
        },
        {
            "question": "df.info() shows Non-Null Count for each column. Why should this be the first thing you check on a new dataset?",
            "options": [
                "Non-Null Count shows which columns have the most unique values",
                "Missing values can cause silent errors in operations and biased results if not handled",
                "Columns with missing values must be removed before any analysis",
                "Non-Null Count determines the correct dtypes for each column"
            ],
            "answer": "Missing values can cause silent errors in operations and biased results if not handled"
        },
        {
            "question": "You want to add an 'age_group' column: 'child' if Age < 18, 'adult' otherwise. What is the correct approach?",
            "options": [
                "df['age_group'] = 'child' if df['Age'] < 18 else 'adult'",
                "Use two boolean masks: assign 'child' where mask is True, 'adult' elsewhere",
                "df['age_group'] = df['Age'].apply(lambda x: 'child' if x < 18 else 'adult')",
                "Both the second and third options are correct; the first option raises a ValueError"
            ],
            "answer": "Both the second and third options are correct; the first option raises a ValueError"
        }
    ]
}

class QuizApp:
    def __init__(self, quiz_name, data):
        self.quiz_name = quiz_name
        self.data = data
        self.current_index = 0
        self.score = 0

        self.header = widgets.HTML(value="<h2>" + quiz_name + "</h2>")
        self.progress = widgets.HTML(value="")
        self.question_text = widgets.HTML(value="")
        self.options = widgets.RadioButtons(options=[], layout={"width": "max-content"})
        self.submit_btn = widgets.Button(description="Submit Answer", button_style="primary")
        self.next_btn = widgets.Button(description="Next Question", button_style="info", disabled=True)
        self.back_btn = widgets.Button(description="Back to Menu", button_style="warning")
        self.feedback = widgets.HTML(value="")

        self.submit_btn.on_click(self.handle_submit)
        self.next_btn.on_click(self.handle_next)
        self.back_btn.on_click(self.handle_back)
        self.render_question()

    def render_question(self):
        clear_output(wait=True)
        q = self.data[self.current_index]
        self.progress.value = "<b>Question " + str(self.current_index + 1) + " of " + str(len(self.data)) + "</b>"
        self.question_text.value = "<p style='font-size:14px;'>" + q["question"] + "</p>"
        self.options.options = q["options"]
        self.options.value = None
        self.submit_btn.disabled = False
        self.next_btn.disabled = True
        self.feedback.value = ""
        display(self.header, self.progress, self.question_text, self.options,
                widgets.HBox([self.submit_btn, self.next_btn]), self.feedback)

    def handle_submit(self, b):
        if self.options.value is None:
            self.feedback.value = "<span style='color:red;'>Please select an option!</span>"
            return
        correct = self.data[self.current_index]["answer"]
        if self.options.value == correct:
            self.score += 1
            self.feedback.value = "<span style='color:green; font-weight:bold;'>Correct!</span>"
        else:
            self.feedback.value = "<span style='color:red; font-weight:bold;'>Incorrect.</span><br>The correct answer was: <b>" + correct + "</b>"
        self.submit_btn.disabled = True
        self.next_btn.disabled = False

    def handle_next(self, b):
        self.current_index += 1
        if self.current_index < len(self.data):
            self.render_question()
        else:
            self.show_results()

    def handle_back(self, b):
        show_menu()

    def show_results(self):
        clear_output()
        pct = (self.score / len(self.data)) * 100
        html = """<div style="border: 2px solid #0366d6; padding: 20px; border-radius: 10px;">
            <h2>Quiz Complete!</h2>
            <p style="font-size:16px;">Your Final Score: <b>""" + str(self.score) + " / " + str(len(self.data)) + """</b></p>
            <p style="font-size:20px; color:#0366d6;"><b>Percentage: """ + "{:.1f}".format(pct) + "%</b></p></div>"""
        display(widgets.HTML(html), self.back_btn)

def show_menu():
    clear_output(wait=True)
    header = widgets.HTML(value="<h2>16 — Intro to ML, NumPy, and Pandas — Practice Quizzes</h2>")
    instruction = widgets.HTML(value="<p>Select a notebook quiz below to begin.</p>")
    quiz_selector = widgets.Dropdown(options=list(all_quizzes.keys()), description="Select Quiz:", layout={"width": "50%"})
    start_btn = widgets.Button(description="Start Quiz", button_style="success")

    def on_start(b):
        QuizApp(quiz_selector.value, all_quizzes[quiz_selector.value])

    start_btn.on_click(on_start)
    display(header, instruction, quiz_selector, start_btn)

show_menu()
```
