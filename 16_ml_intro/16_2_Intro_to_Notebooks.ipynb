## What is a Notebook?

What you are reading right now is called a 'notebook' (also called a 'Jupyter notebook' or a 'Colab notebook'). It is a mix of text and code that is popular in data science. You can code in this notebook and make it do pretty much anything you like.

The way text and code are mixed is through the use of text blocks and code blocks. This is a text block. It does some basic formatting things for text, but it's intentionally pretty simple. If you double click this text, you'll be able to see how to edit, copy, or delete this text block. We will occasionally ask you to edit a text block in response to a question, but mostly we will be asking you to write Python code in code blocks.

### Code Blocks

The block below this (with the `[ ]   1`) is a code block. Go ahead and type the following into the block below: `print('Hello World')`

Once you've typed that out, click on the circle with the triangle in it on the left side of the code block. That will 'run' the program, meaning that you've submitted the program to a computer running Python. Python read the instructions and then executed them. If everything went well, the output of the program appeared below the cell.

Pro Tip: You can typically execute the cell that currently has focus by using Ctrl + Enter, or SHIFT + Enter. I do this without thinking. It is a good shortcut and is much faster than clicking the arrow. 



#### Order of Execution Matters
Notebooks are good for writing short programs in chunks and then displaying their output.

Notice that each time you run a cell, a little number appears in the `[ ]` space in the code cell. This indicates how many cells you have run, and in what order. Code run in one cell can affect how code in another cell works. They are not independent of one another. We will talk about this more later.  For now, you should remember that you need to run cells to have Python run your code, and the order that you run the cells in can sometimes make a difference.

We will typically write, and run, code cells starting at the top of the notebook and working our way down. However, that's not a requirement, we are allowed to run cells in any order.

```python

# here's a code cell full of imports
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# here's a separate code cell that will only run if the previous code cell has been executed
data = [25, 37, 98, np.Nan, 112]
months = ['January', 'February', 'March', 'April', 'May']
df = pd.DataFrame(data, index=months, columns=['Sales'])

sns.lineplot(data=df, x=df.index, y='Sales')
plt.title('Sales Over Time')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

### Text Cells

You can also create and edit text cells. Text cells don't run code, but when you 'run' them (in Google Colab you just hit escape or click outside the text cell when you are done), formatting is applied to the text and it is displayed. We won't often ask you to make text cells in class, but it will come up on projects.

The formatting rules that the text cells use are called 'Markdown'. Most of the Markdown formatting options are accessible in Colab through the menu at the top of each text cell.
